Crea la función `sum_even` que recibe una lista de números y devuelve la suma de los números pares.

Recuerda los 4 elementos de la iteración.

````python
def sum_even(list_number):
    sum = 0
    index = 0
    while index < len(list_number):
        if list_number[index] % 2 ==0:
            sum = sum + list_number[index]
        index = index + 1
    return sum

sum_even([1,2,3,4,5,6,7,8,9])
````

## (3)

Crea la función `less_than(numbers, n)` que recibe una lista de números y un número n. Devuelve la lista de aquellos números que son menores o iguales a n.

1. ¿Qué tipo de función es (compresor, selector o conversor)?
1. ¿Qué debería de devolver cuando recibe una lista vacía?
2. Si te ayuda, impleméntala primero con `for` e identifica claramente los 4 elementos de la iteración.

````python
def less_than(numbers,n):
    list_numbers = []
    index = 0
    while index < len(numbers):
        if numbers[index] <= n:
            list_numbers.append(numbers[index])
        index = index + 1
    return list_numbers

less_than([4,5,6,8,7,9,2,1,5,4,20,25,26],8)
````

## (5)

Crea la función `make_list(start, end, step)` que recibe 3 números y devuelve una lista que se genera de la siguiente manera:

1. Se empieza en `start`
2. Cada elemento es el anterior + `step`
3. Se para cuando se alcanza `end`
4. Si `end` es 0 o negativo, se devuelve la lista vacía
5. Si `end` es menor que `start`, se devuelve la lista vacía

Por ejemplo: 

`make_list(1, 10, 2) --> [1, 3, 5, 7, 9]`

`make_list(1, 10, 0) --> []`

`make_list(10, 10, 2) --> [10]`

`make_list[10, 4, 3] --> []`

**Pistas:**

1. Identifica los 4 elementos de la iteración
2. Valida primero los parámetros y si son erroneos (casos 4 y 5) devuelve la lista vacía.
3. Declara tus variable snada más empezar y un sólo return al final de la función.

````python
def make_list(start,end,step):
    list_new_number = []
    if step > 0:   
        while start < end:
            list_new_number.append(start)
            start = start + step
        if start == end:
            list_new_number.append(start)

    return list_new_number
````

````python
print(make_list(1, 10, 2))
print(make_list(1, 10, 0))
````

````python
assert make_list(1, 10, 2) == [1, 3, 5, 7, 9]

assert make_list(1, 10, 0) == []

assert make_list(10, 10, 2) == [10]

assert make_list(10, 4, 3) == []
````

## (9)

La siguiente función, que recibe una lista y devuelve `True` si alguno de sus elementos es cadena, contiene 1 error. Identifícalo y corrígelo.


````python
def has_string(elements):
    index=0
    found = False
    while not found and index < len(elements):
        if type(elements[index]) == str:
            found = True
        
        index = index + 1
    return found
            
````

````python
print(has_string(['hola','mundo','programador','desde','python']))
print(has_string([5,8,9,8,7,3,2]))
````

Crea la función `has_list` que recibe una lista y devuelve `True` sólo si alguno de sus elementos es también una lista 

````python


def has_list(list_elements):
    index=0    
    bolean=False
    while index < len(list_elements):
        if isinstance(list_elements[index],list):
            bolean = True
            break
        index = index + 1
    return bolean


````

````python
print(has_list([[
    [1,2]],[3,4],[5,6]]))
print(has_list([1,2,3,4]))
print(has_list([]))
print(has_list(['hola','mundo']))
print(has_list([['mundo'],['hola'],['programados']]))
````

1. ¿En qué se parece a la función `reverse`?

````python
def reverse_str(elements):
    index = len(elements) - 1
    reversed = ''
    while index >= 0:
        reversed = reversed + elements[index]  
        index = index - 1
    return reversed
````
