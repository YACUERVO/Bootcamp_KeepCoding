```python
from math import sqrt
def is_prime(number):
    bolean = True       
    if number <= 1:
        bolean = False
    else:
        for num in range(2, int(sqrt(number)) + 1):
            if number % num == 0:
                bolean = False
                break
    
    return bolean   
    
    


print(is_prime(2))   # True
print(is_prime(3))   # True  
print(is_prime(4))   # False
print(is_prime(5))   # True
print(is_prime(8))   # False
print(is_prime(17))  # True


```
