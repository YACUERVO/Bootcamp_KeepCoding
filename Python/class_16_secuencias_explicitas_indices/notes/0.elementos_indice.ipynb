# Indices en iteración

--------------

Recordemos que para iterar se necesitan 4 cosas siempre:

1. Dónde empezar
2. Dónde terminar
3. Cómo pasar de un elemento a otro
4. Qué hacer con cada elemento


Si te fijas verás que *índice* de un elemento dentro de la secuencia no aparece en la lista. Es decir, no es estrictamente necesario. En lenguaje antiguos, para poder implementar esas 4 cosas, era necesario usar el índice, pero el `for` de Python nos libera de eso.

El `for` de Python nos permite concentrarnos en tan sólo lo que de verdad importa: ¿qué hago con cada elemento de la secuencia?

Sin embargo, a veces sí que hace falta saber dónde se encuentra un elemento, además de qué es. 



### *La casa de paja*: gestionamos nosotros el índice

La solución en su momento, fue gestionar nosotros mismos el índice actual, mediante una variable que inicializábamos a 0 antes de empezar el bucle y luego la íbamos incrementando de uno en uno:

```Python
index = 0
for element in elements:
    # haz algo con el element
    # Actualiza el índice
    index = index + 1
```

Hay una solución mejor en la cual es el propio bucle `for` de Python el que nos da, no sólo el elemento de turno, sino también su índice.

### *La casa de ladrillo*: la función `enumerate` de Python

Usando la función `enumerate` con la secuencia sobre la que queremos iterar, el bucle `for` no sólo va a crear una variable con el elemento, también creará otra con el índice de dicho elemento.

````python
for index, item in enumerate(['a', 'b', 'c', 'd']):
    print(index, item)
````
