## Las Virtudes del Código

El buen código es:

1. Claro 
2. Correcto
3. Eficiente
4. Flexible
5. Seguro (más adelante)


### Claridad

Debe de ser nuestro primer objetivo. El código debe de ser claro, sencillo y evidente. 

La **claridad** se logra siguiendo los Pilares de la Ciberkinesis.

### Corrección

El segundo objetivo, y absolutamente imprescindible, es la **corrección**. El código debe de ser correcto, y **hacer lo que tiene que hacer**.

De nada sirve una función muy rápida, si no hace lo que tiene que hacer.

Eso sí, el código claro e incorrecto, es preferible al código confuso y correcto: lo claro puede ser entendido y arreglado, lo confuso (aunque correcto) no podrá ser modificado ni mantenido y  siempre terminará siendo tirado a la basura.

La corrección se logra *comprobando el funcionamiento* del mismo. De momento lo haremos con una celda de comprobación y más adelante lo haremos de forma más sistemática con *testing*

```python
def miles_to_km(miles):
    """
    Pasa de millas a km. Si recibe un valor negativo, devuelve None
    """
    result = None
    if miles >= 0:
        result = miles * 1.60934
    return result
```

```python
# comprobación
assert miles_to_km(0) == 0  #caso trivial
assert miles_to_km(1) == 1.60934
assert miles_to_km(-32) == None # caso erroeo
```

En este curso, nos centraremos sobre todo en los 2 primeros, por ser esenciales y ser los únicos obligatorios.

### Eficiencia

Todo software claro y correcto, debería en la medida de lo posible ser eficiente. Es un extra que sólo debe de ser perseguido si tenemos los dos puntos iniciales y se ha **demostrado** (hay herramientas para ello) que la función en cuestión es el *cuello de botella*. 

Nunca inviertas las prioridades. **Donald Knuth**, dios de la ciberkinesis y uno de los mayores expertos en algorimia y eficiencia computacional ya lo advirtió:

> **La optimización prematura es la raíz de todos los males**

Knuth la mencionó en su libro “The Art of Computer Programming”, citando que aproximadamente el 97% del tiempo, no es necesario preocuparse por la optimización, ya que enfocarse demasiado en optimizar pequeñas partes del código puede llevar a un diseño de software complicado y difícil de mantener.


La eficiencia se logra usando las estructuras de datos más adecuadas, los algoritmos más adecuados y evitando trabajo inútil (en ese orden). Un estudio completo de la eficiencia se sale del ámbito de este curso.


### Flexibilidad

Toda función clara y correcta, debería de ser lo más flexible posible. Es decir, se debe de poder reutilizar sin mayores dificultades y sin  mayores cambios. Al igual que la *eficiencia*, la *flexibilidad* es un extra que sólo debe de ser perseguido si tenemos los dos puntos iniciales y nos vemos ante la necesidad de  reutilizarla en un contexto diferente.

La *flexibilidad* se logra siguiendo los *Pilares de la Ciberkinesis*. Veremos ejemplos concreto más adelante.

### Seguridad

Todo software claro y correcto debe de ser *seguro*: es decir, no debe de tener agujeros o fallos conocidos que permitan a un atacante el robos de datos, pérdida de información o control total por parte de atacantes. 

El estudio de la seguridad también se sale del contexto de este curso, aunque hay un curso específico en la plataforma online de KeepCoding.
