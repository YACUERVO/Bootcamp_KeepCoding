# Datos Sintéticos

----------------

Con la función `all_possible_names` lo que hemos creado ha sido *datos sintéticos*. Estos tienen mucho más utilidad de lo que parece.

Los datos sintéticos son **datos generados artificialmente mediante algoritmos**, en lugar de ser recopilados del mundo real. Son de gran importancia en los ámbitos de testing y de la inteligencia artificial (IA) por ejemplo.




### Testing

**Variedad y Exhaustividad**

En testing, los datos sintéticos han creado una pequeña revolución. Uno de los grandes problemas de los tests, es que lo hacemos normalmente a través ejemplos concretos. Si el test pasa, sabemos que *ese caso concreto funciona*, pero ¿y el resto?.

veamos un ejemplo concreto sencillo: una función de suma. Recibe dos números y devuelve su suma. La forma tradicional de comprobar dicha función sería con ejemplos concretos:

```Python
assert add(3,5) == 8
assert add(-1, 6) == 5
````

Sin embargo, podría darse el caso de que haya valores para los cuales no funcione. A lo mejor con valores muy grandes o muy pequeños. ¿Cómo descubro eso? Ahí es donde entran los *datos sintéticos*: genero una gran cantidad de datos (números que se sumaran) y los pruebo todos.

Para poder probarlos, tenemos que fijarnos en las **propiedades de la función**. *¿Qué es lo que caracteriza la función `add`?*

1. El orden de los parámetros no importa. Sumar a y b es lo mismo que sumar b y a.
2. Sumar cero es lo mismo que no hacer nada.
3. Sumar 1 dos veces, es lo mismo que sumar 2.

Para las mentes más matemáticas, esto es lo mismo que decir:

1. Conmutatividad
2. Identidad
3. Asociatividad

Podríamos generar funciones que comprueban cada una de las propieades de la función `add` y llamarlas con datos sintéticos. Si falla en alguna parte, es que hemos encontrado algún valor concreto que hace que la función no se comporte como es debido. Es decir, hemos demostrado que la función que parecía total es realmente parcial.


`````python
import random

def add(a,b):
    return a + b

# El uso de _ dentro del for, se usa para indicar que no me import
# el valor que se va a asignar a dicha variable, sólo quiero repetir
# el cuerpo del for un cierto número de veces.

def test_conmutativity():
    for _ in range(1000000):
        a = random.uniform(-1000000, 1000000)
        b = random.uniform(-1000000, 1000000)
        assert add(a,b) == add(b,a)

def test_identity():
    for _ in range(1000000):
        k = random.uniform(-1000000, 1000000)
        assert add(k,0) == add(0,k) == k

def test_associativity():
    for _ in range(1000000):
        k = random.uniform(-1000000, 1000000)
        assert add(add(k,1),1) == add(k,2)
`````

`````python
test_conmutativity()
test_identity()
test_associativity()
`````

Aunque un estudio exhaustivo de testing está fuera del objetivo de este material, es un buen ejercicio siempre que diseñes una función, el pensar **cuales son las propiedades que la definen**, más que **cual es el retorno esperado para un cierto parámetro**.

Si sabes cuales son las propiedades, sólo hace falta generar datos sintéticos para someter la función a una comprobación exhaustiva que te impida provocar el nuevo caso "Crowdstrike".

Para la generación de *datos sintéticos* de testing hay herramientas específicas. Generar números es fácil, pero ¿y si tienes que generar nombres, direcciones de email, números de tarjeta, direcciones postales, etc? Para eso es mejor usar una herramienta específica.

La primera de todas se llama *[QuickCheck](https://hackage.haskell.org/package/QuickCheck)* y se inventó para un lenguaje llamado *Haskell*. Todos los demás lenguajes lo copiaron y ahora hay miles. La más común para Python se llama [Hypothesis](https://hypothesis.readthedocs.io/en/latest/).

De todas formas, el uso de estas herramientas es algo que es mejor dejar para un poco más adelante.

**Privacidad y Seguridad**

Los datos sintéticos también se usan en casos donde los datos reales contienen información sensible (por ejemplo, datos personales o financieros), los datos sintéticos pueden ser utilizados para evitar problemas de privacidad y cumplir con las regulaciones de protección de datos.


### Inteligencia Artificial


Los datos sintéticos también se usan mucho en Inteligencia Artificial. Uno de los casos más comunes es la **Ampliación de Datos**.

Los datos sintéticos permiten aumentar el volumen de datos disponibles para el entrenamiento de modelos de IA, lo que puede mejorar el rendimiento y la precisión de los modelos. Cuando no tenemos datos reales suficientes para entrenar un modelo, podemos usar datos sintéticos.

Esto no siempre es factible, ya que a veces los datos sintéticos provocan errores llamados *alucinaciones* en ciertos modelos de IA.

Otro caso común es la generación de **Entornos Virtuales**. En aplicaciones como conducción autónoma, los datos sintéticos pueden ser utilizados para simular entornos y situaciones que serían difíciles, caras o peligrosas de recrear en el mundo real. Antes de soltar al coche autónomo por las calles abigarradas de Calcutta, primero le haces conducir por calles y entornos sintéticos, donde no hay peligro.



## Resumen

* Usando las propiedades que definen una función y datos sintéticos, podemos hacer comprobaciones exhaustivas de nuestro software, algo indispensable en la Ingeniería de Software profesional.
* Cuando el dato es sencillo (cadenas, números), es fácil generar datos sintéticos. Cuando los datos son más complejos (nombres, direcciones, cuentas de banco, etc) es preferible usar herramientas especializadas.
  * La primera de dichas herramientas fue *QuickCheck* y las demás se parecen mucho.
  * En Python, la mejor a fecha de hoy es Hypothesis.
* En IA se utilizan mucho los datos sintéticos, pero no están exentos de problemas, como por ejemplo las llamadas *alucinaciones*
