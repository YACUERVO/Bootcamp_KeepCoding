# Generadores Combinatóricos

-----------------

La función `all_possible_names` es un ejemplo de un tipo de función llamada **Generador Combinatorio**.

Los generadores combinatorios son funciones que permiten la generación sistemática de todas las combinaciones posibles de un conjunto de elementos según ciertas reglas. 

Por ejemplo, `all_possible_names` genera todos los nombres posibles, siguiendo las siguientes normas:

1. Un nombre puede ser simple
2. Un nombre puede ser compuesto mediante la combinación de 2 nombres simples
3. En caso de un nombre compuesto, los dos nombres simples no pueden ser iguales



## Importancia de los Generadores Combinatórios

Permiten explorar de manera exhaustiva todas soluciones posibles a un problema para buscar la mejor de todas, o al menos una muy buena. A este tipo de aplicación se le llama **búsqueda** y **optimización**.

Por ejemplo, de todas las posibles rutas de un entregador de Amazon, cual es la mejor, o al menos una muy buena que optimice los siguientes criterios:

* tiempo
* distancia
* consumo de gasolina
* los tres a la vez

Este tipo de problemas caen dentro del campo de estudio de la IA y una de sus primeras aplicaciones fueron los juegos. 

Supongamos que estás creando un software de Ajedrez para enfrentarte a la nueva campeona del mundo:

![](next_move.png)

Tienes que crear una función llamada `next_move` que a partir de una cierta jugada en el tablero, determina *todas las posibles respuestas* a la jugada de tu oponente. Ahora, para cada una de esas posibles jugadas tuyas, calcula *todas las posibles respuestas* de tu oponente. Repite ese proceso el mayor número de veces que sea posible (es muy costoso y consume tiempo y memoria). Una vez tengas esas secuencias de jugadas, podrás analizarlas y eliminar las que no sirven (un *selector*), porque te lleven a perder o le den ventaja al oponente, y escojer la mejor posible.

Con ese procedimiento, puedes lograr que tan sólo mediante *fuerza bruta* (viendo muchas jugadas más adelante) tu software juegue a nivel de un Gran Maestro.

**Deep Blue**, el ordenador que derrotó a Gary Kasparov era un sistema diseñado para aprovechar precisamente la fuerza bruta para ganar. No sólo usaba capacidad de cálculo para ganar, pero sí era su gran ventaja frente a un Gran Maestro humano.

## Explorar grandes volúmenes de datos

A estas alturas, no sabríamos cómo mejor representar todas esas jugadas mencionadas anteriormente. Podría ser una lista de listas, pero da la impresión que los datos son más complejos, ya que cada jugada genera otras muchas más y así sucesivamente.

Más adelante veremos una estructura de datos más adecuada que la lista para representar este tipo de datos, llamada *árbol*. También veremos formas prácticas de representar secuencias enormes e inclusos *infinitas*. ### Un ejemplo cotidiano: Spotify

*"Pues mi aliado es Lambda y un poderoso aliado es."*

Acabas de ser contratado por Spotify y tu primera tarea es crear una nueva interfaz para que los usuarios puedan crear playlists. El objetivo es el siguiente:

1. El usuario elige una canción cualquiera.
2. El sistema contesta con las 4 canciones más parecidas.
3. El usuario elige una de ellas, que se añade a la playlist.
4. Ahora tienes dos canciones en la lista. El sistema te devuelve las 4 que más pegarían como *continuación*
5. El usuario elige una de las propuestas, y el proceso se retoma hasta que el usuario lo detenga.

Decides implementar la función `continuation` que recibe una lista de canciones y devuelve una lista de *continuaciones* más probables.

Decides aplicar *Divide & Vencerás*. El problema de este *generador* se divide en dos partes:

1. Explorar de forma eficiente los millones de canciones que hay en Spotify. Es un problema de exploración de un espacio de soluciones (canciones en este caso) muy amplio.
2. En cada paso, determinar cuales son las 4 canciones más probables teniendo en cuenta el pasado (las que ya están en la playlist)

Para la primera parte, tenemos todos los conocimientos necesarios (funciones y clausuras léxicas) y lo veremos con detalle en el capítulo dedicado a *evaluación perezosa*. ¡Pocos son los problemas en Ciberkinesis que el poder de Lambda no resuelve!

Para la segunda parte necesitaríamos un modelo probabilístico basado en los datos de gustos del usuario. En una situación real, te lo daría el departamento de ciencia de datos. Si quisieras desarrollarlo, tendrías que aprender más de ciencia de datos e inteligencia artificial.

Se trata de una clase de problemas muy común, el encontrar la *continuación* más probable de una secuencia de datos. Es lo que hacen los sistemas de recomendación, como cuando Youtube o TikTok te recomienda otros videos, cuando Amazon te recomienda productos que cree que te pueden interesar o cuando Netflix te recomienda nuevas series. También es la base de los LLMs: cuando chatGPT te *contesta*, en realidad lo que está haciendo es averiguar cual es la secuencia de palabras más probable para *continuar* lo que tú has escrito.
