### Producto Cartesiano

El producto cartesiano de dos listas es un generador combinatorio que recibe dos listas y devuelve una lista de listas (un LoL). Dicho LoL contiene toda slas combinaciones posible de un elemento de la primera con los de la segunda.

Debes de crear la función `cart_product` que funcione de la siguiente forma:

```Python
# Definimos las dos listas
lista1 = [1, 2, 3]
lista2 = ['a', 'b']

cart_prod(lista1, lista2)
# devuelve la siguiente lista:
# [[1, 'a'], [1, 'b'], [2, 'a'], [2, 'b'], [3, 'a'], [3, 'b']]
```

Cuando cualquiera de las dos listas es la vacía, el resultado ha de ser la lista vacía.

## Objetivo 

Crear una función que genere todos los posibles nombres de Dios es una tarea un poco difícil, sobre todo por el tamaño del valor de retorno (una lista de cadenas). Suponiendo que se usa el alfabeto español (27 letras), y que son palabras, como dice el cuento, de 9 letras, nos saldría la siguiente cantidad de cadenas:

$$27^9 = 7.625.597.484.987$$

Eso viene a ser siete billones seiscientos veinticinco mil quinientos noventa y siete millones cuatrocientos ochenta y cuatro mil novecientos ochenta y siete. Dicho de otra forma, un huevo de nombres.

De hecho es muchísimo más que los 9 mil millones del cuento. Es cierto que en el cuento se menciona al menos una restricción: ninguna palabra puede contener 3 o más consonantes seguidas. Eso debería de reducir el número de posibilidades, pero aun así son demasiadas. Probablemente el *alfabeto especial* que usan los lamas debe de ser más pequeño que el español.

De todas formas, no podemos crear una lista con billones (ni siquiera con miles de millones) de cadenas. No creo que caigamos en la consecuencia que se menciona en el cuento, pero seguro que causa el colapso de tu ordenador.

Lo ideal sería ir creándolos de uno en uno, en vez de todos de un golpe y tener esa burrada de datos en memoria. Además, al hacerlo de uno en uno, *podemos parar en cualquier momento*, evitando así el riesgo de llegar hasta el último (no vaya a ser que los lamas tengan razón).

Más adelante veremos cómo hacer este tipo de cosas, usando el *poder de lambda*.
