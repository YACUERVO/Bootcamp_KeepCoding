En Brasil existe una canción infantil llamado "O sapo não lava o pé" (*el sapo no se lava los pies*) que cuenta la historia de un sapo cuyos pies huelen fatal pero se niega a lavarlos. La letra es la siguiente:

````
O sapo não lava o pé
Não lava porque não quer
Ele mora lá na lagoa
Não lava o pé porque não quer

Mas que chulé!
````

Además de [horriblemente pegadiza](https://www.youtube.com/watch?v=2Vf4VhAGhlY), tiene la siguiente característica interesante:

1. La primera vez, cantas la letra completa tal cual.
2. La segunda vez, la cantas cambiando todas las vocales por una **a**: *a sapa na lava a pa*
3. La tercera, la cantas cambinado todas las vocales por una **e**: *e sepe ne leve e pe*
4. Asi sucesivamente. El bucle continúa *hasta que* se recorran todas las vocales (aeiou), el niño se equivoque, **o** el adulto empiece a gritar y llamar a Herodes para que le ayude. Como ves, se trata de un bucle *while*. 🤓

### Tarea 3.2: Eliminar caracteres acentuados

Como ya hemos visto anteriormente, esto no es una tarea trivial por la forma en que se codifica el texto en Python (y otros lenguajes también). Esa codificación se llama UNICODE y Python tiene módulos propios para tareas relacionadas con unicode.

El módulo en concreto se llama `unicodedata` y ya lo hemos usado en el pasado.

Aquí tienes una función que elimina carateres acentuados.

`````python
import unicodedata

def eliminar_acentos(cadena):
    forma_nfkd = unicodedata.normalize('NFKD', cadena)
    solo_ascii = forma_nfkd.encode('ASCII', 'ignore')
    return solo_ascii.decode()

s = "Mëtàl Hëàd"
s = eliminar_acentos(s)
print(s)  # imprime: "Metal Head"
`````

En este código:

* La función `normalize` convierte la cadena a la forma 'NFKD', que es una forma que separa los caracteres en sus caracteres base y sus acentos.
* La función `encode` luego codifica la cadena en formato ASCII, ignorando cualquier carácter que no pueda ser representado en ASCII (lo cual incluye los acentos).
* La función `decode` se utiliza para convertir el objeto de bytes de vuelta a una cadena.

#### Preprocesamiento en servidores web

Un servidor web se puede modelar como una función que recibe una URL (una cadena con un cierto formato) y que devuelve una cadena con html.

```Python
def serve(url:str)->html:str:
```

La realidad es un poquito más complicada que eso. Empecemos por el *parámetro*. No se trata de una URL, sino que es una cadena, con varias lineas y cada una va cierta información. Una de ellas es la URL que estamos pidiendo.

Si abres un explorador y vas a la página principal de [www.keepcoding.io](https://www.keepcoding.io), lo que tu explorador va a enviar al servidor web es lo siguiente:

```
GET /index.html HTTP/1.1
Host: www.keepcoding.io
User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36
Accept: text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8
Accept-Language: en-US,en;q=0.9
Connection: keep-alive
```

##### Explicación de cada línea:

1. **`GET /index.html HTTP/1.1`**: Esta línea indica el método HTTP (`GET`), la ruta del recurso solicitado (`/index.html`), y la versión del protocolo HTTP (`HTTP/1.1`).
  
2. **`Host: www.keepcoding.io`**: El encabezado `Host` especifica el dominio del servidor al que se está haciendo la solicitud, que en este caso es `www.keepcoding.io`.

3. **`User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36`**: Este encabezado identifica al cliente que hace la solicitud (generalmente un navegador). Aquí se muestra un User-Agent típico de un navegador Chrome en Windows.

4. **`Accept: text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8`**: Este encabezado indica los tipos de contenido que el cliente puede procesar. En este caso, se prefieren HTML y XHTML, seguidos por XML y varios formatos de imagen.

5. **`Accept-Language: en-US,en;q=0.9`**: Indica los idiomas preferidos para el contenido, en este caso, inglés estadounidense.

6. **`Connection: keep-alive`**: Este encabezado solicita que la conexión TCP permanezca abierta para posibles solicitudes adicionales en lugar de cerrarse inmediatamente después de la respuesta.

Esa cadena en bruto es muy dificil de manejar y extraer información de ella es complejo y proclive a errores. Por eso **se pre-procesa para convertirlo en algo útil**.

Tenemos una función llamada `make_request` que recibe una cadena en bruto con una petición http y devuelve un **tipo nuevo**: `request`.

El **tipo `request`** es un tipo que está a su vez *compuesto de otros tipos* (las cadenas de cada linea de la petición en bruto, alguna enum, etc...). Aun no sabemos cómo crear *tipos compuestos a partir de otros*, pero lo veremos más adelante (a esos tipos se les suele llamar *clases*. ¿Has visto la palabra clave `class` en alguna parte? Hmmm). De todas formas, lo importante es un valor de tipo `request` es mucho más manejable y seguro que una cadena en bruto.

La función `make_request` es parecida a la `normalize_string` que hemos creado anteriormente, en el sentido que pre-procesa los datos y **convierte datos en información** que luego podemos usar.

El servidor ahora quedaría así:

```Python
def serve(petition:str)->html:str:
    # preporcesamos la petición en un valor de tipo
    # request, para que sea más manejable
    request = make_request(petition)

    # La función respond es la que hace el trabajo de verdad,
    # leyendo un fichero html, llamando a una base de datos, lo que sea,
    # pero a de devolver html
    html = respond(request)

    # devolvemos el html que es lo que recibe tu navegador
    return html
```

Esta es una visión simplificada, pero representa el funcionamiento básico de todo servidor de http y todas las frameworks web en cualquier lenguaje. Cuando creas una aplicación web, como pueda ser Hotmail (una de las primeras *web apps*) o Google Ads, lo que haces en realdiad, es escribir muchas funciones `respond`. El pre-procesamiento (la función `make_request`) lo hace la framework web. 

La clave está en que el *pre-procesamiento* de los datos (*TRANSFORMA Y VENCERÁS*) **asegura la sencillez y corrección del código nucleo de tu programa** (las funciones `respond` o en nuestro caso `swap_vowels`).



#### Preprocesamiento en Big Data e IA

El pre-procesamiento de los datos es una etapa tan común en Big Data e Inteligencia Artificial que hasta hay un nombre específico para ello: ETL.

Ya no se trata de una simple función, sino que son herramientas completas y varias etapas hasta tener datos de alta calidad para ser usados.

**ETL** es un acrónimo que significa **Extract, Transform, Load** (Extracción, Transformación y Carga). Es un proceso fundamental en la gestión de datos que consiste en tres etapas principales:

1. **Extracción (Extract)**:
   - Consiste en recolectar datos de diferentes fuentes, que pueden ser bases de datos, archivos, APIs, etc. El objetivo es obtener todos los datos relevantes para su posterior procesamiento.

2. **Transformación (Transform)**:
   - Una vez extraídos, los datos se transforman para adecuarlos al formato necesario. Esto incluye limpiar, normalizar, agrupar, agregar, y aplicar reglas de negocio. La transformación asegura que los datos sean consistentes, correctos y estén en un formato adecuado para su análisis.

3. **Carga (Load)**:
   - Finalmente, los datos transformados se cargan en un sistema de destino, como un data warehouse, una base de datos, o un sistema de almacenamiento en la nube. Desde allí, los datos pueden ser utilizados para análisis, informes o alimentar algoritmos de inteligencia artificial.

Para que los modelos de IA funcionen correctamente, necesitan ser entrenados con datos de alta calidad. El proceso ETL asegura que los datos estén en la forma adecuada, sin ruido o inconsistencias, y que se encuentren correctamente etiquetados o preparados para su uso en algoritmos de aprendizaje automático. Sin un proceso ETL correct, los resultados de los modelos de IA pueden ser inexactos o poco confiables.



#### Preprocesamiento y el lavavajillas

Una de las tareas domésticas que más me desagrada, es vaciar el lavavajillas y guardar las cosas de nuevo en su sitio. En concreto, me molesta y da pereza los viajes interminables entre el lavavajillas y el armario para guardar cada cosa.

¿Cómo mejoramos el proceso? ¡Con Ciberkinesis!

Divide el problema en dos:

1. Saca y preprocesa las cosas
2. Guárdalas de una sola vez en el armario

Por ejemplo, primero saca todos los *platos llanos* y apílalos en la encimera, luego los hondos, etc. Una vez *pre-procesado* el contenido de lavavajillas, sólo tienes que hacer un viaje para guardar todos los platos o cualquier otro *tipo* (¿lo pillas?) de elemento de vajilla.

El pre-procesamiento simplifica y reduce el trabajo de la tarea siguiente y la suma de las dos tareas es menor que hacelro todo de un golpe.

#### Preprocesamiento en la cocina

En la cocina, el pre-procesamiento se le llama *mise en place*. Es un término francés que significa literalmente “poner en su lugar” o “poner en orden”. Se refiere al proceso de preparar y organizar todos los ingredientes y utensilios necesarios antes de comenzar a cocinar. Esto incluye picar, medir, lavar, y disponer los ingredientes, así como preparar las herramientas y equipos que se van a utilizar.

Tener todo preparado y organizado antes de empezar a cocinar permite un flujo de trabajo más fluido y eficiente. Los cocineros no necesitan detenerse a buscar ingredientes o utensilios mientras cocinan, lo que ahorra tiempo y reduce el estrés y los errores en la cocina.
