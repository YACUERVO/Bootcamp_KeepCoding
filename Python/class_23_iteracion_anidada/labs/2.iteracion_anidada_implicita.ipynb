### Error en el generador de nombres

Tal como está nuestra función es una función total, pero no es cierto. Hay un requisito que está implícito y que no hemos comprobado.

La lista de nombres que recibimos *tiene que ser una lista de nombres sencillos*. El nombre del parámetro lo da a entender, pero el tipo no lo puede comprobar y el código tampoco lo hace. Tenemos que arreglarlo.

Modifica la función para que cuando reciba una lista que contiene un nombre compuesto (¿cómo lo detectas?) lance una excepción `ValueError` indicando que dicho nombre no es aceptable.

Por ejemplo, la siguiente llamada debería de lanzar un error puesto que *José Luís* no es un parámetro admisible (es compuesto)

```python
all_possible_names(["Antonio", "José Luís", "Pedro"]])
````
