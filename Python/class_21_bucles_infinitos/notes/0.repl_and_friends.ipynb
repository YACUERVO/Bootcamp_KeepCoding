# REPL, RunLoop, EventLoop, Game Loop

-------------

![](repl.webp)


#### REPL

La versión más sencilla de este bucle, es la que se da en programas en modo texto y se le suele
llamar el REPL (Read-Eval-Print_Loop): Bucle de Leer, Evaluar, Imprimir.

Estamos hartos de verlo todos los dias, por ejemplo, cada vez que arrancas un intérprete de Python.
Cuando llamas a iPython, ocurre lo siguiente:

1. Mientras no escribas `quit()`
2. Lee lo que escribes
3. Lo evalúa
4. Imprime el resultado

La linea de comandos, que también has usado, funciona exactamente igual. Lo único que cambia es la
evaluación (no es un programa de Python, sino comandos de Bash) y la señal de salida.




#### RunLoop

En los programas gráficos, tenemos el mismo concepto, pero con otro nombre. Aquí se le suele llamar el
RunLoop (bucle de ejecución). Pensemos en el browser que estás usando para ver esto. Cuando lo arrancas
también entra en un bucle:

1. Mientras no cierras la ventana...
2. *Lee* lo que haces (hacer clic, seleccionar, etc)
3. Lo *interpreta*
4. *Imprime* el resultado (cambia lo que se ve en la ventana)

Así es como funciona TODO software gráfico.



#### Event Loop

En programas que no interactúan directamente con un humano, sino con otro software, también se da este patrón, pero con otro nombre (ganas de complicar la vida): el event loop
o bucle de eventos.

Por ejemplo, el servidor de http que sirve páginas web a los browsers. Cuando nginx (un servidor muy
común) arranca, lo primero que hace es empezar a tocarse los huevos a la espera de un *evento*. 

El *evento* es que reciba una petición, por ejemplo, un browser le pide la página en `/index.html`.

En ese caso, el servidor despierta, lee ese fichero y se lo entrega al browser. Acto seguido, reinicia
su siesta hasta que alguien lo vuelve a molestar.



#### El Game Loop y el futuro

Estos conceptos te los vas a volver a encontrar más adelante. Por ejemplo, si optas por mobile, verás el
run loop, mientras que si optas por Web el event loop será uno de los protagonistas del módulo de backend
avanzado con node.js

Los juegos también tienen siempre ese mismo concepto, ya sea en modo texto o gráfico:

1. Mientras no se acaba la partida
2. *Leo* lo que hace el jugador
3. Lo interpreto
4. Lo *imprimo*

En el caso de los juegos, se usa el nombre *Game Loop*
