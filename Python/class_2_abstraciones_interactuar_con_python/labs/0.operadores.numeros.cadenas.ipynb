# Operadores



## Operadores sobre números

Algunos de los operadores que entiende Python para tratar con números, son los familiares de la aritmética, aunque algunos tienen un símbolo distinto del habitual

* Suma y resta: `+`, `-`
* Multiplicación y división: `*`, `/`
* Potencia: `**`

Calcula lo siguiente usando Python:

* El promedio de 3 y 4
* 101 elevado a 4

````python
101**4
````

## Operadores sobre cadenas de texto

Las *cadenas de texto* o simplemente *cadenas* son otro **tipo de valores** de Python. Al igual que los números, hay una serie de operadores que actúan sobre ellas.

Cuariosamente, suelen tener el *mismo símbolo* que los aritméticos, aunque llevan a cabo cosas **totalmente distintas**.

### Concatenación

Consiste en pegar dos cadenas una a la otra:

> concatenar "hola" "mundo" -> "holamundo"

Se usa el símbolo `+`, aunque la operación es una concatenación, y no una suma.

`"Hola" + " " + "mundo"`

#### ¿De verdad son tan distintas la suma y la concatenación?

SI.

Por ejemplo, `2 + 3` y `3 + 2` son equivalentes, mientras que `"2" + "3"` es totalmente distinto de `"3" + "2"`

La concatenación no es conmutativa.

### Elemento Nulo

La suma tiene un elemento nulo, el cero. Si suma algo a `0`, no pasa nada. La concatenación de cadenas también tiene su **elemento nulo**, la *cadena vacía*.

`"qwerty" + ""` se evalúa a `"querty"`.

### Repetición

Se puede extender una cadena con copias de sí misma, usando el operador de multiplicación:

* `"x" * 4`
* `"qwerty" * 1`
* `"asdf" * 0`

El resultado del último ejemplo nos recuerda que las cadenas, al igual que los números también un elemento nulo. En vez del cero, es la cadena vacía.

### Indexado `[n]`

Nos permite obtener el caracter que está en alguna posición. 

> Recuerda que los programadores emepzamos a contar en el cero

```python
"lucas"[0] #devolvería l
```

> El indexado puede ser negativo para indicar que se empieza por el final, y en ese caso representa la distancia desde el final

```python
"lucas"[-1] #devolvería s
```

### Slicing o *fileteado*

Te permite obtener trozos o subcadenas.

El slicing se realiza utilizando corchetes y dos puntos (:) para separar los índices de inicio, fin y, opcionalmente, el paso.

La sintaxis general para el slicing es la siguiente:

`secuencia[inicio:fin:paso]`

Donde:

* `inicio`: Es el índice del primer elemento que queremos incluir en la porción extraída. Si se omite, se asume el inicio de la secuencia (índice 0).
* `fin`: Es el índice del elemento en el que queremos detenernos, sin incluirlo en la porción extraída. Si se omite, se asume el final de la secuencia.
* `paso`: Es el número de índices entre elementos consecutivos en la porción extraída. Si se omite, se asume un paso de 1.
  
Algunos ejemplos de slicing en Python:


```python
cadena = "Hola Mundo"

# Extraer los primeros 4 caracteres
subcadena1 = cadena[:4]  # Resultado: "Hola"

# Extraer los últimos 5 caracteres
subcadena2 = cadena[-5:]  # Resultado: "Mundo"
```
