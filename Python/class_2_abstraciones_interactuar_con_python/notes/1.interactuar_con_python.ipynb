##     Hipótesis

Creo que lo que hace es simplificar lo que yo escribo. Siempre y cuando tenga un cierto formato que le gusta, intenta simplificarlo hasta que ya no puede más.

Los números y los textos entre comillas ya no se pueden simplificar más y por eso los *repite*. Es su forma de decir *ya no hay nada que hacer*.

* Cuando algo no tiene el formato que le gusta, te da un error y pára.
* Cuando a lo largo de la simplificación se encuentra que no puede seguir, también da un error y se pára
* Cuando no ocurre nada de eso y todo sale bien, simplifica hasta que ya no puede más, y se pára.



##   Definiciones

* Los números y las cadenas de texto son ejemplos de **datos**. Cosas sobre las cuales se puede actuar.
* Los **datos** que ya no se pueden simplificar más, se les llama **valores**. `3`, `-42`, `"hola"` y `"jamón"` son **valores**.
* / * + - son ejemplos de **operadores**. 
* **Los operadores actúan sobre los datos**
* Las combinaciones de datos y  operadores, son **expresiones**
* **Las expresiones se simplifican**


### Ejemplos de expresiones

* `(1 + 9) - 4`
* `4 / 7`
* `4` No es una expresión, es sólo un valor
* `+` No es un expresión, puesto que no hay ningún valor

Toda **expresión** se puede simplificar a un **valor**.

### Ejemplos de Valores

Algunos de los valores que ya hemos visto son los siguientes:

* Números, ya sean enteros o decimales
* Cadenas de texto
* Quizás sorprendentemente, los *errores* también son valores.

Más adelante veremos muchos más.

## Evaluación

Al proceso de *simplificación* que lleva a cabo Python, le llamamos **evaluación**.  

Cuando le damos una *expresión* a Python, éste la **evalúa** hasta reducirla a un **valor**.

El proceso se lleva a cabo por partes, siguiendo una serie de reglas, que de momento parece incluir las reglas de la artimética (prioridad de operadores, prioridad de paréntesis, etc). A estas *reglas* las llamamos **Algoritmo de Evaluación**.

### Algoritmo de Evaluación

Esta es una primero versión, que no incluye todas las reglas de evaluación de Python, pero que nos sirve para evaluar el tipo de expresiones sencillas que manejamos ahora.

> Las expresiones se evalúan de izquierda a derecha, respetando la prioridad de los operadores aritméticos y de los paréntesis

Cuando introduzcamos nuevos elementos del lenguaje, tendremso que actualizar nuestro algoritmo.

Este proceso, que se lleva a cabo paso a paso, genera lo que llamamos una *torre de evaluación*.  


### Torres de Evaluación

````
(3 + 3) * 4 * (8 / 2)
6 * 4 * (8 / 2)
6 * 4 * 4
24 * 4
96
````

````
2.5 * (1 + 5) * (8 / 0)
2.5 * 6 * (8 / 0)
2.5 * 6 * Error
Error
````


Es muy conveniente hacer este ejercício con papel y lápiz para entender bien el proceso de **Evaluación**. Es un cocepto fundamental.


### Equivalencia de Expresiones

**Dos expresiones son equivalentes, si se simplifican al mismo valor**

Todas estas expresiones son equivalentes:

* (4 * 2) +2
* 8 + 2
* 10
