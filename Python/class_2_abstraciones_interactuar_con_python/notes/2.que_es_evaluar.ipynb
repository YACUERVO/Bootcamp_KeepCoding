# ¿Qué es la evaluación?


------------


Se trata del proceso que lleva a cabo Python para que tú te puedas comunicar con una inteligencia alienígena llamada *procesador*.

Se hace en varias etapas que ocurren en *dos mundos distintos*, el de los humanos (Python) y el de la máquina (lenguaje máquina). Para poder entender mejor lo que ocurre, hay que definir un concepto muy importante en la *Ciberkinesis*: la **representación**.




### Representación

¿Qué tienen en común las siguientes *expresiones*?

* `(8 + 8) * 2 + 1`
* `16 * 2 + 1`
* `40 - 7`
* `(60 + 6) / 2`
* `33`

Todos son *equivalentes* y son diferentes representaciones de una misma cosa. Lo más normal es que prefiramos la más sencilla (`33`), pero a veces puede resultar más conveniente otra, según lo que estemos haciendo.

Cambiar la *representación de unos datos* puede hacer que un problema que parecía imposible de resolver se convierta en trivial, o mejor aún, que se transforme en otro problema para el cual ya tenemos solución. El **cambio de representación** es la base del *Principio 1 de la Ciberkinesis: Transforma y Vencerás*.

Veremos muchos ejemplos a lo largo de nuestro viaje y consiste en *aprender a mirar el mismo problema con otros ojos*. Antes de avanzar, veamos algunos ejemplos que se puedan entender a estas alturas:

* **Invertir los dígitos de un número** (de `13` a `31` por ejemplo): Es mucho más fácil si conviertes el número a una cadena como paso intermedio.
* **Reconocimiento de imágenes**: cambia la representación de las imágenes a tablas de números y caes de paracaidas en una rama de las matemáticas que está super estudiada (Algebra Lineal) y para la cual hay hardware específico, potente, abundante y relativamente barato, las GPUs. Acabas de reinventar el Deep Learning.
* **Predicción de texto, tal y como hace chatGPT, Grokk y otros**. Busca una forma de representar el texto como tablas de números y estás en la situación anterior.
* **Criptografía**: En criptografía (la base de toda la seguridad informática) es muy importante calcular rápidamente el *Máximo Común Divisor* de dos números gigantes. Representa esos números, no como un número larguísimo, sino como una lista de números más pequeños, sus factores primos. Acabas de simplificar enormemente la tarea original.

> Aprende a mirar con ojos diferentes y a usar la imaginación.
