
# Jupyter lab: la primera herramienta

Vamos a usar varias herramientas, de complejidad creciente, para comunicarnos con el lenguaje. La más sencilla, y que no requiere ningún tipo de instalación en nuestro equipo, es [Google Colab](https://colab.research.google.com/).

Colab es un [Notebook de Jupyter](https://jupyter.org/) alojado en los servidores de Google y que puedes usar como si fuese un fichero de Google Drive.

👽 *¿Y qué es un Jupyter Notebook?*

Me lo temía... Es una herramienta para crear *libretas* (notebooks) que contienen texto, código ejecutable (en varios lenguajes, no sólo python), imágenes, gráficos etc. En realidad, estás mirando uno ahora mismo.

  Los notebooks de Jupyter son una herramienta muy útil y son muy utilizados en Big Data e IA. Usálos para apuntes.

> Abre un notebook en google colab, y empecemos con una celda de código



En las celdas de código iremos haciendo pruebas para intentar entender qué hace Python (o cualquier otro lenguaje de programación).

También puedes crear celdas de texto y usar la herramienta para apuntes. Es una herramienta muy común, especialmente en Big Data e IA y todo el material de este curso ha sido creado con Jupyter.
