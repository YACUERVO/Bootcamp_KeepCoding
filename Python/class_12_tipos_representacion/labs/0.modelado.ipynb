### (0) bmi en sistema internacional y sistema norteamericano

Los norteamericanos padecen de una inexplicable fobia al sistema métrico. Como muchos de tus clientes son de USA, decides modificar la función `bmi` para que se pueda usar tanto en un sistema como el otro. Tu primera reacción es cambiar el tipo de los parámetros a cadenas, que incluyan la información de las unidades:

```python
# versión antigua
def bmi(height: float, weight: float)->float:
    pass
```

Pretendes cambiarla por algo así:
````python
def bmi(height:str, weight: float)->float:
    pass
`````

El objetivo es poder llamarla de esta forma:

```python
bmi("6 ft", "200 lbs") # sistema usa
bmi("187 cm", "84 kg") # sistema internacional
````

Algo te dice que esto es un mojón de mucho cuidado. ¿Por qué?


**PISTA**

No leas esto antes de pensar por tu cuenta.






1. Los parámetros, peso y altura, son números y se deberían de representar como tal
2. El sistema de unidades es otra cosa, y no debería de estar mezclado y revuelto
3. Nadie te asegura que el usuario no mezcle sistemas de unidades al llamar la función
4. Habrá que analziar las cadenas para extraer el valor numérico y la unidad, y a partir de eso, deducir el sistema usado.
5. Estar convirtiendo de cadenas a números es ineficiente y proclive a errores.
6. Es una locura.


1. Crea una enum para reprsentar los sistemas de unidades (Métrico y Guiri)
2. Haz que los parámetros sean números como tiene que ser
3. Añade un tercer parámetro que indica el sistema de unidades (la enum)
4. Opcionalmente, si el uso de un sistema es mucho más común, dale al tercer parámetro ese valor por defecto.


### (1) Mal uso de cadenas y malos nombres


Estás creando una app web y hay contenido y funcionalidades que no deben de estar disponibles para los usuarios menores de edad. Decides crear una función para determinar si alguien es mayor o menor de edad. Tu primera versión es la siguiente:


```python
def age_type(age:int)->str:
    if age >= 18:
        return "major"
    else:
        return "minor"
```

Miras por encima del hombre y me ves, cual [Fantasma de la Fuerza](https://starwars.fandom.com/es/wiki/Fantasma_de_la_Fuerza) reaparecido para impartir justicia y castigo, farfullando algo así como *"¡Eras el elegido, tenías que acabar con las ñapas, no unirte a ellas!"*.

Enumera los defectos de dicha solución y piensa en cómo arreglarlo.


**PISTA**

No leas esto antes de pensar por tu cuenta.





1. El ser mayor o menor de edad es algo booleano: si no es una cosa, es la otra. Devuelve un booleano.
2. Cámbiale el nombre para que con sólo verla ya se sepa qué hace: `def is_major(age):bool`






### (3) Enumeraciones para valores fijos

¿Qué tiene de malo este código? Corríjelo.

```python
def handle_event(event_type):
    if event_type == "start" or event_type == "Start":
        print("Event started")
    elif event_type == "stop":
        print("Event stopped")
`````
