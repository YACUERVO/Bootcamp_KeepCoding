# Representar mediante Tipos


-------------

1. Usa siempre el tipo más específico posible. Por ejemplo para representar un entero, usa un `int`, no un `float`.
2. Si tienes que representar algo que puede tener varios valores diferentes, usa una `enum` específica.
3. Si tienes que representar valores verdaderos o falsos, usa un `bool`. Si no son verdadero o falso, aunque sean dos casos, usa una `enum` específica.
4. Para representar una coleción de valores, usa una lista.
5. Usa las cadenas única y exclusivamente para representar lo que es texto.
6. Si tienes que representar algo que no encaja en ninguna de esas categorías, puedes crear tipos propios, cosa que veremos más adelante.

## Datos Compuestos


¿Qué pasa si tenemos que representar algo que está compuesto por diferentes partes? Por ejemplo, supongamos que tenemos que representar un usuario dentro de un aplicación web. Un usuario tiene los siguientes componentes:

1. Nombre: `str`
2. Apellidos: `list[str]`
3. Email: ¿`str`?
4. Estado (De pago o gratuito): `enum`

La mayoría de los componentes sé cómo representarlos mediante las reglas anteriores. El único que me genera un poco de duda es el email, y ahora veremos por qué. Sin embargo, el usurario en sí, no sé cómo hacerlo, no hay ninguna regla que se aplique a cosas compuestas por otras.

La única opción, sería una lista de Nombre, Apellidos, Email, Estado. No tiene buena pinta, ya que padece muchas de las taras que tenía el sobreuso de las cadenas:

* No hay una estructura clara
* No es factible comprobar la validez mediante el intérprete o compilador
* Requiere de mucha documentación y es frágil

Está claro que nos falta algo.

Nos falta una forma de **crear tipos compuestos por otros tipos**. Lo veremos más adelante, cuando lleguemos a las tuplas y las clases.

### ¿Qué pasa con el email?

¿Por qué no me gusta el uso de una `str` para el email? Porque no logra modelar todas las propiedades de un email dentro de un sistema online. Para empezar, un email es un *tipo de cadena* y no toda cadena podrá ser un email válido.

De entrada, parece que además de la cadena, necesitaríamos un *predicado* que nos diga si ese email tiene un formato válido o no.

Si pensamos un poco más, nos daremos cuenta que dentro de la aplicación web, habrá servicios que sólo estarán disponibles para los usuarios que hayan confirmado su email. El estado de "pendiente de confirmación" y "confirmado" tampoco lo puede modelar correctamente una cadena. Hace falta una `enum`.

Es decir, por lo visto, un email es algo más complejo y compuesto de varias partes:

* dirección: `str`
* estado: `enum`
* validez: un predicado

Esto también tendrá que esperar a tuplas y clases.
