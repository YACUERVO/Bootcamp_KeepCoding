### Una herramienta para cada cosa, un tipo para cada dato

El error de Otilio es muy común entre los pogramadores novatos, que se niegan a usar los tipos de los que disponen para representar datos. En concreto, hay un caso tan común, que hasta hay un nombre para él: *stringly-typed code*. Consiste en usar cadenas para todo, tanto si es texto como si no.

Esto es un error gravísimo que debemos de extirpar de inmediato.



### La maldición del abuso de las cadenas o *stringly typed code*


La expresión "stringly typed" (tipado mediante cadenas), es un juego de palabras con "strongly typed" (fuertemente tipado). Este término se usa informalmente en programación para describir una situación donde los valores que deberían ser representados con tipos distintos (como enteros, enums, bools, etc) se representan mediante cadenas.

#### ¿Qué es "Stringly Typed"?

Un sistema "stringly typed" es aquel donde las operaciones que deberían usar tipos de datos más específicos o estructurados se implementan con cadenas. Por ejemplo, en lugar de usar un booleano o un tipo enumerado, un programa podría representar un estado con cadenas como "true", "false", "on", "off", lo cual puede llevar a varios problemas.



#### ¿Por qué se Considera una Mala Práctica?

1. **Propenso a Errores**: Comparar cadenas para operaciones clave es propenso a errores debido a posibles errores tipográficos, problemas de sensibilidad a mayúsculas y minúsculas, y preocupaciones de localización.

2. **Falta de Verificación en Tiempo de Compilación**: Los lenguajes fuertemente tipados pueden detectar muchos errores en tiempo de compilación. Al usar cadenas, estos errores se convierten en problemas de tiempo de ejecución, que son más difíciles de detectar y depurar.

3. **Legibilidad y Mantenimiento**: El código que abusa de las cadenas para el control de flujo o el estado puede volverse difícil de leer y mantener. Puede ser poco claro cuáles son los valores válidos y cómo deben usarse.

4. **Rendimiento**: Las operaciones con cadenas (como comparación, concatenación) suelen ser menos eficientes que las operaciones con otros tipos como enteros o enumerados.

5. **Falta de Claridad**: Usar cadenas para múltiples propósitos puede hacer que el código sea menos intuitivo y más difícil de entender y manejar para los nuevos desarrolladores.



### Ejemplo

En lugar de:

```python
def handle_event(event_type):
    if event_type == "start":
        # iniciar el evento
    elif event_type == "stop":
        # detener el evento
```

Un enfoque mejor sería:

```python
from enum import Enum

class EventType(Enum):
    START = 1
    STOP = 2

def handle_event(event_type):
    if event_type == EventType.START:
        # iniciar el evento
    elif event_type == EventType.STOP:
        # detener el evento
```

En el enfoque mejorado, el uso de un enumerado hace que el código sea más legible, menos propenso a errores y más fácil de mantener.



### ¿Por qué los novatos caen tanto en este error?

Una de las causas más comunes es el **miedo a la complejidad**. A veces los principiantes pueden percibir (¡erroneamente!) que el uso de tipos específicos y correctos como enumeraciones, booleanos y otros,  es una complicación innecesaria para tareas pequeñas o simples.

**No lo es**. Al usar tipos incorrectos, generas complejidad. Lo más sencillo es usar la herramienta adecuada para cada tarea.

Otra razón muy común es *porque sobrestimas a la máquina*. Si al crear un *predicado* devuelves una cadena `"YES"` eres capaz de inferir su significado, pero ni Python ni el ordenador pueden hacerlo. Ambos sólo ven que esa función devuelve un texto. El lenguaje sólo ve los tipos, así que cuando usas los tipos adecuados, **tanto tú como el ordenador entendeis el significado de tu código**. Eso es esencial para prevenir errores.
