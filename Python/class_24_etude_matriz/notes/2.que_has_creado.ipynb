## Representación de imágenes


Una imagen digital se representa comúnmente mediante una matriz de píxeles, donde cada píxel en la matriz corresponde a un punto de color en la imagen. Aquí te explico cómo funciona esta representación:

1. **Matriz de Píxeles**: La imagen se divide en una cuadrícula de pequeños puntos llamados píxeles. Cada píxel en esta cuadrícula tiene un valor específico que representa su color.

2. **Dimensiones de la Matriz**: La matriz es bidimensional donde el número de filas representa la altura de la imagen y el número de columnas representa el ancho. Por ejemplo, una imagen de 800x600 tiene 800 píxeles de ancho y 600 de alto.

3. **Valores de los Píxeles**:
   - **Imágenes en Escala de Grises**: Cada píxel se representa como un único número que mide la intensidad de la luz, donde valores más bajos (cercanos a 0) son más oscuros y valores más altos (cercanos a 255 en una imagen de 8 bits) son más claros.
   - **Imágenes a Color**: Para imágenes en color, cada píxel típicamente contiene tres o cuatro valores, correspondientes a los canales de color Rojo, Verde y Azul (RGB), y a veces un cuarto canal, Alfa, que representa la transparencia. Cada canal normalmente varía de 0 a 255, donde 0 significa ninguna intensidad de ese color y 255 es la máxima intensidad.

4. **Almacenamiento**: La matriz de píxeles se almacena en la memoria de la computadora como un gran array de datos. Cuando la imagen se muestra, el sistema gráfico interpreta estos datos para renderizar los colores apropiados en la pantalla.

Esta representación matricial de píxeles permite manipular imágenes a nivel de píxel, como en operaciones de edición, filtrado y transformación.


## Filtros de imágenes

El proceso de transformar cada punto en el promedio de sus vecinos es un filtro de imágenes muy común. 

La gran diferencia está en el tamaño de las imágenes, representadas como matrices de píxeles (3 números que representan los componentes rojo, verde y azul), son mucho mayores que las que hemos visto aquí.

Al ir promediando los valores, lo que hacemos es eliminar detalles y bordes, es decir, **difuminamos la imágen**. Has creado un filtro para ¡eliminar poros abiertos, arrugas y patas de gallo en el instagram!


![](difuminar.png)

### Pixelar y censurar

Si en vez de procesar cada elemento, lo hubiésemos hecho por *bloques*, en vez de suavizar, estaríamos *pixelando la imagen*. 

Si además, solo lo aplicas a un rango de índices (en vez de toda la matriz), habríamos *pixelado una parte de la imagen*, como la cara.


![](pixelar.png)

### Ampliar imágenes

Esto también es útil cuando amplias una imagen. Supon que tienes una imagen de 800x800 píxeles (una matriz de 800x800) y quieres ampliarla a 1600x1600.

Si distribuyes los píxeles originales de forma homogenea, van a *quedar huecos*: hay píxeles para los cuales no tienes datos y habrá que rellenarlos de alguna forma.

![](ampliar.png)

El problema con el que te enfrentas, es el de extrapolar puntos a partir de algunos datos conocidos. Una forma de *rellenar* esos huecos, es darle a esos píxels el valor ¡promedio de los originales que tiene de vecino!

De esa forma desaparecen los huecos y aparece algo que tiene sentido. Eso sí, a medida que amplías más y más, la imagen se difumina y pierde calidad.

### Dime con quien andas y te diré quién eres

El problema de extrapolar puntos a partir de datos conocidos es una de las tareas esenciales de la IA y el ML. Una forma de hacerlo es con un algoritmo llamado **K-NN**, que significa: *k nearest neighbours*. Eso significa los *k vecinos más próximos*.

Supón que tienes los siguientes datos de miles de clientes de banca:

* ingresos
* gastos 
* ha dejado o no de pagar un préstamo

Te llega un nuevo cliente y tienes que *adivinar* si es uno que te va a dar un pufo o no. Podrías hacer lo siguiente:

* guardo todos esos datos en una matriz
* inserto al nuevo cliente con sus ingresos y gastos
* compruebo si sus vecinos más próximos me han dado un pufo: si lo han hecho, considero que él también lo hará.


![](k_nn.png)

¿Cuantos vecinos miro? Pues eso depende de cada caso, y lo normal es probar varios valores (de `k`) y elegir el que funciona mejor.

Comparando al nuevo con sus k vecinos más próximos, extrapolo cual será su comportamiento.

K-NN suena muy bien, pero no es más que el viejo *dime con quién andas y te diré quién eres* de la abuela. Las abuelas saben mucho.

En la práctica, no tendrás sólo tres datos (llamados *features* en ML) de cada cliente. Tendrás cientos (estado civil, hijos, dirección, estudios, otros préstamos, ahorros, etc, etc) o miles y el espacio de datos no será bidimendisonal (una matriz) sino que tendrá **cientos de dimensiones**.

Sin embargo, eso no es un problema, porque lo único que necesitamos hacer es calcular la distancia entre dos puntos. Si el espacio no es un plano, sino que tiene cientos de dimensiones, la fórmula será parecida pero más larga. No será sustancialmente más compleja.

Para calcular la distancia cartesiana entre dos puntos en un espacio euclidiano de cualquier número de dimensiones, usamos la fórmula general de la distancia euclidiana y se basa en el teorema de Pitágoras. Cada letra representa una dimensión.

**distancia cartesiana entre dos puntos en un plano (2 dimensiones)**

$$d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$$


**distancia cartesiana entre dos puntos en 3 dimensiones**

$$d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2}$$


**distancia cartesiana entre dos puntos en 10 dimensiones**

$$d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2 + \ldots + (j_2 - j_1)^2}$$

**PS**

Si lo de cientos o miles de dimensiones te ha dejado patidifuso, piensa que en modelos de LLMs, como GPT o LLama, son **miles de millones**. Ahí es ná.
