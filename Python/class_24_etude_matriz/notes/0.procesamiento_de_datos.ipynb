# Procesamiento de datos

----------------

El objetivo final es crear la función `average_matrix` que recibe una matriz de `float`s y devuelve una nueva matriz de las mismas dimensiones, pero con cada punto sustituído por el promedio de sus vecinos y él mismo.

Se trata de *promediar* los valores de la matriz.


### Transformación

La transformación que aplica es la siguiente: cada elemento de la matriz pasa a tener como valor el promedio de su antiguo valor y los valores de sus vecinos.

### Vecinos 

Se considera vecino, aquel elemento con el se comparte una arista.

##### Tipos de elementos

* Los elementos que se encuentran en el borde, tienen 3 vecinos. Son una excepción.
* Los elementos que se encuentran en las esquinas, tienen 2 vecinos. Son otra excepción.

## Antes de Empezar

Para poder comprobar que nuestra función devuelve los valores correctos, necesitaremos crear algunas matrices. Vamos a diseñar dos funcione spara crear matrices de floats.


### `make_constant_matrix(rows: int, cols: int, value: float)->list[list[float]]`

Esta función recibe el tamaño de la matriz que debe de crear (filas y columnas) y en cada posición mete un valor constante (`value`).

Por ejemplo, `make_constant_matrix(2,2,0)` genera la siguiente matriz `[[0,0], [0,0]]`



### `make_random_matrix(rows: int, cols: int)->list[list[float]]`

Hace lo mismo, pero inserta en cada elemento un `float` aleatorio entre 0 y 1.

## Comprobación

Cuando tengas to función `average_matrix` lista, comprueba que fnciona con `asserts`:

* las dimensiones de la matriz de salida son iguales a las de la matriz de entrada
* cuando le pasas una matriz constante (creada con `make_constant_matrix`), la matriz de salida es igual a la de entrada
