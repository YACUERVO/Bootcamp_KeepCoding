#### Tokenizar (Transforma & Vencerás)


Nuestra tarea de romper el texto entre palabras se vería muy simplificada si en vez de tener una sóla cadena gigante
tuviésemos una *lista de cadenas*, siendo cada una de ellas una palabra. 

En vez de `"Hola Mundo!"`, tendríamos `["Hola", "mundo!"]`.

Para ello el primer paso sería *romper la frase* cortando en cada espacio en blanco.

El transformar una cadena larga en una lista de subcomponentes (palabras o lo que sea) es algo muy común. Se llama **tokenizar** una
cadena.

Tokenizar una cadena significa dividirla en partes más pequeñas, conocidas como "tokens". En programación, esto suele referirse a separar una cadena de texto en palabras, frases u otros elementos según ciertos delimitadores, como espacios, comas, puntos, etc. Esta operación es fundamental en el procesamiento de texto para análisis, búsqueda y otras operaciones.

También es la primera tarea que lleva a cabo Python cuando intenta interpretar un programa tuyo: lo primero es **tokenizar** tu código.

La forma más sencilla en Python es usar el método `split()`, que divide la cadena en tokens basados en un delimitador especificado. Si no se especifica un delimitador, por defecto se usa el espacio en blanco:

```python
# Ejemplo de tokenización de una cadena en Python usando split()

# Definimos una cadena de texto
texto = "Hola, este es un ejemplo de tokenización."

# Tokenizamos el texto usando el espacio como delimitador
tokens = texto.split()

# Imprimimos los tokens obtenidos
print(tokens)
```

En este ejemplo, `split()` divide la cadena `texto` en palabras, utilizando el espacio como delimitador por defecto. El resultado es una lista de tokens: `['Hola,', 'este', 'es', 'un', 'ejemplo', 'de', 'tokenización.']`.
