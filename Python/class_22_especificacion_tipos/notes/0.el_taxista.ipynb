# El taxista portugués

----------------------------


Durante un viaje a Portugal, pillé un taxi para ir de Aveiro a Coimbra. Como el trayecto era un poco largo, decidí charlar con el conductor. Al pasar por un pueblo llamado Mealhada, le pregunté si era ahí donde se preparaba un famoso cochinillo asado (siempre pensando en lo mismo).

- *No, me contesta.*
- *¿Ah si? Pues yo había entendido que...*
- *Para nada. En el pueblo están los **restaurantes**, y es **en los restaurantes** donde se prepara el cochinillo. (sic)*

![](pikachu-surprised.jpg)

Ni qué decir que no volví a abrir la boca en el resto del viaje.


Algunos lenguajes de programación tienen una forma de ser similar a la del taxista en cuestión. Tu código tiene que ser elaborado con *extrema precisión y lógica*, porque ninguna vaguedad será perdonada.


### Python no es así

En Python, se nos permiten ciertas vaguedades a la hora de expresarnos. Por ejemplo esta función:

```Python
def add(a,b):
    return a + b
```

No hemos especificado en ningún momento si lo que recibe y devuelve son números, aunque todo parece indicarlo. Aunque si lo piensas, bien podrían ser cadenas...

Python no entra en corto circuito y simplemente acepta que estás sumando o concatenando *cosas*.

Es fácil y rápido crear el código, aunque luego te expones a errores en tiempo de ejecución, como si llamas a la función de la siguiente manera: `add([3, 4], "hola")`.

Esto no es un fallo de Python, es una decisión de diseño. Python te facilita el empezar de forma rápida y presupone que sabes lo que estás haciendo. Si luego usas mal tus funciones, pues asunto tuyo: ya eres mayorcito y Python no es tu niñera.

> Los lenguajes como Python se llaman Lenguajes de Tipado Dinámico

**Ejemplos de Lenguajes de tipado dinámico**

* Python
* Ruby
* Javascript
* Objective-C


### Los lenguajes niñera

No todos los lenguajes tienen esa filosofía. Otros de hecho tienen exactamente la opuesta: parten del principio que no tienes idea de lo que haces y que hay que protegerte de tí mismo. Son una especie de [Señorita Rottenmeier](https://es.wikipedia.org/wiki/Se%C3%B1orita_Rottenmeier).

![](https://www.elocuent.com/wp-content/uploads/2013/06/rotenmeier.jpg)

De personalidad seca, áspera e intransigente, exigen que todos los detalles hayan sido especificados hasta la saciedad no dejando ninguna duda, por irrelevante que sea, si aclarar.

Un buen ejemplo de esta personalidad es [Swift](https://www.swift.org/).

La función de arriba quedaría así en Swift:

```Swift
func suma(num1: Int, num2: Int) -> Int {
    return num1 + num2
}
```

Al código se le añade información extra de tipado: los dos parámetros son de tipo entero y el resultado también ha de serlo.


> Los lenguajes como Swift se llaman Lenguajes de Tipado Estático

**Ejemplos de Lenguajes de Tipado Estático**

* Swift
* Kotlin
* Scala
* Java
* Typescript


### ¿Cual es mejor?

Asegún y adepende. Cada uno tiene sus ventajas.

* El *tipado estático* hace que ciertos bugs no lleguen al tiempo de ejecución, pero hacen que el proceso de crear el software sea más cansino y aburrido. El compilador constantemente te está llamando la atención sobre cualquier posible incoherencia.
* El *tipado dinámico* te da más libertad y hace el proceso de crear el software más rápido, sin tener que detenerse todo el rato con pequeños detalles. Con esa libertad extra, puedes hacer grandes virguerías, pero también grandes cagadas.

Hace unos años la moda era el tipado dinámico y los que no lo usaban eran unos pringados. Ahora es todo lo contrario, y si no usas tipado estático, eres un inútil.

> (...)“toda virtud es un termino medio entre dos extremos, cada uno de los cuales es un vicio.”
> Aristóteles

¿Lo mejor? Dejarse de fanatismos idiotas y tirar por el [Camino Medio](https://es.wikipedia.org/wiki/Camino_medio), aprovechando las ventajas de cada uno.
