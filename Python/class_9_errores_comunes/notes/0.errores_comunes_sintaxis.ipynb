# Sintaxis de Python

------------------

Todo lenguaje tiene una sintaxis, es decir una estructura fija que hay que seguir para que podamos usarlo para transmitir información. En Español por ejemplo, se sigue la regla general SVO (Sujeto-Verbo-Objeto) que es la más común en todos las lenguas del mundo:

* Español: "Juan (sujeto) come (verbo) manzanas (objeto)."
* Inglés: "She (subject) loves (verb) reading (object)."
* Chino Mandarín: "我 (sujeto) 喜欢 (verbo) 茶 (objeto)" ("Yo amo té").
* Francés: "Marie (sujet) regarde (verbe) la télévision (objet)."

Si yo digo "como pizza yo", es normal que otro hispanohablante no me entienda. De la misma forma es vital seguir las normas de la sintaxis de Python para que nos entienda.


Vamos a ver algunos de los errores más comunes cometidos por los novatos cuando empiezan con Python y las funciones.


1. **Omitir los dos puntos (`:`) después de la declaración de la función**
   - En Python, y otros lenguajes que usan una sintaxis similar, los dos puntos al final de la declaración de la función son esenciales
   - porque indican que se acabó la *firma* de la función (la primera linea que incluye nombre y parámetros).
   ```python
   def mi_funcion()  # Error
   def mi_funcion():  # Correcto
   ```



2. **Errores en la indentación**
   - La indentación es crucial en Python para definir el bloque de código que forma parte de la función.
   ```python
   def mi_funcion():
   print("Hola")  # Error, falta la indentación
   ```
```Python
    def mi_funcion():
       print("Hola")  # Correcto
   ```

   



3. **Confundir la definición de la función con la llamada**
   - Algunos novatos intentan pasar argumentos en la definición de la función en lugar de en la llamada. Una función es una plantilla, como un formulario de código, con huecos (con nombre) que se rellenarán cuando se llame. 

   - Por eso es un sinsentido poner **valores** (los números) en vez de **nombres** en los parámetros (los huecos).

   ```python
   def mi_funcion(5):  # Error
   def mi_funcion(x):  # Correcto
   mi_funcion(5)       # Llamada correcta
   ```



4. **No definir parámetros esperados en la declaración de la función**
   - Si una función espera un argumento, debe estar definido en la declaración.
   ```python
   def mi_funcion():  # Definición sin parámetros
       print(x)        # Error, x no está definido
   def mi_funcion(x):  # Definición correcta
       print(x)        # Correcto
   ```



5. **Llamar a funciones antes de definirlas**
   - En muchos lenguajes, incluido Python, las funciones deben ser definidas antes de ser llamadas.
   ```python
   mi_funcion()       # Error si la función aún no está definida
   ```

   ```python
   def mi_funcion():
       print("Hola")
   ```

   



6. **Usar el nombre incorrecto de la función al llamarla**
   - Esto puede parecer obvio, pero es común cometer errores tipográficos.
   - Comprueba siempre que usas los nombres correctos y **lee los errores que te da Python**: **ahí está toda la información** que necesitas para resolver el problema.
   ```python
   def mi_funcion():
       print("Hola")
   mi_funccion()  # Error, nombre incorrecto
   ```

   



7. **Mezclar tipos de datos en la llamada y definición**
   - Es importante que los tipos de datos de los argumentos pasados coincidan con los que la función espera.
   ```python
   def sumar(a, b):
       return a + b
   sumar("1", 2)  # Error, tipos incompatibles
   ```





8. **Omitir argumentos requeridos en la llamada de la función**
   - Si una función requiere ciertos argumentos, deben ser proporcionados al llamarla.
   ```python
   def mi_funcion(a, b):
       return a + b
   mi_funcion(5)  # Error, falta un argumento
   ```




9.  **Olvidar retornar un valor desde la función**
    - Si se espera que una función devuelva un resultado, debe incluirse explícitamente una declaración de retorno.
    ```python
    def sumar(a, b):
        resultado = a + b  # Error, no retorna nada
    ```
    ```python
    def sumar(a, b):
        resultado = a + b  # Error, no retorna nada
        return resultado    # Correcto
    ```




10.  **Referenciar la función sin llamarla**
   - Los novatos a veces olvidan incluir los paréntesis al final del nombre de la función cuando desean llamarla. Esto resulta en que en lugar de ejecutar la función, lo que se obtiene es una referencia al objeto de la función misma.
   - **Estás confundiendo la función en sí, con el acto de llamarla**
   ```python
   def mi_funcion():
       return "Hola mundo"

   print(mi_funcion)  # Imprime algo como <function mi_funcion at 0x10c866550>, que es la referencia al objeto de la función
   print(mi_funcion())  # Correcto, imprime "Hola mundo"
   ```



11. **Pasar la función como un valor a otra función sin llamarla**
   - Esto puede ser útil en casos avanzados, pero a menudo resulta en confusión si el novato espera que la función se ejecute cuando en realidad está pasando la función misma.
   - **Estás confundiendo la función en sí, con el acto de llamarla**
   ```python
   def saludar():
       return "¡Hola!"
       print(saludar) # No va a imprimir ¡Hola!
   ```

   


12. **Confundir el imprimir con devolver**

Es quizás el más insidioso, porque es dificil de percibir la diferencia, pero la diferencia es inmensa.

Todos los lenguajes tienen alguna función que permite mostar algo en pantalla. En el caso de Python, se llama `print`.

Supongamos que tienes estas dos funciones:


```python
def foo():
    print("Hello")

def bar():
    return "Hello"
```

Cuando las llamas, da la impresión que hacen lo mismo, puesto que ambos casos aparece en pantalla la palabra *Hello*, sin emabrgo, la forma en que ocurre es muy distinta.

#### Función `foo()`
- **Qué hace:** La función `foo()` imprime la palabra "Hello" directamente en la consola cada vez que se llama.
- **Uso:** Es útil si simplemente quieres mostrar algo en pantalla, pero **no puedes usar el "Hello" impreso para otra cosa en tu programa**.
- `foo()` no devuelve **nada**
  
```python
foo()  # Esto muestra "Hello" en la consola.
```


#### Función `bar()`
- **Qué hace:** La función `bar()` devuelve la palabra "Hello" a quien llame a la función.
- **Uso:** Esto es útil si quieres usar el valor "Hello" para otras operaciones o decisiones dentro de tu programa.
- `bar()` **devuelve una cadena**, que podrás usar a posteriori.

```python
mensaje = bar()  # Esto no muestra nada en la consola por sí mismo.
print(mensaje)  # Ahora imprime "Hello", usando el valor devuelto almacenado en `mensaje`.
```

### Diferencia Clave

La diferencia clave aquí es que `foo()` sólo cambia el aspecto del terminal y su "efecto" termina ahí; no puedes hacer nada más con el "Hello" porque no es capturado en una variable ni devuelto. En contraste, `bar()` permite que el valor "Hello" sea capturado y reutilizado, incluso para imprimirlo.


### ¿Quién imprime cuando yo llamo a `bar`?

Cuando llamo a `bar` en una celda de Jupyter, aunque yo no haya llamado a `print`, la cadena devuelta por `bar` (`Hello`), sí que aparece. ¿Quién la está imprimiendo?

Lo hace algo llamado *repl* que conoceremos más adelante.

**REGLA GENERAL: Toda función debería de devolver ALGO.**
