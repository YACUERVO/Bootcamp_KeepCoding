# El mayor error: chulería

----------------

El error más dañino es una mezcla de exceso de confianza en uno mismo y desconfianza en el proceso. Eso puede tirar por tierra todo tu trabajo.

Suele darse en aquellos que *ya saben programar* porque

* *"sé html y css"*
* *"he visto videos en youtube"*
* etc...

Es posible que a estas alturas ya te des cuenta de lo ridículas que son esas afirmaciones. Html y css no son lenguajes de programación y ver videos en youtube es *ocio*, no es esfuerzo, trabajo y aprendizaje. Ver videos en youtube de MMA no te capacita para meterte en una jaula y enfrentarte a Connor McGreggor.

![](mma.jpeg)



Señales inequívocas de que estás cayendo en ese error:

#### Resuelves ejercicios usando cosas que no se han visto aun

> Debes desaprender lo aprendido. Yoda


Suele darse en los que *saben mucho Python y poca Ciberkinesis*. Resuelven un problema esquivándolo, mediante alguna funcionalidad ingeniosa del lenguaje. Acabas de malgastar tu tiempo.

Si algo no se ha enseñado aun, no es por accidente ni por descuido. Aquí no hay accidentes, todo tiene una razón de ser y a menudo, muy buena. 

Confía en el proceso y recuerda el Pacto: *Yo digo, tu haces. Sin preguntas*.

**Nunca resuelvas algo usando cosas que no se hayan visto aun.**


Tomar contacto con la Ciberkinesis, a igual que con la Fuerza,  requiere en todos los casos dejar a un lado muchas de las cosas en las que uno cree y en las que le han educado. Solo así se aprende de manera correcta lo que es de verdad importante.

#### Escribes código innecesariamente corto o complejo

Existe entre los novatos la creencia, equivocada, de que la calidad del código se mide por su tamaño o por la dificultad de compreenderlo. El hecho de que se pueda escribir en una linea, no es una razón para hacerlo. El hecho de que se pueda resolver mediante un truco ingenioso y poco conocido, no es una razón para hacerlo.

Si te pido que resuelvas problemas mediante un técnica concreta, aplicando un patrón computacional específico (compresor, conversor y otros que veremos más adelante), no es por accidente. Es porque tienes que aprenderlo. No hay accidentes, no hay casualidades. Recuerda el Pacto.

No te dejes llevar por la tentación de soluciones cortas e indescifrables. Eso no hace de ti un *programador avanzado*, hace de ti un *novato descarriado*.

Emociones como la prisa y la arrogancia son turbulencias en la Senda del Mago de la Ciberkinesis, así como en la del Jedi. Fija tu vista sólo en la Senda.



## El Programador Humilde


![](dijkstra.jpg)


Este es un extracto del discurso de Edsger Dijkstra (uno de los dioses de la Ciberkinesis) al recibir el Premio Turing (el Nobel de la computación):

-----------

"El programador competente es plenamente consciente del tamaño estrictamente limitado de su propio cráneo; por lo tanto, **aborda la tarea de programar con total humildad** y, entre otras cosas, *evita trucos ingeniosos como si fueran la peste*. 

(...) Esto toma una de dos formas diferentes: un programador coloca un programa de una sola línea en el escritorio de otro y, o bien le dice con orgullo lo que hace y añade la pregunta ‘¿Puedes codificar esto con menos símbolos?’ —*¡como si esto tuviera alguna relevancia conceptual!*—, o simplemente pregunta ‘Adivina qué hace’. 

A partir de esta observación debemos concluir que este lenguaje, como herramienta, es una invitación abierta a trucos ingeniosos; y aunque esto puede ser precisamente la explicación de parte de su atractivo, es decir, para aquellos a quienes les gusta mostrar lo inteligentes que son, lo siento, pero debo considerar esto como una de las cosas más condenatorias que se pueden decir sobre un lenguaje de programación.”

-----------


Si quieres enseñar tu código a alguien para averiguar si puede mejorarlo, pregúntale si puede hacerlo más *claro*, más *correcto*, más *eficiente* o más *flexible*. Y si de verdad, sois grandes maestros los dos, averiguad si podeis hacerlo más *seguro*.

Y si alguna vez al ver tu código alguien tiene que *adivinar qué hace*, tu código es un truño.
