# Homegeneidad en el retorno

----------------------

Para que pueda ser usada, una función debe **devolver siempre el mismo tipo**: debe de ser *homogenea*.

La razón es muy fácil de entender si pensamos en una analogía: imagínate que te comprar un transformador que pasa el voltaje de 220V a 110V. ¿Te parecería sensato que a veces, muy de vez en cuando, pero a veces, devolviese la corriente a 560V? ¿Te atreverías a conectar algún electrodoméstico a él?


Por la misma razón, las funciones deben de **devolver siempre el mismo tipo**.



```python
def inadmisible(n):
    if n >= 0:
        return n + 59
    else:
        return "Hola holita, vecinito"
```

## ¿Si es tan malo, por qué puedes hacerlo?

Puedes hacerlo en Python, aunque sea horrible, porque es una consecuencia de una filosofía de diseño de Python. Es una casualidad, no un objetivo del lenguaje.

## ¿Cuando y por qué sientes la tentación de hacerlo?

Porque no sabes *cómo gestionar errores*. Cuando te encuentras con una *función parcial*, aun no sabes cómo actuar cuando recibe un valor que no es capaz de procesar.

Vamos a ver algunos casos de *funciones parciales* y cómo actuar con ellas.

### Valor que causa un error

Por ejemplo, la siguiente función, que cuando recibe un cero provoca un error:

```python
def inv(n):
    return 1 / n
```

Si al recibir un cero, provoca un error, **¡deja que salte ese error!**. Es lo correcto y más adelante veremos como actuar cuando al llamar a una función, salta un error.

De ninguna manera debes de devolver un mensaje de error (una cadena) en caso de recibir un valor inadecuado.

```python
def inv_indamisible(n):
    if n == 0:
        return "No puedes pasar un cero, burro!"
    else:
        return 1 / n
```

Aunque es correcto el avisar que algo ha salido mal, **esa no es la forma**. Lo correcto es lanzar un erro, como está haciendo Python en este caso de forma automática.

Lo único que sí debes de hacer, es documentar que no acepta el valor de 0.

```python
def inv(n):
    """
    Devuelve el inverso. Si recibe 0, lanza un error.
    """
    return 1 / n
```

### Valor que no causa un error, pero genera resultados sin sentido

Imagínate una función que pasa de millas a km:

```python
def miles_to_km(miles):
    return miles * 1.60934
```

¿Tiene sentido pasar un *valor negativo*? En absoluto, pero no causará un error, simplemente la función responderá con algo sin sentido.

¿Qué hacer en ese caso?

**Buscar un valor dentro del tipo de retorno que indique el error**

Podríamos buscar dentro del tipo del retorno (un número decimal), un valor que sirva para indicar que algo ha salido mal, pero sin romper la homogeneidad de la función.

Por ejemplo, en este caso, podríamos tal vez devolver `0` cuando nos pasan un valor negativo.


```python
def miles_to_km(miles):
    """
    Pasa de millas a km. Si recibe un valor negativo, devuelve 0
    """
    result = 0
    if miles > 0:
        result = miles * 1.60934
    return result
```

Sólo hay un problema: ¿como distingo los dos casos?

* `miles_to_km(0)`: valor trivial, pero correcto
* `miles_to_km(-145)`: valor erroneo

Es necesario que los casos excepcionales y erroneos se distigan claramente de los correctos.

**Devolver `None`**

Si no es posible encontrar un valor de retorno del tipo correcto y que jamás se pueda confundir con los valores delf uncionamiento normal, una opción es devolver `None` en caso de error. 

`None` es el único valor del dipo `NoneType` y sirve para representar *la nada*, o la ausencia de un valor correcto.

De esta forma, la función quedaría así:

```python
def miles_to_km(miles):
    """
    Pasa de millas a km. Si recibe un valor negativo, devuelve None
    """
    result = None
    if miles >= 0:
        result = miles * 1.60934
    return result
```

Este *approach* tiene la ventaja de distinguir los casos correctos de los excepcionales. Tiene la desventaja de que *rompe la homgeneidad*, pero el uso de `None` es una **ruptura mucho menor** que devolver cualquier otro tipo (como pueda ser una cadena). 

La razón por la cual devolver `None` en caso de error es más aceptable, se sale del ámbito de este curso.
