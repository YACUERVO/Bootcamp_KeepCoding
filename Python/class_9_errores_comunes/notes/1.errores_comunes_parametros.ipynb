# Alterar los parámetros

------------------

Otro error muy común y dañino, es el alterar o intentar alterar lo parámetros de una función.

Supón que le prestas tu coche, recien comprado a un amigo para que haga un recado. Te lo devuelve todo tuneado y con una de esas fundas horrendas de bolas en el asiento.

![](hortera.png)

Sería absolutamente inaceptable. No sólo porque es una horterada, pero sobre todo porque el coche no era suyo y sólo estaba prestado.


**Lo mismo ocurre con los parámetros**

Cuando tu función los recibe, **no son suyos**. No sabes que quién pertenecen esos valores ni qué se va a hacer con ellos. Sólo te los han prestado. Es de sentido común que **no lo alteres de ninguna manera**.

Por ejemplo, en esta función:

```python
def add(a, b):
    a = a + b # ya la has cagado
    return a
```

Aunque es cierto que los lenguajes (Python y demás), toman medidas para que impedir que cambies de *forma definitiva* los parámetros (al salir de la función, Python limpia el coche y tira a la basura la funda de bolas y la cintita de las medidas del Pilar que has colgado del retrovisor), puede que algo se le escape. Por si acaso, no lo hagas nunca.

**Nunca cambies el valor de un parámetro**

Crea una nueva variable, es gratis.
