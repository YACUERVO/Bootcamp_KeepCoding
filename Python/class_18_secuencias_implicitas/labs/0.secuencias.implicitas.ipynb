## (0)

Crea una función `sum_all` que recibe 3 parámetros:
* Un valor de inicio
* Un valor final
* Un valor de paso

y devuelve la suma de todos los valores de inicio a fin  tomados de paso en paso.

Es decir,

`sum_all(1, 10, 2)` sumaría 1,3,5,7 y 9 y devuelve = 25

Recuerda usar `range`

````python
def sum_all(start,stop,step):
    result = 0
    for n in range(start,stop,step):
        result = result + n
    return result
print(sum_all(1,10,1))
````

## (1)

Crea la función `factorial(n)` que recibe un número y devuelve su factorial.

Recuerda que el factorial (representado con un !) de un número n es la multiplicación de todos los números enteros de 1 hasta n:

$$4! = 4 * 3 * 2 * 1$$

sabiendo que $1! = 1$ y $0! = 1$

Hazlo usando `range` yendo de 1 hasta `n`.

````python
def factorial(n):
    result = 1
    for numero in range(2 , n+1):
        result *= numero
    return result

print(factorial(5))
````

Según él, tu función *no está mal*, pero hace trabajo innecesario:

Imagina que estás buscando pares de números que, al multiplicarse, formen un número grande, como 36. Estos pares se llaman "factores" y siempre vienen en dos partes: una más pequeña y otra más grande. Por ejemplo:

- $ 1 \cdot 36 $
- $ 2 \cdot 18 $
- $ 3 \cdot 12 $
- $ 4 \cdot 9 $
- $ 6 \cdot 6 $
--------
- $ 9 \cdot 4$
- $ 12 \cdot 3$
- $ 18 \cdot 2$
- $ 36 \cdot 1$

¿Notas algo? Una vez que llegamos a $ 6 \cdot 6 $, hemos alcanzado el medio. Después de eso, los pares simplemente se invierten: $ 9 \cdot 4 $, $ 12 \cdot 3 $, y así sucesivamente.

El punto donde se produce esa inversión es $\sqrt{n}$, en nuestro caso, $\sqrt{36} = 6$

Cuando estás comprobando si un número como $ n $ es primo, realmente estás preguntando: "¿Tiene algún factor distinto de $ 1 $ y de sí mismo?" Si lo tiene, no es primo. Aquí es donde entra la idea de que solo necesitas revisar hasta la raíz cuadrada de $ n $:

Si hay un factor grande de $ n $, como $ 18 $ en nuestro ejemplo, su factor compañero (como $ 2 $) tiene que ser más pequeño que la raíz cuadrada. Una vez que revisas todos los números más pequeños o iguales a la raíz cuadrada, habrás encontrado todos los pares posibles. Si no encuentras ningún factor hasta ese punto, no encontrarás ninguno después.

Por ejemplo, para un número como $ n = 36 $:
- La raíz cuadrada de $ 36 $ es $ 6 $. Entonces, solo necesitas revisar $ 1, 2, 3, 4, 5, 6 $.
- Si ninguno de estos divide a $ n $ (excepto $ 1 $ y $ 6 $), $ n $ es primo.

Para un número que no tiene una raíz cuadrada entera, como por ejemplo $n = 23$:

- La ríz cuadrada de $23$, $\sqrt{23} = 4.7958315233$. Entonces sólo necesitas probar $1, 2, 3, 4$. $4$ es el entero resultante de redondear $\sqrt{23}$.
- Si ninguno de estos divide a $23$, éste es primo.

Este atajo funciona porque los factores siempre vienen en pares. Revisar más allá de la raíz cuadrada es como buscar pares que ya encontraste—¡no es necesario hacer trabajo extra!

1. Modifica tu función para que no comprueba más números de los indispensables.
2. Comprueba que sigue funcionando.


````python
from math import sqrt
def is_prime(number):
    bolean = True       
    if number <= 1:
        bolean = False
    else:
        for num in range(2, int(sqrt(number)) + 1):
            if number % num == 0:
                bolean = False
                break
    
    return bolean   

print(is_prime(9))
````

## (7)

Crea la función `rev_list` que recibe una lista y la devuelve invertida:

`rev_list([1,2,3,4]) == [4,3,2,1]`


Pista: Se trata de un conversor y puedes usar `range` con el tercer parámetro para iterar por los índices hacia atrás.

Asegúrate que comple los siguientes `asserts`

````python
def rev_list(list_numbers):
    list_new = []
    for number in range(len(list_numbers)-1,-1,-1):
        list_new.append(list_numbers[number])
    return list_new
print(rev_list([1,2,3]))
````

````python
assert rev_list([]) == []
assert rev_list([1,2,3,4]) == [4,3,2,1]
assert rev_list(rev_list([1,2,3,4])) == [1,2,3,4] # ¿Entiendes este último?
````

## (8)

Crea la función `rev_str`, que hace lo mismo, pero para cadenas:


`rev["Hola"] == "aloH"`

Recuerda crear los `assert`s correspondientes.

````python
def rev_str(list_strings):
    list_new = ''
    for str in range(len(list_strings)-1,-1,-1):
        list_new += list_strings[str]
    return list_new

print(rev_str('hola'))
````
