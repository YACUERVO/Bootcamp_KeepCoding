*Klaatu* desembarca. 

Vamos a empezar a interactuar con un lenguaje de programación, en este caso Python, y hay que decidir por donde empezar.

Mi situación es similar a la siguiente:


## La visita del marciano

Supón que estás un día en el jardín de tu casa, ya tu lado se posa un platillo volante. El alienígena, en vez de pedir que lo *conduzcas a tu lider*, hace una pregunta desconcertante:

👽 *¿Qué es una frase?*

*Una frase es un conjunto de palabras, separadas por espacios en blanco*

👽 *Ah, bien... ¿y qué es una palabra?*

*Pues una palabra es un conjunto de caracteres, ninguno de los cuales puede ser un espacio en blanco*

Ya empiezas a estar irritado, porque en el fondo sabes cual será su próxima pregunta.

👽 *¿Qué es un caracter?*


Cada vez que te intentas explicar, tienes que mencionar un concepto más sencillo, que inmediatamente el alienígena te pide que definas. Parece la historia de nunca acabar.


Para evitar este tipo de diálogo de besugos, nosotros vamos a empezar por lo más sencillo posible: los elementos *atómicos* de un lenguaje de programación. A partir de ahí, iremos subiendo de complejidad.
