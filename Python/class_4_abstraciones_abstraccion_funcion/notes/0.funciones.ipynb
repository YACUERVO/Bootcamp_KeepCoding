## Sintaxis para definir una función

Al ser algo tan común, todos los lenguajes poseen una sintaxis especial para definir funciones. En Python el patrón básico es 
el siguiente:

```Python
def ___(___):
    ___
```

nada: están reservadas, como el nombre indica para uso exclusivo del lenguaje.

### Los 3 huecos de la definición


```Python
def ___(___):
    ___
```

Vemos que hay tres huecos, indicados por `___`.  

1. El primero se usa para el nombre de la función,
2. El segundo es una lista opcional de parámetros de la función
3. El tercero, es el cuerpo de la función y pueden ser varias lineas de código. Aquí es donde ponemos las *expresiones* que queremos reaprovechar mediante el nombre de la función. El *cuerpo de la función* siempre está **indentado**.


### Función `perimeter` versión 1.0

Vamos a reaprovechar la expresión que habíamos usado para calcular el perímetro de un círculo creando una función para ello. 

Teníamos una expresión para calcular el perímetro de un círculo de radio 10:

`2 * 3.1415926 * 10`

Vamos a transformar esto en una función.


````python

def perimeter():
    return 2 * 3.1415926 * 10

````


Nuestra función:

* tiene un *nombre*: `perimeter`
* tiene una *lista vacía de parámetros* (no tiene parámetros): `()`
* su *cuerpo* tiene una sóla linea que empieza con `return` y sigue con una expresión: `return 2 * 3.1415926 * 10`




#### Nombre de la función

Las mismas reglas que se aplicaban a los nombres de variables, se aplican a los nombres de la funciones.El nombre debería ser claro y explicar qué hace dicha función. 

> Ve acostumbrándote a poner los nombres (variables, funciones y lo que sea) en Inglés


### Las funciones son valores

Si intentamos evaluar la función recién creada, vemos un comportamiento similar a cuando intentamos evaluar un número o una cadena de texto: se evalúa a si mismo.

````python
perimeter
````

La representación cambia un poco:

* no me muestra el código que he escrito
* me muestra un texto diciendo que es una función: `<function __main__.perimeter()>`

Sin embargo, lo más importante es que **una función se evalúa a sí misma**. Dicho de otro modo, **las funciones son valores**, al igual que los números o las cadenas.

En la mayoría de los lenguajes modernos, Python incluído, *las funciones son valores*, al igual que los números o las cadenas de texto. Esto tiene implicaciones enormes, que ahora mismo no podemos entender, pero que sepas que es algo **muy bueno**.


Sin embargo, nadie se toma la molestia de crear una función para evaluarla directamente. Lo que que queremos es **llamarla**.


###

### Las funciones se pueden llamar


Cuando creamos una función lo hacemos para poder reaprovechar el código que hay en su cuerpo. Eso lo logramos **llamando a la función**.

La sintáxis para llamar a una función es la siguiente:

```
nombre_de_la_funcion(lista_de_parámetros)
```

En nuestro caso, como la función `perimeter`no tiene parámetros, sería los paréntesis vacíos.


````python
perimeter()
````

Aquí vemos que ocurre algo totalmente distinto a lo de antes. En vez de evaluarse a sí mismo, se evalúa a un número: hemos logrado que se evaluáse el cuerpo de la función.
Es decir, evaluar `perimeter()` es equivalente a evaluar `2 * 3.1415926 * 10`.

Es muy importante entender la diferencia entre *evaluar la función* y *llamar la función*. EN el segundo caso, estamos evaluando el *código contenido dentro de la función*.

> Una función es un contendor de código que tiene un nombre

Una cosa es una olla, y otra muy distinta es el cocido madrileño que haces dentro de ella. Son cosas distintas y se hacen cosas diferentes con ellas; no puedes comerte la olla, ni puedes lavar el cocido.

* la función es la olla
* el código del cuerpo de la función es el cocido.

### La palabra clave `return`


La palabra clave `return` en los lenguajes de programación tiene una función crucial: termina la ejecución de una función y especifica el valor que será devuelto a quien llamó a dicha función.  Es una herramienta fundamental en la programación que permite a las funciones comunicar el resultado de su computación.

Una función siempre devolverá el resultado de evaluar la expresión que sigue a `return`. 

No tiene sentido poner código después de un `return`, ya que jamás será evaluado y muchos lenguajes lo considerarán un error.

La linea del `return` debe de ser siempre la última de una función, puesto que es su *puerta de salida* por donde sale la información qu eha generado.

### Función `perimeter` versión 2.0

Vamos a modificar la función para que su cuerpo tenga varias lineas de código, es decir, varias expresiones.



````python
def perimeter():
    pi = 3.1415926
    result = 2 * pi * 10
    return result
````

Hemos creado dos variables y devolvemos una de ellas. Dentro de una función podemos crear el código que queramos, no es
más que *un trozo de código con nombre (y puerta de salida)*.

Podemos crear variables e incluso ¡crear otras funciones!

Esto último, es algo un poco más avanzado y lo veremos más adelante.

### Función `perimeter` versión 3.0

Nuestra función sólo sirve para calcular el perímetro de un círculo que tenga 10 de radio. ¿Pero y si tiene otro radio?

Para eso están los parámetros. Para que ciertos identificadores dentro del cuerpo de la función tengan valores distintos según como se llama
a la función.



````python
def perimeter(radius):
    pi = 3.1415926
    result = 2 * pi * radius
    return result
````

Esto convierte la función de un *trozo de código* en una **plantilla de código** con huecos (parámetros) que se rellenan cuando se llama a la función.

````python
perimeter(19.4)
````

> Las funciones son plantillas de código con huecos (parámetros) que se rellenan cuando se les llama. Tienen un nombre y tienen una *puerta de salida* que 
> permite comunicar información con el mundo exterior (quien le haya llamado).

### Parámetros

Puede ser 0 (poco común), 1 o varios parámetros, también llamados argumentos.

* Son *variables* que sólo valen dentro de la función. 
* Desde fuera **nadie** los ve.

**Función sin parámetros** (poco común y poco útil)
```python
    def f():
        return 42
```

**Función con varios parámetros** (van separados por comas)
```python
def add(a,b):
    return a + b
```

### Documentación de la función

Es opcional pero de muy buen gusto. Se trata de una `cadena de texto` con comillas triples (para poder meter varias lineas) que explica lo que hace la función. 

> Documenta siempre tu código


````python
def perimeter(radius):
    """
    Calcula el perímetro de un círculo a partir de su radio.
    No comprueba que el radio sea positivo.
    """
    pi = 3.1415926
    result = 2 * pi * radius
    return result
````

### Funciones sin parámetros

Aparentemente las funciones sin parámetros no sirven para nada. No es del todo cierto y sí que tienen una enorme utilidad. Tanto es así, que 
existe un nombre para dichas funciones: *thunk*. Su uso sin embargo, es para cosas más avanzadas de lo que vemos aquí.

Si quieres, averigua el origen y significado del nombre *thunk*. Aprenderás algo sobre la historia de la programación y la cultura hacker.

## Funciones como patrón universal de computación

Las funciones son un elemento universal de computación y existen y son muy usadas en todos los lenguajes de programación. La sintaxis es siempre muy parecida, siempre hay un nombre, parámetros y un *trozo de código* que abstrae y reutiliza (el cuerpo de la función).

Veamos la función `perimeter` en otros lenguajes:


**Javascript**

```javascript
function perimeter(radius) {
    const pi = 3.14159;
    let perimeter = 2 * pi * radius;
    return perimeter;
}

```
**Swift**

```swift
func perimeter(radius: Double) -> Double {
    let pi = 3.14159
    let perimeter = 2 * pi * radius
    return perimeter
}
```

Aunque no entendamos algunos detalles, el patrón general es evidente. Aprende bien a manejar las funciones y tendrás medio camino andado.


## Uso del Inglés 

¿Por qué a nuestra función la hemos llamado `perimeter` y no `perímetro`?  

Usar el inglés para nombres de variables, funciones, y en general para todos los identificadores en la programación es 
una práctica recomendada por varias razones importantes:


- **Universalidad**: El inglés es el idioma universal en la programación, lo que hace el código accesible globalmente.
- **Colaboración**: Facilita el trabajo conjunto de desarrolladores de distintos países al ser un idioma común en la industria.
- **Herramientas y Soporte**: La mayoría de las herramientas de desarrollo, documentación y recursos están en inglés, simplificando la búsqueda de soluciones y el acceso a soporte comunitario.
