# Variables y Constantes

------------

En muchos lenguajes existen dos formas de *reaprovechar valores*:

* variables
* constantes


### Constantes

Las *constantes* se usan para aquellos casos en los que sabemos que el valor en cuestión no va a cambiar a lo largo de la ejecución del programa. Un buen ejemplo es `PI`. Es algo que jamás cambiará.

Cuando definimos algo como una constante, el propio lenguaje se encarga de que no lo cambiemos por accidente.

Por ejemplo, cuando en javascript escribo este código:

```javascript
function perimeter(radius) {
    const pi = 3.14159;
    let perimeter = 2 * pi * radius;
    return perimeter;
}
```

la linea `const pi = 3.14159;` **avisa al compilador** que `pi` jamás va a cambiar y si se intenta cambiar, debe de avisar de inmediato porque se trata de un error.



### Constantes en Python

En Python **no existen las constantes**.  Es decir, no hay ninguna forma de *avisar al compilador* que algo no va a cambiar jamás.

Esto no es un defecto o un *olvido*. Es un decisión de diseño del lenguaje.

Aunque no podemos avisar al compilador que algo no cambiará, **sí que podemos avisar al programador**.  Esto último se hace mediante una convención de nomenclatura.

**En Python las constantes se escriben en mayúsuculas**

Cuando veas un identificador todo en mayúsculas, eso quiere decir que es algo constante.

```python
def perimeter(radius):
    PI = 3.14159 # es algo constante: mejor no lo cambio
    perimetro = 2 * PI * radius
    return perimetro
```
