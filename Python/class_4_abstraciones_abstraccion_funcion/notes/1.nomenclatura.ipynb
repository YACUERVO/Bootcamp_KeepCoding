### El error de todo novato

En Computación tenemos el mismo desafío: poner buenos nombres a variables, funciones y demás.  Es más dificil de lo que parece, como bien dice la cita de Phil Karlton, y **muchísimo más importante de lo que piensas**.

¿Cual es el error que todo novato comete y que hace que su código cante a léguas? No *dedicar tiempo a poner buenos nombres a sus identificadores*. Tenemos que pensar que estamos poniendo un mote de escuela a esa función, un mote de esos que por ser tan acertado te persigue hasta la tumba y más allá.



### Poner buenos nombres

Elegir buenos nombres para tus identificadores es **crucial** porque estos nombres sirven como el **primer nivel de documentación de tu código**. Un buen nombre puede decirte "qué" hace una variable o función, e incluso "por qué" existe, sin necesidad de leer y entender el código detalladamente. 

Esto es especialmente importante para mantener el código comprensible no solo para ti, sino también para otros programadores que puedan trabajar en tu proyecto en el futuro.

Veamos algunos ejemplos de nombres inadecuados para variables, parámetros o funciones, cómo se deberían corregir, y por qué.

1. **Nombre inadecuado**: `x`
   - **Corrección sugerida**: `total_ventas`
   - **Razón**: El nombre `x` no proporciona ninguna información sobre lo que representa la variable. `total_ventas`, por otro lado, indica claramente que la variable guarda el total de ventas.

2. **Nombre inadecuado**: `a1`, `a2`
   - **Corrección sugerida**: `precio_producto`, `cantidad_producto`
   - **Razón**: `a1` y `a2` son nombres que no comunican el propósito de las variables. Nombres como `precio_producto` y `cantidad_producto` hacen explícito lo que cada variable está almacenando.

3. **Nombre inadecuado**: `funcion1()`
   - **Corrección sugerida**: `calcular_total()`
   - **Razón**: `funcion1()` no indica qué hace la función. `calcular_total()`, en cambio, deja claro que la función se encarga de calcular un total de alguna operación.

4. **Nombre inadecuado**: `data`
   - **Corrección sugerida**: `datos_usuario`
   - **Razón**: Aunque `data` es ligeramente más descriptivo que los ejemplos anteriores, sigue siendo muy genérico. `datos_usuario` especifica qué tipo de datos está manejando la variable.

5. **Nombre inadecuado**: `procesar()`
   - **Corrección sugerida**: `procesar_pago()`
   - **Razón**: `procesar()` es un nombre muy genérico que no aporta información sobre lo que hace la función. `procesar_pago()`, por el contrario, indica claramente que la función está relacionada con el procesamiento de pagos.


### Normas generales

* **Descriptivo**: Un buen nombre es descriptivo y preciso. Un nombre debe de reflejar claramente lo que hace la función o variable, sin ser demasiado corto ni demasiado largo. Usa el sentido común y piensa que le estás poniendo un mote para toda la vida.
* **Funciones**: Las funciones hacen cosas, transforman datos. Por lo general, es una buena idea usar *verbos* para las funciones. Por ejemplo `filtrar_pedidos`, `eliminar_usuarios`, etc...
* **Variables**: Las variables representan algo estático y no un proceso como es el caso de las funciones. Es una buena idea usar *sustantivos* para las variables. Por ejemplo: `producto_seleccionado`, `total_ventas`, etc...
* **Evita Abreviaturas Confusas**: Las abreviaturas pueden hacer tu código innecesariamente difícil de entender.
* **Evita Números o Letras Sin Sentido**: Nombres como `variable1`, `funcion2` no aportan valor y dificultan la comprensión del código.

Siguiendo estas prácticas, contribuirás a que tu código sea más legible, mantenible, y comprehensible tanto para ti como para otros desarrolladores.

**No lo olvides nunca: tu primera responsabilidad es crear código correcto y comprehensible. Si eso falla, nada más importa.**

No serás mejor programador por escribir código hiper corto y dificil de entender, sino todo lo contrario.

Como bien dijo **Terry Davis**, *An idiot admires complexity, a genius admires simplicity*.  Se un genio, hazme el favor.

### Convenciones

Además del nombre en sí, tenemos que decidir cómo representarlo, especialmente cuando está compuesto por varias palabras. Hay básicamente dos convenciones:

**CamelCase**

La *conveción del camello* es una convención de nomenclatura en la que la primera letra de la primera palabra está en minúsculas y la primera letra de cada palabra subsiguiente comienza con mayúscula, sin espacios entre ellas.

No es la habitual en Python.

```Python
def calcularEdad(annoActual, annoNacimiento):
    return annoActual - annoNacimiento

edadUsuario = calcularEdad(2024, 1990)
print(edadUsuario)
```


**snake_case**

 Es otra convención de nomenclatura en la que las palabras se separan con un guión bajo (_) y todas las letras están en minúsculas. Es la convención de nomenclatura preferida en Python, especialmente para nombres de variables y funciones.

 Ésta es la que usaremos nosotros.

 ```Python
 def calcular_edad(anio_actual, anio_nacimiento):
    return anio_actual - anio_nacimiento

edad_usuario = calcular_edad(2024, 1990)
print(edad_usuario)
 ```



### Uso del Inglés


Usar el inglés para nombres de variables, funciones, y en general para todos los identificadores en la programación es 
una **práctica recomendada** por varias razones importantes:


- **Universalidad**: El inglés es el idioma universal en la programación, lo que hace el código accesible globalmente.
- **Colaboración**: Facilita el trabajo conjunto de desarrolladores de distintos países al ser un idioma común en la industria.
- **Herramientas y Soporte**: La mayoría de las herramientas de desarrollo, documentación y recursos están en inglés, simplificando la búsqueda de soluciones y el acceso a soporte comunitario.


```Python
def calculate_age(current_year, birth_year):
    return current_year - birth_year

user_age = calculate_age(2024, 1990)
print(user_age)
```


A partir de ahora, usaré siempre el Inglés en el código de ejemplo, y recomiendo que hagas lo mismo.
