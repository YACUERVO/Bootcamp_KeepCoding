### (0) Función area

Crea una función llamada `area` que recibe el radio de un círculo y calcula su area. Recuerda que la fórmula es:

$$ \text{Área} = \pi r^2 $$

Define `pi` como una variable dentro de la función.

-----------------------------

````python
def area(radius):
  """
  Calcula el area de un círculo a partir del radio.
  No comprueba que el radio sea mayor que cero
  """
  pi = 3.141592611
  return pi * radius ** 2
````

### (1) Función volume

Crea una función llamada `volume` que recibe el radio de una esfera y calcula su volumen. Recuerda la fórmula:

$$ V = \frac{4}{3} \pi r^3 $$


Define `pi` como una variable dentro de la función.

--------------------------


````python
def volume(radius):
  """
  Calcula volumen de un esfera
  No comprueba la validez del parámetro
  """
  PI = 3.141592611
  return PI * (4 / 3) * radius ** 3
````

### (4) Fahrenheit a Celsius

Crea la función `fahrenheit_to_celsius` que recibe una temperatura en Fahrenheit y lo convierte a Celsius. La fórmula es la siguiente:

$$ C = \frac{5}{9} (F - 32) $$

-------------------

````python
def fahrenheit_to_celsius(Fahrenheit):
    celsius = 5/9*(Fahrenheit - 32)
    return celsius

````

````python
fahrenheit_to_celsius(50)
````

De este hecho aparentemente anodino, podemos sacar dos cosas de enorme importancia:

### (10) Azúcar en sangre

El azúcar en sangre, o glucosa en sangre, es la cantidad de glucosa presente en el torrente sanguíneo. Se mide comúnmente en dos unidades: milimoles por litro (mmol/L) y miligramos por decilitro (mg/dL).
* mmol/L (milimoles por litro) es la unidad del sistema internacional y se usa en muchos países.
* mg/dL (miligramos por litro) es la unidad más común en países como EE.UU, pero también en España.



**¿Qué es un mol?**

Un **mol** es un nombre propio para un número (muy grande). De la misma forma que una *docena* es un nombre para 12 cosas, un *mol* es un nombre para un número muy grande: $6.022 \times 10^{23}$.

En química, un mol es una unidad fundamental que representa una cantidad específica de un compuesto químico. Se define como la cantidad de ese compuesto que contiene exactamente $6.022 \times 10^{23}$ entidades elementales (átomos, moléculas o iones). Este número se conoce como **Constante de Avogadro**.

Para entenderlo mejor, pensemos en un mol como una "docena", pero en lugar de 12 unidades, contiene $6.022 \times 10^{23}$ unidades.

**Ejemplos:**
- 1 mol de moléculas de glucosa $C_6H_{12}O_6$ contiene $6.022 \times 10^{23}$ moléculas de glucosa.
- 1 mol de oxígeno molecular $O_2$ tiene $6.022 \times 10^{23}$ moléculas de oxígeno.

**¿Cómo se relaciona con la glucosa en sangre?**

En el caso de la glucosa en sangre, la concentración se mide en **milimoles por litro (mmol/L)**. Un **milimol (mmol)** es simplemente **una milésima parte de un mol**.

Si un análisis de sangre indica **5.5 mmol/L de glucosa**, esto significa que en **cada litro de sangre hay 5.5 milimoles de moléculas de glucosa**. Como cada mol de glucosa pesa **180.16 gramos**, podemos convertir esta cantidad a **mg/dL** con la siguiente relación:

$$\text{Glucose} \ (\text{mg/dL}) = \text{Glucose} \ (\text{mmol/L}) \times 18.016$$

Este factor de conversión **18** proviene de la **masa molar de la glucosa (\(180.16\) g/mol)** dividida entre **10** para ajustar a las unidades de mg/dL. La *masa molar* es el peso de 1 mol de algúncompuesto químico.

**Ejercicio Python**
Escribe dos funciones en Python para convertir entre **mmol/L y mg/dL**.


1. Define `glucose_to_mgdl(mmol_l)` que reciba un valor en **mmol/L** y lo convierta a **mg/dL**.
2. Define `glucose_to_mmol(mg_dl) que haga la conversión inversa.
3. Prueba las funciones con los valores:
   - **7.99 mmol/L** → debería retornar aproximadamente **144 mg/dL**.
   - **144 mg/dL** → debería retornar aproximadamente **7.99 mmol/L**.

**Ejemplo esperado**
```python
print(glucose_to_mgdl(7.99))  # Debería imprimir aproximadamente 144
print(glucose_to_mmol(144))   # Debería imprimir aproximadamente 7.99
```

------

````python
def glucose_to_mgdl(mmol_l):
    mol = mmol_l * 18.0182
    return mol

def glucose_to_mmol(mg_l):
    mmol = mg_l / 18.0182
    return mmol 

print(glucose_to_mgdl(7.99))
print(glucose_to_mmol(144))
````
