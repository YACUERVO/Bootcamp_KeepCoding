# Velocidad de los registros, RAM y almacenamiento masivo

-----------------

Ya hemos visto que la escala de tamaño de los 3 componentes principales son muy diferentes. En concreto los registros son diminutos si comparados con cualquier de los otros dos.

La escala de velocidades también es muy distinta. Una vez más lo veremos con una analogía con el *mundo real*.

## El oficinista

Una buena analogía para la CPU es un oficinista o administrativo que lleva a cabo tareas sencillas sobre los documentos que le pasan: firmar, sellar, etc.



### Datos en los registros

Un dato que está en un registro, sería como *un documento que está sobre su mesa*. Puede pillarlo de inmediato y procesarlo. No hay un atraso perceptible.

### Datos en RAM

El cargar un dato desde la RAM a un registro, es como si nuestro oficinista se tuviese que levantar e ir a un armario dentro de su despacho para buscar el documento en cuestión. Ya no es inmediato.

### Datos en el SSD

El cargar un documento desde el SSD a la RAM y luego a registro, es como si el dato estuviese en otro despacho dentro del mismo edificio. El oficinista, tendrá que salir, pillar el ascensor, ir al otro despacho, buscar el documento y volver a traerlo. Como tenga que hacer esto muy a menudo, estamos perdidos. Mejor sería dejar una copia en su propio armario (la RAM) y eso es precisamente lo que intenta hacer el sistema operativo.

### Datos en la red

Imagínate que tu programa se tiene que descargar una imagen de internet. ¿A qué equivale eso?

Sería como si nuestro oficinista estuviese en Madrid, y la imagen en Cartagena de Indias. Nuestro heroe se montaría en un burro y iría hasta Cádiz, donde pillaría una caravela para ir a Cartagena de Indias. Y luego volver... ¡Esa es la escala de tiempo! 😲

Por si fuera poco, como te puedes imaginar, en ese viaje puede pasar de todo y es posible que no vuelva, así que habría que mandar a otro.

No estaría de más dejar una copia de los documentos más usados un poquito más a mano, como el armario del despacho de al lado (el ssd o disco duro). Eso es lo que hacen los exploradores al guardar un *caché* en disco duro.

Por eso la mayoría de las aplicaciones web están hechas con lenguajes relativamente lentos, como Python o Javascript. No ganas nada en tener un oficinista muy rápido si tiene que andar en burro y velero todo el rato: el cuello de botella será siempre el acceso a la red.

### Mantener los datos lo más cerca de la CPU

Si conseguimos mantener los datos cerca de la CPU, ganamos muchísimo. Por eso las CPUs modernas tienen una capacidad llamada *vectorización*. Se trata de la habilidad de cargar varios elementos de una lista en registros y aplicar una misma operación sobre todos ellos a la vez. A esto también se le llama SIMD: *Single Instruction, Multiple Data*.

¿Donde se puede usar esto? Por ejemplo en un *conversor*, donde se aplica una misma operación sobre todos los elementos de la lista. Esto es muy beneficioso al procesar grandes cantidades de datos.

Por lo general, no es algo de lo que te tengas que preocupar, ya que lo hace la CPU, el lenguaje y el sistema operativo.

En el caso de Python, el SIMD lo hace una framework usada para procesamiento de datos llamada [NumPy](https://numpy.org/) que se usa en Big Data, IA y computación científica.

Python y NumPy se usaron por ejemplo para [generar la primera imagen de un agujero negro](https://numpy.org/case-studies/blackhole-image/) o para confirmar la existencia de las [ondas gravitacionales](https://numpy.org/case-studies/gw-discov/) previstas por Einstein en el siglo pasado.
