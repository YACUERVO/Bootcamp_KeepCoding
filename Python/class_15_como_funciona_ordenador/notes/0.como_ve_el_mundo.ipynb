# ¿Cómo ve el mundo la CPU?

--------------------------

La CPU (*Central Processing Unit*) también llamado procesador, es el principal cerébro de un ordenador. Al igual que los dinosaurios, puede tener *cerebros secundarios* para otras tareas especializadas, siendo el más importante de éstos la GPU (*Graphics Processing Unit*).

Aqui vamos a centrarnos en la CPU y su funcionamiento básico.


### La CPU y los datos

La CPU *procesa*, es decir, *modifica* datos. Las modificaciones que hace son súmamente sencillas, mucho más de lo que crees, aunque no vamos a entrar en esos detalles ahora. 

Si la CPU procesa los *datos*, **¿donde se guardan los datos?**

Los datos se almacenan en 3 posibles lugares:

1. Los *registros* de la CPU
2. La memoria (RAM)
3. El *almacenamiento masivo* (el disco duro, un ssd, etc)



### Los registros de la CPU

Los registros son pequeñas celdas de memoria donde se puede almacenar hasta 64 bits. 

64 bits es poquísimo, una auténtica birria. Ahí cabe un número y nada más. De ninguna manera cabe una lista de números. Tampoco cabe una cadena, cabe un caracter ¡y gracias!

**El contenido de sus registros es TODO lo que puede ver una CPU**

Lo único que puede ver y modificar, es lo que está en uno de sus registros. Piensa en lo limitante que es eso.

### La memoria RAM

Son celdas de memoria, parecidas a los registros, pero mucho más lentas y baratas. Precisamente por eso, podemos tener muchas.

Para que te hagas una idea de la diferencia entre 64 bits (un registro) y 16 Gigabytes (memoria habitual en ordenadores hoy en día), lo mejor es una analogía:

* Si 64 bits equivalen a un metro
* 16 Gbytes serían... **¡más de 2 millones de Km!** 😲



### Almacenamiento Masivo

Los antiguos discos duros o los modernos SSD, son similares a la memoria RAM, con algunas diferencias:

* son mucho más baratos
* son mucho más lentos
* no pierden su contenido al cortar la corriente

Por eso se usan para almacenar los datos.

Si un registro (64 bits) equivaliese a un metro, un SSD de 1Terabyte equivalendría aproximadamente a unos 140 millones de km, es decir ¡**la distancia promedio de la Tierra al Sol**!

Las escalas de tamaño de los registros y los demás componentes del ordenador son muy diferentes.

### Cómo se procesa una lista

Supongamos que queremos ejecutar una función como la siguiente:

```python
def find(haystack, needle):
    found = False
    for element in haystack:
        if element == needle:
            found = True
            break
    return found
```

Lo que ocurrirá será lo siguiente:

1. La función se carga del disco duro o ssd a RAM
2. En un registro de la CPU se carga el valor inicial de la variable `found`
3. El `for` va cargando en otro registro el elemento de turno de la lista: `element`
4. La CPU compara los valores de los dos registros
   1. Si coinciden, el resultado de `found` se guarda en un registro especial de retorno (el `return`) y se termina
   2. Si no, se vuelve a cargar otro elemento de la lista en el registro

Así de ineficiente es, pero en compensación es muy rápido

### ¿Qué ve la CPU?

La CPU *sólo puede ver el contenido de los registros* y nada más. Es decir, tiene una visión diminuta del problema en cada momento.
