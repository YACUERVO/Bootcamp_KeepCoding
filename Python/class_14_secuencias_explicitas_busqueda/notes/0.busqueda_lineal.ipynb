# Búsqueda lineal


----------

La búsqueda lineal (o búsqueda secuencial) es un algoritmo de búsqueda simple que examina cada elemento de una lista de manera secuencial, recorriendo los elementos de uno en uno, hasta encontrar el elemento deseado o hasta que se hayan revisado todos los elementos. 

En el mejor de los casos, encontraríamos el elemento en la primera posición. En el peor de los casos, el elemento es el último, o no está: en ese caso tendríamos que recorrer todos los elementos. En promedio, la búsqueda lineal recorre la mitad de los elementos.


## Usos en la vida real

Utilizamos a menudo la búsqueda lineal en nuestra vida. Veamos algunos ejemplos:

* **Buscar un libro en una estantería**: vas repasando todos hasta que das con el que buscas, o llegas al final y a la conclusión que no está.
* **Verificar asistencia**: recorres una lista de nombres para verificar si un alumno en concreto está en clase.
* **Encontrar un contacto**: revisar una lista de contactos en un teléfono móvil para encontrar el número de una persona específica.

Intuitivamente sabes que si la lista es pequeña todo irá bien, pero si la lista es muy grande se puede volver agotador e ineficiente. Se puede mejorar si nos aseguramos de que la lista esté ordenada (¿por qué?), pero a partir de un cierto punto, hay que buscar otra solución.

## Ventajas y desventajas

1. Es muy sencilla y fácil de implementar: basta con un `for`.
2. Funciona muy bien con listas de tamaño razonable.
3. En promedio sólo hay que recorrer la mitad de la lista: si tienes suerte y es el primero, recorres un elemento, si no tienes suerte te toca recorrer toda la lista. El caso promedio está entre ambos extremos: recorres más o menos la mitad de la lista.
1. Para listas muy grandes, se vuelve inmanejable y hay que buscar otra solución.

## Importancia de la búsqueda en la Ciberkinesis

La búsqueda (lineal o no) es una operación extremamente común en Ciberkinesis. Mucho más de lo que puedas creer. Vamos a ver un ejemplo muy común: abres tu explorador, escribes `www.amazon.es` y le das a intro. Esto es lo que ocurre:

1. Tu explorador hace una primera conexión a un servicio online llamado *servidor de DNS*. Dicho servidor actúa como una función que recibe una URL y devuelve una cadena con algo llamado dirección IP. Esa dirección IP es la que permite conectarse a un servidor en internet (los nombres de dominio, como keepcoding.io o google.com son sólo para los humanos, a las máquinas no les sirve de nada).
   1. El servidor de DNS tiene algo así como una lista de dominios y su IP asociada. El DNS **busca** en su lista de dominios a `www.amazon.es`, lo encuentra y devuelve la IP: `108.157.105.110`.
2. Tu browser se conecta a `108.157.105.110` y el servidor de http de Amazon responde con una página web (html, css y js). Junto a tu petición tu browser envía cierta información sobre ti (las cookies) con un *identificador de cliente*.
3. Amazon **busca** tu *identificador de cliente* entre los millones de clientes que tendrá.
4. Amazon **busca** tus últimas compras entre todo lo que hayas comprado y te muestra productos nuevos que te puedan interesar.

Si esto te ha parecido más complejo de lo que esperabas, que sepas que es una versión **muy simplificada** del proceso real. De todas formas, la clave está en ver la importancia y lo común que es **buscar** cosas. 

Algunas de estas búsquedas se realizan sobre volúmenes de datos enormes y no es práctico usar la *búsqueda lineal* en ellos. Más adelante veremos otras formas de buscar que son *infinitamente más rápidas*.
