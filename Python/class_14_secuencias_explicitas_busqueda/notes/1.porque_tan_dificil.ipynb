### Hablarle a un alienígena

Te sorprende la dificultad para *explicarle* tareas triviales, porque **sobreestimas sus capacidades**. Es un error de novato en el fondo, crees que estás tratando con un humano. En realidad se trata de un alienígena. Es torpe, muy torpe, pero rápido. No es como tú.

En la siguiente sección vamos a aprender un poco sobre cómo funciona este alienígena, lo cual nos ayudará a entenderle un poco mejor y cómo tratar con él.
