#### 1 `find_one`

Implementa tú la función `find_one`, usando un `for`. Recuerda lo siguiente:

1. Recibe dos parámetros, el *haystack* (la lista que actúa como pajar) y la *needle* (la aguja o elemento que estamso buscando).
2. Devuelve un booleano.
3. Si la aguja no está en el pajar, devuelve `False`.
4. Si el pajar es la lista vacía, devuelve `False` (en la lista vacía no hay nada).
5. Si encuentras la aguja, ¡**no sigas buscando**! Ya tienes la respuesta que quieres.

Implementa tu función y comprueba que funciona correctamente. Prueba valores que están y valores que no están, así como un pajar vacío. Asegúrate de que tu función tenga una cadena de documentación correcta.

Cuando lo tengas, responde a las siguientes preguntas:

1. ¿Es una función reversible o irreversible?
2. ¿Es una función total o parcial? En caso de ser parcial, ¿qué valor causa problemas problemas y por qué?



```python
def find_one(haystack,needle):
    '''Reciba una lista (pajar) y un numero (aguja)
    devuelve True si encuentras la aguja en el pajar'''
    bolean = False
    for element in haystack:
        if element == needle:
            bolean = True
            break
    return bolean
```

```python
print(find_one([4,5,6,1,2],9))
print(find_one([4,5,6,1,10,11,15],15))
print(find_one([15],15))
print(find_one([15,12,15,14,18,15,15],33))

```

#### 2 `find_n`

Ahora vamos a por `find_n(haystack, needle, n)`. Debes de encontrar `n` instancias de `needle` dentro de `haystack`, en la posición que sea.

**Pistas**

Además de un booleano para saber si has encontrado o no la aguja, vas a necesitar un contador para saber *cuantas* agujas has encontrado.

```python
def find_n(haystack,needle,n):    
    boleano = False
    contador = 0
    for element in haystack:
        if element == needle:
            contador += 1
            if contador == n:         
                boleano = True
                break
    return boleano
```

```python
def find_n_2(haystack,needle,n):    
    contador = 0
    for element in haystack:
        if element == needle:
            contador += 1
            if contador == n:         
                break                
    return contador == n
```

```python
print(find_n([5,4,6,8,5,5,9,2,1,3,5,6,4,4],4,3))
print(find_n_2([5,4,6,8,5,5,9,2,1,3,5,6,4,4],4,3))
print(find_n_2([5,4,6,8,5,5,9,2,1,3,5,6,4,4],4,4))
```

#### 4 `find_streak`

Ahora ya podemos dar el salto final y meterle mano al problema inicial: encontrar n instancias *seguidas* de la aguja dentro del pajar.

**Pistas**

1. La única diferencia entre `find_n` y `find_streak` es que en la segunda, hay una condición más: que estén seguidas.
2. Necesitarás una nueva variable, en este caso un booleano para indicar si estás o no en racha (las agujas están seguidas).
3. Cuando te encuentres con un elemento del pajar,
   1. Si es igual a la aguja, activas el booleano de racha e incrementas el contador de instancias
   2. Si el contador pasa a ser igual o mayor que `n` has terminado (¡no sigas buscando!)
   3. Si no es igual a la aguja, desactiva el booleano de racha y pon a cero el contador

```python
def is_consecutive(string, char):
    bolean = string != ""
    for element in string:
        if element != char:
            bolean = False
            break
    return bolean

```

```python
assert is_consecutive('aaaaa','a') == True
assert is_consecutive('aaaaa','b') == False
assert is_consecutive('aaabcaa','a') == False
assert is_consecutive('','a') == False
assert is_consecutive([1,1,1,1,1],1) == True
```

```python
def find_streak(haystack,needle,streak): 
    contador = 0
    in_streak = False
    for element in haystack:
        if element == needle:
            contador = contador + 1
            if contador >= streak:
                in_streak = True
                break
        else:
            in_streak = False
            contador = 0
    return contador == streak


assert find_streak([1,2,3,2,2,2],2,3) == True
assert find_streak([1,2,3,1,2,3],3,2) == False
assert find_streak([],'A',3) == False
assert find_streak([1,2,3,2,2,2],2,0) == False
assert find_streak([1,2,3,2,2,2],9,0) == True
assert find_streak([1,2,3,9,2,2],9,0) == False

    
        
        

```

```python
print(find_streak([5,4,4,4,5,5,9,2,1,3,5,6,4,4],4,4))
print(find_streak([5,4,6,8,5,5,9,2,1,3,5,6,4,4],4,3))
print(find_streak([5,4,6,8,5,5,9,2,1,3,5,6,4,4],5,2))
```

#### 5 `find_streak` refactorizado

El hacer cambios para mejorar código que ya funciona (es decir, no estás arreglando errores) se le llama *refactorizar*. Pues bien, vamos a refactorizar la función `find_streak`.

Si la has creado usando un booleano para saber si estás en racha o no, y un contador para saber cuantas ocurrencias has encontrado, hay una forma más sencilla: **no necesitas el booleano**, dado que **contador ya tiene toda la información** que necesitas. Piensa un poco y lo verás.



```python
def find_streak(haystack,needle,streak): 
    contador = 0
    for element in haystack:
        if element == needle:
            contador = contador + 1
            if contador >= streak:              
                break
        else:        
            contador = 0
    return contador == streak


assert find_streak([1,2,3,2,2,2],2,3) == True
assert find_streak([1,2,3,1,2,3],3,2) == False
assert find_streak([],'A',3) == False
assert find_streak([1,2,3,2,2,2],2,0) == False
assert find_streak([1,2,3,2,2,2],9,0) == True
assert find_streak([1,2,3,9,2,2],9,0) == False
```
