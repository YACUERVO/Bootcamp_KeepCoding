### Función `decouple`

`decouple` recibe una lista y un predicado y devuelve una lista con dos sublistas. La primera está compuesta por aquellos elementos que han pasado el test del predicado y la segunda contiene aquellos que no lo han pasado.

Por ejemplo, si aplicamos `decouple` a la lista `[1,2,3,4,5,6,7]` con el predicado `lambda x : x % 2 == 0`, obtendremos un *LoL* de longitud 2, con los elementos pares en la primera sublista y los impares en la segunda. Es decir, devolvería `[[2,4,6], [1,3,7]]`.

**Requisitos**

1. La longitud de salida es siempre 2 (es una lista con dos sublistas).
2. Si se aplica a la lista vacía, devuelve `[[], []]`.
3. Si el predicado devuelve `True` para todos los elementos, todos ellos irán a parar a la primera sublista y la segunda estará vacía.
4. Si el predicado devuelve `False` para todos los elementos, todos ellos irán a parar a la segunda sublista y la primera será la vacía.
5. El número de elementos en la lista de entrada y la de salida nunca cambia.
