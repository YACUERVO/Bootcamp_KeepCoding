##### Redescubrimiento I: el *proxy* también conocido como el pringao

En *Lista de Clientes II*, nos hemos inventado un valor nuevo, que no era parte de la lista que representa a un usuario, *simplemente para ordenar*: el valor del bmi. 

Se trata de algo temporal que hemos inventado simplemente para que haga un trabajo que no queríamos hacer o para el que no estábamos capacitados. Se trata de un caso de *Transforma Y Vencerás*, tan común, que tiene nombre propio: **El Patrón de Diseño del Proxy**.

Es muy util y lo verás muchas veces en tu exploración de la Ciberkinesis.

Eso sí, no dejes que el nombre bonito te lleve a engaños, tan solo se trata del viejo truco de endosarle un muerto a un pringao más tonto que tú.

##### Redescubrimiento II: el *adaptador* 

Otra situación común es encontrarnos con sistemas que se tienen que comunicar entre sí pero son incompatibles entre sí. El caso más sencillo es el que hemos visto: dos funciones que *no encajan*. La solución es la que hemos visto y que ya conocemos del mundo real: un adaptador.

A menudo los adaptadores son más complejos que éste, pero la idea es la misma: poner algo entre los dos sistemas incompatibles que adapte uno al otro para que ambos puedan seguir funcionando sin necesidad de cambios.

Esto es aun más común que el *proxy* y aparece mucho a la hora de crear interfaces gráficas, ya que sea para Web, Apps o Aplicaciones de Escritorio. Se llama el **Patrón de Diseño del Adaptador (o Adapter, si quieres ser más fino)**.
