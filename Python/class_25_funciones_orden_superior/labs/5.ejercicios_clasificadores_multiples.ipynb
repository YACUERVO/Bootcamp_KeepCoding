# Clasificadores Múltiples

-----------------------

A menudo necesitamos clasificar datos en más de dos grupos (sí o no). Por ejemplo, supongamos que estamos analizando el correo entrante a un servidor, y usando métodos estadísticos debemos clasificarlo en alguna de las siguientes categorías para que el usuario pueda filtrar:

* **spam**: pastillas azules, príncipes nigerianos, etc...
* **social**: no es spam y proviene de alguna red social
* **actualizaciones**: no es spam y es algún mensaje que te pone al tanto de algún proceso (un envío, una compra)
* **foros**: no es spam y proviene de algún foro
* **otros**: no cae en ninguno de los grupos anteriores

No podemos clasificarlos usando un booleano. Necesitamos algún otro tipo para ello. Al ser algo muy específico, tendremos que crear un nuevo tipo mediante una `Enum`:

`````python
from enum import Enum
from enum import auto # sirve para darle un número cualquiera de forma automática (¿lo pillas?)

class EmailTag(Enum):
    """
    Etiquetas que usaremos para clasificar los emails
    """
    SPAM = auto()
    SOCIAL = auto()
    UPDATES = auto()
    FORUM = auto()
    OTHER = auto()
    
`````

La *clasificación* sería una lista de valores de `EmailTag`.

Otro ejemplo podría ser un juego de turnos (como Ajedrez, Damas, Connecta4, etc) en el cual el ordenador debe de jugar contra un humano. Es vital clasificar las jugadas para saber cuales te llevan a ganar o perder. Antes de jugar, el ordenador se crea una lista de las posibles jugadas y las clasifica en los siguientes grupos:

* **WIN**: si juego aquí, gano de inmediato.
* **LOSE**: como juegue aquí, pierdo de inmediato.
* **BAD**: si juego aquí, no pierdo de inmediato, pero me acerca a la derrota.
* **GOOD**: no gano de inmediato, pero estaré en mejor situación.
* **MAYBE**: no tengo suficiente información.



`````python
from enum import Enum


class GameMoveTag(Enum):
    """
    Etiquetas de las diferentes jugadas. Les pongo valores que indican lo deseables que son. Así podré ordenar la clasificación
    de tal forma que lo mejor sea lo primero.
    """
    WIN = 100
    GOOD = 50
    MAYBE = 10
    BAD = 0
    LOSE = -1
`````

#### Clasificación de BMI

Vas a recibir una lista de BMI (índice de masa corporal) numéricos. Según el valor, deberás clasificarlo en los siguientes grupos:

1. **Bajo peso**: IMC menor que 18.5
2. **Normal**: IMC entre 18.5 y 25 (no incluído)
3. **Sobrepeso**: IMC entre 25.0 y 30.0 (no incluído)
4. **Obesidad grado I**: IMC entre 30.0 y 35.0 (no incluído)
5. **Obesidad grado II (severa)**: IMC entre 35.0 y 40.0 (no incluído)
6. **Obesidad grado III (mórbida)**: IMC mayor a 40.0 y menor que 80 (incluído)
7. **Erroneo**: Cualquier valor fuera de estos índices.

#### Clasificación de BMI II: El BMI Contraataca

Te avisan que dejarás de recibir los BMIs como listas de números, sino que vas a recibir una lista de *usuarios*.

Un usuario se representa como una lista con la siguiente estructura: nombre, apellidos, altura en cm, peso en Kg.

```Python
user1 = ["Fernando", "Rodríguez Romero", 187.0, 84.0]
````

Por lo tanto, la *lista de usuarios* será una *lista de listas* o *LoL*. 

Debes de seguir usando la función `classify`, lo único que tienes que cambiar es la función que categoriza los elementos. Antes cada elemento era un número y se transformaba en un valor de la `Enum` de *BMI*.

Ahora, la función `categorize_user` :

* recibe un *usuario*, es decir una lista con 4 elementos: nombre, apellidos, altura y peso
* devuelve un *usuario categorizado*, una lista con 6 elementos: nombre, apellidos, altura, peso, bmi, categoría

Por lo tanto, la clasificación de los usuarios, será una lista de usuarios categorizados (un LoL). Un ejemplo podría ser:

```Python
classification = [
    ["Fernando", "Rodriguez Romero", 187, 84, 24.0, BMI.NORMAL],
    ["Jabba", "The Hutt", 190, 245, 67.9, BMI.MORBID_OBESE]]
````

**Pistas**

1. Es muy similar a la anterior. La única diferencia es que un elemento ahora es un dato compuesto y tendrás que extraer la info necesaria y luego añadir lo calculado a dicho dato compuesto.
2. Para acceder a los componentes del *usuario* (una lista), no uses los índices directamente. Es mucho mejor usar constantes para ello: `NAME = 0`, `SURNAME = 1`, `HEIGHT = 2`, etc...
3. Recuerda los casos especiales (LoL vacío), la cadena de documentación, y los requisitos
4. No te preocupes por recibir un elemento que no tenga la estructura de un usuario. Ese problema lo resolveremos más adelante.
