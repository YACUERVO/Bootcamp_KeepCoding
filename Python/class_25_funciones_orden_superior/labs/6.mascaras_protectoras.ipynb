Un uso muy común de las *máscaras binarias* (listas de booleanos) es àra indicar sobre qué parte de una secuencia se debe de actuar y sobre cual nó. Es el equivalente ciberkinético a la *cinta de carrocero* que sirve para proteger enchufes y otras partes de una pared a la hora de pintarla.

#### Función `double_with_mask`

Crea la función `double_with_mask` que recibe una lista de números y una máscara de longitud *igual o superior* a la lista de números.

Devuelve una lista de números de la misma lista original, pero con aquellos valores cuyo índice en la máscara es `True`, duplicados.

Es decir,

`double_with_mask([1,2,3,4], [True, True, False, False, True])` devolvería `[2, 4, 3, 4]`.

Casos especiales:

* Si la lista de números es la lista vacía, devuelve la lista vacía
* Si la máscara es más corta que la lista de números, lanza `ValueError`

#### ¿Qué estamos haciendo?

¿Qué tipo de función es `apply_with_mask`? ¿Recuerdas haber visto algo así en el pasado?

Se trata de un *conversor condicional*, que sólo actúa si la máscara se lo permite.

`apply_with_mask` con una `false_mask` sería un *conversor nulo*, que no convierte a ningún elemento (la máscara no se lo permite) y `apply_with_mask` con una `true_mask` sería un *conversor incondicional*, puesto que la máscara siempre se lo permite.

#### ¡¿Qué hemos hecho?!

Vamos por partes.

1. Lo que hemos llamado un *punto* (una lista de 3 números) es en realidad un *pixel*. Se trata de tres valores que representan el color de un punto en una imagen: rojo, verde y azul (RGB o Red Green Blue).
2. La función `mistery` lo que hace es convertir un pixel en color a escala de grises. Los coeficientes usados son los que resultan más adecuados para el ojo humano.
3. La función `secret` convierte los puntos centrales a escala de grises y los extremos a negro (deja un margen negro).

Si en vez de una lista pixels, tuviésemos una matriz de píxels, habrías creado un filtro de imagen que convierte a escala de grises dejando un margen negro en los bordes.



#### Procesamiento de Imágenes con Máscaras

Las máscaras se usan mucho para procesar imágenes, pero limitando dónde se puede aplicar el filtro (donde lo permita la máscara). Cambiando la máscara puedes aplicar efectos distintos a diferentes regiones de la imagen. El invertir una máscara para aplicar un filtro sobre la zona que no ha sido tocada previamente, es una operación común.
