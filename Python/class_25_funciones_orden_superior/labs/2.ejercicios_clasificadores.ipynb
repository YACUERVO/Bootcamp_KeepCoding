# Ejercicios de clasificadores


-----------------------

### Función `is_positive`

Crea el clasificador `is_positive` que recibe una lista de números y devuelve la clasificación de si son positivos o negativos. Considera el `0` como positivo y si recibe la lista vacía, devuelve la clasificación vacía (`[]`).



### Función `is_all_caps`

Crea el clasificador `is_all_caps` que recibe una lista de cadenas y clasifica las que están todas en mayúsculas. Es decir, `"Hola"` se clasificaría como `False` y `"HOLA"` se clasificaría como `False`.

### Función `starts_with(strings: list[str], prefix:str)->list[bool]`

Clasificador que recibe una lista de cadenas y una cadena. Detecta aquellas cadenas en `strings` que empiezan por la cadena `prefix`. Ignora diferencias de mayúsculas y minúsculas.

### Clasificador genérico

Crea la función `classify` que recibe una lista y un predicado y devuelve la clasificación de esa lista según el predicado.
