#### ¿Qué son `and_mask`, `or_mask` y `not_mask` ?


Una *clasificación* o *máscara* son como **booleanos extendidos**, listas de booleanos. El equivalente al `not`, `and` y `or` de los booleanos son las funciones `not_mask`, `and_mask`  y `or_mask` en las listas de booleanos.

![](bools_masks.png)

`bool` y `list[bool]` son tipos relacionados y tienen operaciones parecidas. Esto ya lo hemos visto en el pasado, por ejemplo con los números y las listas de números. De la misma manera que tenemos `+` y `*` para los números, tenemos a funciones como `add_all` y `prod_all` para las listas de números.

El tratar una lista de números (llamada vector) o matrices de números como si fuesen números y definiendo funciones para sumarlos, multiplicarlos y otras funcione similares, es la base de una rama de las matemáticas llamada **Algebra Lineal** que es de importancia vital en buen aparte de la IA. De hecho los *modelos de IA* por detrás los LLMs como chatGPT, no son más que matrices (en mucho más que dos dimensiones) inmensas de números.
