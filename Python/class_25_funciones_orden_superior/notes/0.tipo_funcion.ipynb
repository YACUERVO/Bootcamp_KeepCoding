# Tipo Función

------------------------------

Ya hemos usado funciones para resolver todo tipo de problemas, pero para poder avanzar, ahora debemos dedicar un momento a las *funciones en sí* y el uso que les hemos estado dando.

En Python así como otros lenguajes relativamente modernos (como Swift, Kotlin o Javascript), las funciones son consideradas "ciudadanos de primera clase", lo que significa que se pueden manejar de manera similar a otros tipos de datos como números, cadenas y listas. 

Las funciones son **valores**.


### Tipo `function`
Cuando defines una función en Python, estás creando un valor de tipo `function`. Esto se puede demostrar asignando la función a una variable y utilizando la función `type()`.

```python
def saludar():
    return "Hola, mundo!"

print(type(saludar))  # <class 'function'>
```


Por esto podemos asignar una función a una variable o incluso guardar funciones en una lista:


````python
def add42(n):
    return n + 42

def invert(n):
    return 1 / n

funcs = [abs, add42, invert]

print(funcs)

for func in funcs:
    print(func(42))
````

El que las funciones sean un valor, tiene muchas implicaciones y estamos a punto de estudiar la más útil de todas: **las funciones pueden aceptar otras funciones como parámetros**.

Otra consecuencia, es que una función pueda devolver a otra como su retorno, pero eso lo veremos más adelante.

### Operaciones sobre el tipo función

Sabemos que un tipo se compone de dos cosas:
* las instancias del tipo
* las operaciones que se pueden hacer con ellos

Por ejemplo, 

**tipo `bool`**

* instancias: `True`, `False`
* operaciones: `and`, `or`, `not`

**tipo `int`**

* instancias: todos los enteros posibles
* operaciones: todas las operaciones aritméticas( +, *, -, /, etc)

El **tipo función** también tiene esos dos componentes:

* instancias: todas las funciones que puedas crear
* operaciones: ya habíamos visto que las funciones se *pueden llamar* con el operador `()`, sin embargo hay otras dos operaciones que nos permiten combinar funciones entre sí.

**Composición**

Una forma de combinar dos o más funciones donde la salida de una función se convierte en la entrada de otra. Esto es útil para crear operaciones más complejas a partir de funciones más simples.

**Convolución**

Se trata de otra forma de combinar dos funciones, aunque mucho menos común.

Ambas se usan en Deep Learning.

De momento nos vamos a concentrar en las *funciones que reciben a otras funciones como parámetros*.
