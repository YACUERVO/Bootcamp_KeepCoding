# Selector vs Clasificador

----------------------



Se trata de dos estrategias para hacer básicamente lo mismo: aplicar un predicado a todos los elementos de una colección. En el caso de *selector* el proceso se hace de un golpe, obteniendo de **forma directa** la lista de elementos que pasan el predicado.

Por el otro lado, la estrategia del *clasificador* es **indirecta**: primero obtenemos la **máscara booleana** (o *clasificación*) y luego podemos usarla para obtener la lista de elementos que pasan el test, pero *también se podría usar para muchas otras cosas*.

**¿Cual de las dos estrategias es mejor y cual debes de usar?**

*De lo que quieras hacer, los datos y las herramientas que vayas a usar, todo dependerá. Sí, hmm.*



### Enfoque de Máscara Booleana

Esta es muy común en lenguajes enfocados a procesar grandes cantidades de datos en formato de lista. Se llaman lenguajes enfocados a *arrays* o *vectores* (especies de listas) y el ejemplo más popular hoy en día es [q](https://code.kx.com/q/) de KX Systems. Se utiliza mucho para procesar datos financieros a menudo en tiempo real y es popular en la industria financiera.

En la industria financiera, y otras aplicaciones de Big Data, es frecuente encontrarse con **series temporales**. Las series temporales son como listas de datos a menudo larguísimas *que van llegando de uno en uno al sistema*. Por ejemplo, los datos de precios de una acción son una serie temporal. Cada elemento tiene asociado un momento, fecha y hora y a menudo llegan a tu sistema de uno en uno en tiempo real.

Otro ejemplo serían datos de redes sociales que van apareciendo a medida que son generados, o los datos que recibe el avión sobre el funcionamiento en tiempo real de sus turbinas.

El enfoque de las máscaras es común en este tipo de escenarios.

**Paralelismo mediante Vectorización**

La razón principal por la cual se utiliza la técnica de la máscara booleana para listas enormes de datos, es la posibilidad de aplicar el predicado *en paralelo*: el procesador ¡no tiene que recorrer la lista aplicando el predicado de elemento en elemento!

Existe la posibilidad de aplicar el predicado a *todos los elementos de un golpe*, usando una técnica de la cpu llamada *vectorización*.

la *vectorización* es una funcionalidad de las CPUs modernas que les permite ejecutar ciertas tareas a la vez, en vez de una detrás de la otra.

Ciertos lenguajes, como *q* están diseñados para hacer uso de la vectorización de forma automática.

**Flexibilidad en Operaciones Subsecuentes** 

La máscara booleana puede usarse para realizar múltiples operaciones diferentes una vez calculada. Además de extraer los elementos que cumplen con una condición, la misma máscara puede aplicarse para contar, reemplazar u otras operaciones vectorizadas sin recalcular el predicado.

**Reutilización una vez calculado** Una vez que la máscara booleana está calculada, se puede reutilizar para varios fines sin necesidad de reaplicar el predicado. Esto puede ser especialmente útil si necesitas realizar múltiples operaciones diferentes basadas en la misma condición. Esto puede ser especialmente importante para un predicado inherentemente lento, ya sea por su complejidad, costo computacional o necesidad de acceder a recursos externos.





### Python y demás lenguajes no vectoriales: selector

En Python y los demás lenguajes más generales, se suele optar por la estrategia de la función selectora. Es lo que deberías de usar por defecto, a no ser que estés tratando con un enorme volumen de datos que requieran el procesamiento en paralelo.

Su gran ventaja es que es más directo, especialmente en el caso más común de *seleccionar los elementos que cumplen* con un predicado: los obtienes de un golpe sin tener que pasar por el paso intermedio de una máscara.

Aunque es más lento precisamente por no poder ser paralelizado, su rendimiento es adecuado para listas de tamaño no muy grande.

### ¿Y si tengo que procesar listas enormes en Python?

No hace falta irse a lenguajes tan específicos como `q`. De hecho Python es el lenguaje más usado para procesamiento de datos, pero tenemos que usar *frameworks* específicas para ello, como **Pandas** o **Polars**.

Estas tienen sus propias versiones de las listas de Python (llamadas *dataframes*) que son mucho más eficientes. Más que listas, son como *hojas de cálculo* empotradas. Estos dataframes usan la estrategia de las máscaras booleanas, aplican los predicados en paralelo mediante vectorización y tienen funcionalidades equivalentes a las funciones `and_mask` y `or_mask`, pero mucho más rápidas.

Es decir, podemos usar Python para aquellos casos de uso típicos de lenguajes como `q`, pero usando herramientas extra.


### Conclusión

En contextos donde pueden ser necesarias múltiples manipulaciones diferentes después del filtrado, o donde las operaciones necesitan ser vectorizadas para el rendimiento, el enfoque de máscara booleana de los lenguajes basados en arrays o las frameworks comos pandas o Polars es ventajoso. 

En contraste, el enfoque de filtrado directo en Python a menudo es más conveniente para la programación de propósito general donde la claridad del código y la manipulación directa de datos son prioridades.
