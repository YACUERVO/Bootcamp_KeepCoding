# Funciones de Orden Superior

---------------

![](alto_orden.jpg)

Según nos había contado **Lambda el Supremo**, existe una *Llanura de la Abstración* donde reina el *Cálcula Lambda*. Gracias a ello, las funciones son libres y todo es posible: las funciones pueden ser pasadas como argumentos, retornadas como resultados, y todo interactúa en armonía. "Este es el corazón de nuestro zoológico", nos dijo.

Uno de los habitnates más representativos, son las **Funciones de Orden Superior**. Como el Ying y el Yang, la Luz y la Oscuridad, vienen en dos formatos:

* funciones que reciben otras funciones como parámetros
* funciones que devuelven otras funciones



### Funciones Aplicadoras

A las *funciones que reciben otras funciones como parámetros* les llamaremos **Funciones Aplicadoras**, puesto que consumen funciones y las *aplican* a otros parámetros.

Son enormemente útiles y nos permite subir subir un peldaño más en la Torre de la Abstración y así generalizar cosas que hemos estado tratando como específicas. En concreto, podremos borrar todos los *compresores* y sustituirlos por una sola función, un *comrpesor universal*. Haremos lo mismo con las demás familias de funciones que hemos visto en el tratamiento de secuencias (conversores, selectores, etc), gracias a la magia de las funciones aplicadoras.

Éstas son muy usadas en el procesamiento de datos, programación concurrente y también en la gestión de interfaces de usuario complejas.




### Constructores Funcionales

El otro lado de la moneda, son aquellas funciones que *devuelven una nueva función*. Como fabrican o construyen funciones, les llamaremos **Constructores Funcionales**.

No son de uso tan universal como las **Funciones Aplicadoras**, y requieren algo más de soltura con esta rama de la Ciberkinesis. Por ello, las dejaremos para un poco más adelante.
