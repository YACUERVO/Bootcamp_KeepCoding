# El Adaptador

-------------



Mientras aprendíamos sobre cómo ordenar de forma personalizada, hemos aprendido, entre otras cosas, un concepto súmamente útil y común: **el adaptador**.

![](adapter.jpg)

Un **Adaptador**  es un concepto muy común en Ciberkinesis, y de enorme utilidad. A menudo se le conoce con distintos nombres según el contexto (adaptador, controlador, *middleware*, y hasta *view-model*), pero son todos el mismo perro con el mismo collar.

Un **Adaptador** es un componente sencillo que actúa como intermediario entre dos sistemas (normalmente más complejos, o que no puedes cambiar) que permite que ambos trabajen juntos.

* El adaptador de la imagen es un cacharro de unos pocos euros que permite que el sistema eléctrico norteamericano funcione a la perfección con los electrodomésticos europeos, sin que ninguno de los dos tenga que cambiar en lo más mínimo.
* El casquillo es un cacharro sencillo y barato que permite que el sistem eléctico pueda interactuar con todos los tipos de bombilla que hay hoy en día, sin tener que ocuparse de sus detalles ni cambiar en lo más mínimo (led, incandescente, halógena, lo que sea)
* El gestor de producto (*product manager*) es un encargado de que los departamentos de desarrollo, diseño y negocio de una empresa interactúen entre ellos sin preocuparse por lo que hacen los demás.



## La Ley de Wheeler y los Adaptadores

![](wheeler.jpg) David Wheeler fue un pionero británico de la Ciberkinesis. A él le debemos lo que por aquel entoces se llamaba la *subrutina cerrada*. Traducido al idioma actual, quiere decir que inventó el mecanismo mediante el cual se llaman a las funciones y éstas devuelven un valor. Casi ná, el amigo David.

A él tambien se atribuye una ley, la **Ley de Wheeler**:

**En computación, todos los problemas se resuelven con un nivel extra de indirección**.

Si de nuevo lo traducimos a términos más familiares, nos quedaría como

**En ciberkinesis, todos los problemas se resuelven mediante un adaptador**

Así de importante y universalmente útil, es el concepto del *adaptador*

#### ... pero no te pases

Recuerda que la ley en cuestión, también tiene un corolorario:

**LEY**

*En computación, todos los problemas se resuelven con un nivel extra de indirección*

**COROLARIO**

*Excepto el exceso de niveles de indirección*


Es decir, no te vuelvas loco a poner adaptadores sobre adaptadores: eso ya es una ñapa. Como siempre, usa el setido común.
