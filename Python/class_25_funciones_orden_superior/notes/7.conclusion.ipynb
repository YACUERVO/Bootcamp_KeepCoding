# Conclusiones

-------------

Al poner en práctica los aplicadores, terminamos viendo otros muchos conceptos igualmente importantes:

* ordenación
* clasificación y máscaras booleanas
* manipulación de imágenes
* patrón del adaptador

También hemos vislumbado un potencial problema: Cuando tenemos que aplicar varias operaciones sobre una secuencia muy larga, el tener que recorrerla muchas veces, puede convertirse en un cuello d ebotella. Podríamos aplicar todas las operaciones de un golpe, pero perderíamos la sencillez y claridad a cambio de una mayor eficiencia. Más adelante, veremos que existen formas de preservar ambas cosas (claridad y eficiencia).
