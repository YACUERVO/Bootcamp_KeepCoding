# Clasificadores Booleanos

----------------------

En esta sección veremos una variación de un concepto ya familiar: el *selector*.

Un *selector* es una función que recibe una lista y devuelve una nueva lista con aquellos elementos que superan un test predicado. Es decir, *selecciona algunos elementos y descarta a otros*.

Por ejemplo, este *selector* selecciona las cadenas cuya longitud es mayor que `max_len`:

```python
def long_strings(strings:list[str], max_len:int)->list[str]:
    """
    selecciona las cadenas cuya longitud es mayor que `max_len`
    """
    result = []
    for string in strings:
        if len(string) > max_len:
            result.append(string)
    return result
```


### Clasificador Booleano

Mientras que un *selector* sólo devuelve aquellos elementos que pasaron el test del predicado, un *clasificador* nos devuelve el resultado del predicado *para cada uno de los elementos*.

El clasificador equivalente a `long_strings` sería:

```python
def classify_long_strings(strings:list[str], max_len:int)->list[bool]:
    result = []
    for string in strings:
        result.append(len(string) > max_len)
    return result
```


El *clasificador* devuelve una lista, de la misma longitud que la de entrada, de valores booleanos, cada uno el resultado de aplicar el test o predicado a los elementos de la lista de entrada.

A dicha *lista de booleanos* la llamamos la *clasificación*.

Es otra forma de *clasificación binaria* (en dos grupos) de los datos según un test. En vez de meter cada uno en un grupo, esa información se guarda en una lista de booleanos.
