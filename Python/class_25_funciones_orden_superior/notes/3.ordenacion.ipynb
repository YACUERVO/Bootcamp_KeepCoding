### Funciones Destructivas y No Destructivas

Recordemos que una *función destructiva* modifica (destruye) sus parámetros, mientras que una *no destructiva* no los altera. Hasta ahora nosotros sólo hemos creado funciones *no destructivas* y eso es lo correcto y lo civilizado. 

El alterar parámetros es muy indeseable y pone en grave peligro la **corrección del software**. La razón es que *los parámetros no pertenecen a la función*. El código que la llama y se los *presta* no se espera que sean cambiados. De la misma forma que si le prestas tu coche a un amigo, no te esperas que lo tunee o peor aun, que loo choque.

La única razón que nos puede llevar a usar una función destructiva, es **el rendimiento**. Si una función es el cuello de botella y el hacerla destructiva, resuelve el problema, pues lo hacemos, pero siempre **asegurando la corrección**.

Una caso común de esto, suele ser la ordenación.



### Ordenación en Python

Hay dos funciones para ordenar listas:

* `sort()`: es destructiva y no devuelve nada. En vez de eso, modifica (ordena) la lista que recibe como parámetro.
* `sorted()`: similar a la anterior pero es *no destructiva* y devuelve la lista ordenada. 

**Siempre usaremos `sorted`**.

### Ordenación Trivial

Supongamos que tenemos una lista de números o de cadenas. Podemos obtener una versión ordenada, llamando a `sorted`.

```python
nums = [4, 1 ,0, -45, 78, 2.09]
ordered = sorted(nums)

print(f"Original: {nums}")
print(f"Ordenada: {ordered}")

```

Funciona tal y como uno se espera, siempre y cuando te esperases que ordenase de menor a mayor. ¿Y si quiero que ordene de *mayor a menor*?

Veamos con una *lista de cadenas*

```python
words = ["Hola", "hola", "que", "ase", "holgura"]
print(sorted(words))
```

Ordenar,  sí que ordena, pero su concepto de orden es un poco peculiar. Veamos algunas de las cosas que hemos observado:

* **Las mayúsculas son menores**: Se considera que las letras en mayúsculas van antes que la misma letra pero en minúsucula. Es decir, `"H" < "h"` y `"H" < "a"`. Es un poco extraño, pero es algo que todos los ordenadores hacen. Comprúebalo ordenando ficheros en tu ordenador y verás que empieza por aquellos que tienen el nombre empezando por una mayúscula. 
* **Las palabras se comparan hasta encontrar una diferencia**: Aunque `hola` y `holgura` empiezan igual, cuando se llega a la primera letra distinta, la `a` y la `g`, respectivamente, la `a` es considerada menor y `"hola" < "holgura"`.

### Hacer que `sort` y `sorted` sean más flexibles

Necesitamos que ambas funciones sean más flexibles, y que podamos indicar cosas como las siguientes:

* ordena de mayor a menor
* compara de una forma específica, por ejemplo en el caso de las cadenas, sin distinguir mayúsculas o minúsculas.

**Siempre que quieras que una función sea más flexible, añade algún parámetro**. Además, dale a ese parámetro un valor por defecto para el caso más habitual.

Eso es exactamente lo que han hecho los creadores de `sort` y `sorted` con dos parámetros:

* `reverse`
* `key`



##### reverse

`reverse` es un booleano que indica si se quiere ordenar de mayor a menor. Por defecto vale `False`

```python
nums = [4, 1 ,0, -45, 78, 2.09]
print(sorted(nums, reverse=True)) # mayor a menor
print(sorted(nums, reverse=False)) # mayor a menor (por defecto)
print(sorted(nums)) # mayor a menor sin usar el parámetro
```

##### key

`key` es una función que recibe un elemento y lo *transforma en aquello que queremos comparar*.

Por ejemplo, supongamos que queremos ordenar una lista de cadenas *por la longitud de la cadena* y de mayor a menor:

```python
ingredientes = ["cilantro", 'ajo', "sal", "canela", "perejil"]

# Ordenación por defecto
print(sorted(ingredientes))

# ordenación por longitud de la cadena
print(sorted(ingredientes, key=len))

# ordenación por longitud y de mayor a menor
print(sorted(ingredientes, key=len, reverse=True))
```

**¿Qué hace `sorted`?**

A cada elemento le aplica la función `len` y usa su valor de retorno para ordenar. Al final, está transformando una cadena en un entero para poder ordenar.

> ¡`key` nos permite aplicar *Transforma & Vencerás* a la ordenación!
