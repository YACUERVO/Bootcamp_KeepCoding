##   Variables

Una variable es un mecanismo para *darle un nombre a un dato* y poder referirnos siempre a ese dato mediante dicho nombre.

`pi = 3.141592653589793`

A partir de este momento, ya nunca más tendré que escribir `3.141592653589793`, podré usar siempre el nombre de variable `pi`.

Prueba a escribir lo de arriba en una celda.
Ahora prueba a escribir `pi`



1. Crear una variable, se llama **definir una variable**. Hemos definido una variable llamada `pi`.
2. Las variables están compuestas de dos cosas:
   1. Un nombre (`pi`)
   2. Un valor (`3.141592653589793`)
3. Las variables se **simplifican a su valor**.
4. Se puede cambiar el **valor** de  la variable, repitiendo la definición con un valor nuevo: `pi = 'hola'`
5. No se puede cambiar el **nombre** de una variable, sólo crear una nueva.
6. Los **nombres** pueden estar compuestos por letras, números y el guión bajo, pero no pueden empezar por un número:
   1. `pi` OK
   2. `clave_de_usuario` OK
   3. `32_piso` NO
   4. `ñapa` OK, pero mejor evita caracteres acentuados. 


> Las variables nos permiten reutilizar un valor

##   Concepto de Abstracción

Es un concepto vital en programación. Tenemos que identificar qué es la esencia de una cosa y qué son detalles irrelevantes. Una vez hecho eso, le damos un nombre y nos olvidamos de los detalles. 

Por ejemplo, 

* **Mamífero** es el nombre que le damos a lo verdaderamente esencial que comparten conceptos como un ornitorrinco, una vaca y tú: todos perteneceis a especies cuyas hembras producen leche.
* Un Ford Fiesta, un Tesla modelo X y un Dos Caballos son **coches**. **Coche** es una abstracción que nos permite referinos a todos ellos sin preocuparnos por detalles como marca, modelo, año de fabricación, combustible, etc..
* El número **Uno** captura la esencia de un virus, una patata, una multa y un disgusto: todos son 1.

Abstraer es un proceso de amnesia selectiva, en el cual decidimos olvidarnos de ciertas cosas y desentendernos de ellas.

> Es un ejemplo de *Divide y Vencerás*: reduzco el problema olvidando detalles insignificantes.  Es la única forma de gestionar cosas complejas: subdividir y olvidar lo irrelevante.
