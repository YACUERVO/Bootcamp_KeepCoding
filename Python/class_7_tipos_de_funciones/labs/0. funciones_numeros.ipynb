## Función Sigmoide

La **función sigmoide** es una función matemática que convierte cualquier número real a un valor entre 0 y 1. Se utiliza ampliamente en **aprendizaje automático**, especialmente en regresión logística y redes neuronales, donde sirve como **función de activación**.

La función sigmoide se define como:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

Donde:
- $ e $ es el número de Euler (aproximadamente 2.718).
- $ x $ es el valor de entrada.

#### Propiedades de la función sigmoide:
- **Rango**: La salida siempre está entre 0 y 1.
- **Forma**: Tiene una curva en forma de "S".
- **Aplicaciones**: Se utiliza en problemas de clasificación binaria para mapear las predicciones a probabilidades.

![](sigmoid.png)


Crea la función `sigmoid` en Python usando la fórmula de arriba. Debe recibir un número y devolver otro en 0 y 1. Puedes usar la función `math.exp`.

Comprueba que sólo devuelve números entre 0 y 1.

##### Uso en IA y Aprendizaje Automático (*Machine Learning*)

En **aprendizaje automático**, la función sigmoide se utiliza principalmente convertir un valor numérico arbitrario en una **probabilidad** entre 0 y 1, lo que facilita la interpretación de las predicciones como probabilidades.

Imagina que estás resolviendo un problema donde el objetivo es clasificar si una acción va a subir o bajar de precio (0 o 1). La función sigmoide toma la salida de un modelo (que podría ser cualquier número) y lo convierte en una probabilidad. Si el resultado de la función sigmoide es mayor que 0.5, el modelo clasifica el resultado como 1 (sube). Si es menor, lo clasifica como 0 (baja).

Por ejemplo, si la función sigmoide devuelve 0.8, significa que el modelo asigna una probabilidad del 80% a que la acción subirá. Esto permite interpretar la salida de manera más intuitiva.


`````python
def funcion_sigmoid(x):
    '''Calcular la funcion sigmoid'''
    sigmoid = 1 / (1 + math.exp(-x))
    return sigmoid

funcion_sigmoid(1)
`````

## La ecuación que cambió el mundo

Vas a calcular la energía $E$ utilizando la famosa ecuación de Einstein $ E = mc^2 $, donde:

- $E$ es la energía en **julios (J)**.
- $ m $ es la masa en **kilogramos (kg)**.

Crea la función `energy` que recibe la masa en kg y devuelve la energía equivalente en Julios. Usa una constante para `c`definida dentro de la función.



`````python
def energy(mass, c=299792458):
    '''Calcular la energia segun la ecuacion de Einstein'''
    energy = mass * c **2
    return energy

energy(0.1)
`````

## Distancia entre dos puntos en el plano cartesiano

![](point_distance.gif)


 Si tenemos dos puntos:

 $p_{1}(x_{1}, y_{1})$

 $p_{2}(x_{2}, y_{2})$


 su *distancia cartesiana* sería la siguiente:

 $$dc = \sqrt{(x_{1} - x_{2})^{2} + (y_{1} - y_{2})^{2}}$$


 1. Crear una función que recibe la xe y del primer punto y la x e y del segundo punto
 2. Devuelve la distancia cartesiana


### Curiosidad

Esta es la *distancia cartesiana*, pero hay otros tipos de distancias definidas entre puntos.

La *distancia cartesiana* es la primera que te muestra Google Maps o Waze cuando pides que te lleve a algún sitio. Normalmente cuando pides que calcule la ruta, te llevas un chasco, porque te da otra distancia mayor. ¿¿Por qué??

Porque la distancia cartesiana sólo vale si puedes volar. Si tienes que ir entre calles, tendrás que dar vueltas y eso incrementa la distancia.

> ¿Por qué crees que Google hace eso?


Otra distancia, algo más realista para la navegación por ciudades, es la *distancia manhattan*.

Averigua en la wikipedia qué es la distancia manhattan, por qué se llama así y cómo calcularla. Crea la función Python correspondiente.


`````python
def distancia_cartesiana(x1, y1, x2, y2):
    '''Calcular la distancia cartesiana entre dos puntos'''
    distancia = math.sqrt((x2 - x1) **2 + (y2 - y1) **2)
    return distancia

def distancia_manhattan(x1, y1, x2, y2):
    '''Calcular la distancia manhattan entre dos puntos'''
    distancia = abs(x2 - x1) + abs(y2 - y1)
    return distancia
print(distancia_cartesiana(1,2,4,6))
print(distancia_manhattan(1,2,4,6))
`````
