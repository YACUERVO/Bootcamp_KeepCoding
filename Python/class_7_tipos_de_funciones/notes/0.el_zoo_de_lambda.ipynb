# El Zoo de Lambda

-----

![](ada_zoo_lambda.jpg)


En el vasto Reino de Turingia, existe un lugar mágico conocido como el Zoológico de Lambda, un sitio donde las funciones de todos los rincones del reino se encuentran representadas.

Un día, una curiosa estudiante del arte de la Ciberkinesis llamada Ada decide visitar este zoológico para entender mejor la diversidad de funciones y su importancia. A su llegada, es recibida por el guía, un anciano conocido como Lambda el Supremo, quien se ofrece a mostrarle los diferentes hábitats.

El primer lugar que visitan es el Bosque de las Funciones Totales, donde cada función tiene la capacidad de producir un resultado para cada entrada posible. "Estas funciones son confiables", explica Lambda, "pues sabemos que siempre responderán, sin importar la pregunta".

Continuando su paseo, llegan a la Caverna de las Funciones Parciales, un lugar oscuro y traicionero, habitado por criaturas como la temida Esfinge de Church: si no le gusta tu pregunta, te devora. Lambda señala que estas funciones, a diferencia de sus primas del bosque, pueden no tener una respuesta para cada entrada. "Aquí es donde la necesidad de manejar errores se vuelve crucial", advierte, "pues aventurarse sin precaución puede llevar a resultados desastrosos".

Más adelante, encuentran el Valle de las Funciones Irreversibles, hogar de seres como el Dragón de Hash. Estas funciones son especiales, explica Lambda, ya que transforman la información de manera que no puede ser revertida a su forma original. "Este poder", dice con un tono de misterio, "es fundamental en el mundo de la criptografía, protegiendo nuestros secretos de ojos indiscretos".

Finalmente, llegan a la Llanura de la Abstracción, un amplio valle donde reina el Cálculo Lambda. Aquí, todo es posible: las funciones pueden ser pasadas como argumentos, retornadas como resultados, y todo interactúa en armonía. "Este es el corazón de nuestro zoológico", declara Lambda con orgullo, "donde el concepto de funciones de alto orden permite que el zoológico florezca en toda su complejidad".

Ada, maravillada por la diversidad y las lecciones, agradece a Lambda por el tour. Ha comprendido que en el mundo de la computación, las funciones no solo realizan tareas; son los pilares sobre los cuales se construyen universos enteros de lógica y magia.

Al salir del Zoológico de Lambda, Ada reflexiona sobre cómo cada tipo de función, desde las totales hasta las irreversibles, juega un papel vital, y cómo la habilidad para manejar errores y entender la naturaleza de estas funciones es esencial para cualquier practicante de la ciberkinesis. Con una sonrisa, Ada se promete a sí misma regresar algún día, quizás como una experta en la misteriosa y poderosa magia de las funciones.


# La Magia de la Funciones


De todas las formas de magia que componen el Arte de la Ciberkinesis, **las funciones son la más poderosa**. Todo lo demás se deriva de ella y en ella está contenida toda la Ciberkinesis. Quien la domina, tiene asegurada su entrada al Olimpo de la Ciberkinesis.

La base matemática sobre la que se sostienen las funciones es el **Cálculo Lambda**, creado por *Alonzo Church*. En dicha obra se describen todos los habitantes del Zoo de Lambda: las funciones totales y parciales, las funciones reversibles e irreversibles, las funciones de orden superior o las funciones *hash*.

Hay muchos más habitantes en el zoo, pero se encuentran en secciones prohibidas para el visitante novel. En ellas habitan seres como las funciones recursivas o los combinadores. Se dice incluso, que en un pasillo recóndito, al amparo de la oscuridad, se haya el mítico Combinador Y. Pocos lo han vislumbrado y muchos creen que no es real y que es tan sólo [una aceleradora de startups](https://www.ycombinator.com/). ¡Ignaros!

Sin embargo, la sola visión de uno de estos seres convertiría instantanenamente el cerebro de un aprendiz en puré de lentejas recalentado. Mejor será que vayamos por partes y sin dejar un día a cero. Veamos a los primeros habitantes del zoo:

* funciones totales
* funciones parciales
* funciones reversibles
* funciones irreversibles
* funciones irreversibles estimables
