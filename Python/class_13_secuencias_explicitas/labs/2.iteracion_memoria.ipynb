Cuando estamos iterando por una secuencia, el ámbito del `for` sólo tiene acceso al elemento actual:

```python
for actual in [1,2,3,4, 5,6,7]:
    # Aquí dentro sólo "vemos" el elemento actual de la
    # secuencia. Por ejemplo, si `actual` vale `5`, no hay
    # forma de saber que antes valió `4`.
    # Aquí es donde vive Leonard
```


Hasta ahora esto jamás había sido un problema: aunque el ámbito del `for` es como una *ventana* que se va moviendo por la secuencia, con saber donde estamos *ahora* nos ha sido suficiente. Sin embargo, hay casos en los que hace falta saber cuales fueron los valores anteriores. ¿Cómo podemos resolver esto?

### Nuestros mementos

En vez de tatuajes, lo que haremos nosotros es ir guardando lo que ya hayamos visto, en variables declaradas antes del `for`.

Por ejemplo, supongamos que tenemos que recorrer una lista y comprobar que **cada elemento es igual o mayor** que el anterior. Necesitaríamos una variable para guardar el elemento anterior que hayamos visto.

````python
def is_increasing(numbers):
    """
    Predicado que devuelve True si cada elemento es igual o mayor que el anterior
    En caso de la lista vacía devuelve False
    """
    result = True

    if numbers == []:
        return False
    else:
        anterior = numbers[0] # el primer elemento
        for actual in numbers:
            # Aqui sólo veo el de turno (actual), pero tengo en 
            # `anterior` el que acabo de ver
            if anterior > actual:
                # Acabamos de encontrar un ejemplo que demuestra que la lista
                # no cumple con la condición: cambiamos el resultado y salimos
                result = False
                break
            else:
                # la pareja `anterior` & `actual` cumple. Seguimos buscando
                # pero hay que actualizar `anterior`!
                anterior = actual
    return result
````

````python
# Comprobamos que funciona
# assert es una palabra clave que recibe una expresión booleana y si
# no es True, devuelve un error de inmediato. Es buena para comprobaciones rápidas
assert is_increasing([]) == False
assert is_increasing([1]) == True
assert is_increasing([1,1 ,1, 1]) == True # si esto te sorprende, revisa el enuciado
assert is_increasing([1,2,3,4]) == True
assert is_increasing([1,2,3, 2, 4]) == False


````
