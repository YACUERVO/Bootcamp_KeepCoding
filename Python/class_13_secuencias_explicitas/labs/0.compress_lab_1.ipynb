### Suma de cuadrados

Crea la función que recibe una lista de números y devuelve la suma de los cuadrados de dichos números. Si recibe la lista vacía, devuelve 0.

````python
def sum_squares(numbers):
    total = 0
    for number in numbers:
        total = total + (number ** 2)
    return total

sum_squares([2,2,3,4])
````

### Suma módulo 10

Ahora se trata de crear una función `sum_mod_10` que recibe una lista de números y tiene que devolver la suma módulo 10. Es decir, suma los elementos y luego obtiene el módulo 10 del resultado parcial:

`(a + b) % 10`

* Considerando que se va sacando el módulo 10 de las sumas parciales, ¿cual debería de ser el rango del resultado final? 
  * ¿Podría ser el resultado alguno de éstos números: 10, 21, 42, 65?
  * ¿Por qué?

````python
def sum_mod_10(list_numbers):
    sum_model = 0
    for number in list_numbers:
        sum_model = (sum_model + number) % 10
    return sum_model 

sum_mod_10([1,10,20,15,10])
````

### Al menos un `True`

Ya tenemos una función que recibe una lista de booleanos y devuelve `True` cuando **todos son `True`**. Ahora debes de diseñar la función `some_true` que recibe una lista de booleanos y sólo devuelve `True` cuando al menos uno de ellos es `True`.

* ¿Qué operador booleano usarías en este caso y por qué?
* ¿Cual debería de ser el valor inicial del acumulador y por qué?
* Comprueba que funciona tal como se indica:
  * `some_true([]) == False`
  * `some_true([False, False, True]) == True`
  * `some_true([3 > 2, 2 <= -1]) == False`

````python
def some_true(list_bolean):
    '''Algun verdadero'''
    result = False
    for boelan in list_bolean:
        result = result or boelan
    return result

print(some_true([False, False, False ,False, False]))
print(some_true([]))
print(some_true([False, False, True]))
print(some_true([3 > 2, 2 <= -1]))
````

### CSV: Comma Separated Values

Crea la función `to_csv`. CSV significa *comma separated values* y es un formato común para representar secuencias y tablas. Lo volveremosa ver en el futuro.
La función debe de recibir una lista de cadenas y devuelve una sola cadena con los elementos separados por por comas y un espacio:

`["nombre", "apellidos", "edad"] --> "nombre,apellidos,edad,"`

* Queda un poco feo que no haya espacios entre los valores. 
  * Modifica tu función para que el resultado anterior sea `"nombre, apellidos, edad, "`
* También queda feo que el *último elemento* termine en coma. 
  * Averigua por tu cuenta cómo eliminar esos últimos caracteres de la cadena resultante, para que quede así `"nombre, apellidos, edad"`

````python
def to_cvs(text_string):
    text = ''
    for values in text_string:
        text = text + values + ' ,'
    return text[:-2]

print(to_cvs(["nombre", "apellidos", "edad"]))
````

`None` es la única instancia del typo `NoneType` y se usa para indicar la nada.

Entre los tipos de Python, juego el mismo papel que el 0 entre los números.

1. Refactoriza tu función `max_value` para que en el caso de que la lista de entrada sea la vacía, que devuelva `None`.

````python
def max_value(list_numbers):
    '''Funcion que recibe una lista de numero y devuelve el mayor'''
    max = None
    if len(list_numbers)> 0:
        max = list_numbers[0]
        for value in list_numbers:
            if value > max:
                max = value  
    return max

print(max_value([10,11,25,5,7,12,13,4,15]))

print(max_value([-10,-11,-25,-5,-7,-12,-13,-4,-15]))

print(max_value([]))


````

### Compresores

Todas estas funciones tienen una cosa en común **reciben una lista de elementos y devuelven un solo elemento**. 

*Reciben una lista* y la **comprimen** a un sólo elemento. Dicho elemento se obtiene *combinando mediante alguna operación* a los elementos de la lista.

Para la operación de compresión usamos todos los elementos. 

Para seleccionar, usamos un **predicado**.

Para ir construyendo el valor comprimido, usamos una variable donde vamos *acumulando* los resultados temporales. A esa variable le llamamos el **acumulador**. 

Cuando llegamos al final,  tenemos el *valor comprimido final* dentro del *acumulador*, y lo devolvemos.

El patrón general de un **compresor** es:


```python

def compress(elements):

    # Acumulador con valor inicial
    accum = <initial>

    # iteramos
    for element in elements:
        # combinamos. La operación de combinación
        # puede incluir un predicado que seleccione
        # o no el valor actual para la combinación
        accum = <combine>(element, accum)

    # Tenemos en accum el valor comprimido
    return accum
```







### Esqueleto común y partes específicas

* el valor inicial del acumulador
* la operación de combinación (que puede incluir un predicado)
