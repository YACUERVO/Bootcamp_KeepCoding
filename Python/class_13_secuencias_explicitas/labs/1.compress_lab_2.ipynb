# Compresores Completos e Incompletos

----------------------


En en laboratorio anterior, descubrimos el concepto de los *compresores*: funciones que **reciben una secuencia de elementos y los comprimen a uno sólo**.  Lo hacen combinando **todos** los elementos mediante alguna **operación binaria** (opera sobre dos valores). A éstos, les llamaremos **compresores completos**.

Sin embargo existen otros, y son muy comunes, a los que podríamos llamar **compresores incompletos**: combinan solo **una parte de los elementos de la secuencia**. Existen dos tipos básicos:

* Los que aplican un *test* a cada elemento, y sólo lo combinan si pasa el test.
* Los que paran *antes de llegar al final* de la secuencia, al ser innecesario seguir.

Vamos a ver ejemplos de ambos.

### Suma de los múltiplos

Crea la función `sum_mult_of_3`:

* Recibe una lista de números
* Devuelve la suma de aquellos que son **múltiplos de 3**

#### DIVIDE & VENCERÁS

Resuelve primero el subproblema "saber si un número es múltiplo de 3", creando el predicado `is_multiple_3`. Recuerda el operador `%` para obtener el resto de una división.



````python



def is_multiple_3(element): 
   return element % 3 == 0
is_multiple_3(9)
````

````python
def sum_mult_of_3(numbers):
    '''Saber si un numero es multiplo de tres'''
    sum = 0
    for number in numbers:
        if is_multiple_3(number):
            sum = sum + number
    return sum


assert sum_mult_of_3([]) == 0
assert sum_mult_of_3([2,4,7,8]) == 0
assert sum_mult_of_3([3,6,9,8,7]) == 18
print(sum_mult_of_3([3,6,9,2,4,5]))
````

### Concatenar las cadenas que contienen una x minúscula

Debes de crear la función `concat_x` que recibe una *lista de cadenas* y devuelve *una cadena* resultado de haber concatenado todas las cadenas que contengan la `x` minúscula.

* Recuerda que puedes averiguar si una cadena contiene una letra usando el operador `in`: `"a" in "abba"` se evaluará a `True`
* Recuerda también que ese `in` no tiene nada que ver con el del bucle `for` (aunque se llamen igual). Son cosas distintas que se usan en contrextos distintos, como el `+` en los números (suma) y con las cadenas (concatenación).

1. ¿Cual debe de ser el valor inicial de tu acumulador?
2. ¿Qué devuelve cuando recibe la lista vacía?

````python
def concat_x(cadena_texto):
    '''Funcion que valida si la letra x se encuentra en una cadena y devuelve las palabras que cumplan con la letra x'''
    result = ''
    for cadena in cadena_texto: 
        if 'x' in cadena:
           result = result +  cadena + ' '
    return result[:-1]
````

````python
assert concat_x(['xiomara']) == 'xiomara'
assert concat_x(['Xiomara']) == ''
print(concat_x(['xiomara' , 'yamith', 'xinofono','Xan'])) # == 'xiomara xinofono'
````

````python
def is_vocal(cadena_texto):
    '''Funcion que Verifica si una cadena de texto contiene una vocal'''
    result = False
    for vocal in cadena_texto:
        if vocal.lower() in 'aeiou':
            result = True
            
    return result

assert is_vocal('PERRO') == True
assert is_vocal('perro') == True
assert is_vocal('XYBN') == False

````

````python
def concat_vowels(cadena_texto):
    '''Validar si una cadena de texto contiene una vocal sea mayuscula o minuscula'''
    result = ''
    for vocal in cadena_texto:
        if is_vocal(vocal):
            result = result + vocal + ' '
    return result[:-1]

print(concat_vowels(['Perro', 'Gato', 'hghg','GATOTE']))
````

### Contar las cadenas más largas que un cierto número

Crea la función `count_long_strs(strings, max_len)` que recibe una lista de cadenas y un número (`max_len`), y devuelve el número de cadenas cuya longitud es mayor o igual a `max_len`.





````python
def count_long_str(strings,max_len):
    '''Funcion que cuenta la cantidad de string que cumpla con max_len'''
    result=0
    for string in strings:
        if max_len < len(string):
            result = result + 1
    return result


````

````python
#Test
print(count_long_str(['cadena','yamith','oso','ojo'],3))
assert count_long_str(['cadena','yamith','oso','ojo'],3) == 2

print(count_long_str(['cadena','yamith','oso','ojo','caracoles','serpiente','hipopotamo'],6))
````

### Tus funciones, el polimorfismo y el botón de encender y apagar

Como sin duda te habrás dado cuenta, no se parecen, sino que son la misma función. De hecho, talvez esa función debería de ser una sóla, y llamarse `count_long_seq` de *contar secuencias largas*.

**¿Por qué son exactamente iguales?**

De entrada porque *hacen lo mismo*: recorren una secuencia, averiguando la longitud de las secuencias dentro de ella.

La clave está en la **función `len`**: mide la longitud de cualquier secuencia. Sin duda debe de haber diferencias entre cadenas y listas (¡y las hay!), pero `len` es capaz de gestionar esas diferencias y hacer su trabajo independientemente de los detalles. Gracias a eso no tenemos dos funciones con nombres distintos para hacer fundamentalmente lo mismo: `len_str` y `len_list`.  Tener que acordarse de una sola `len`es mucho más cómodo.

Es como si `len` estuviese implementada de una forma parecida a esto:

```python

    if type(elements) == str:
        # es una cadena
        # código para medir longitud de cadenas
    elif type(elements) == list:
        # una lista
        # código para medir longitud de cadenas
    elif...

```

A las funciones que como `len` pueden trabajar sobre distintos tipos, se les llama **Funciones Polimórficas**. Polimorfismo viene del Griego y significa *muchas formas*. Es como si `len` tuviese una *forma para las cadenas* y una *forma para las listas* (en realidad, tiene *formas* para cada tipo de secuencia).

El **polimorfismo** hace que un sistema sea mucho más sencillo, al sólo tener que recordar un nombre para un conjunto de operaciones similares.

### Suma de todos los elementos pares

Crea la función `sum_even` que recibe una lista de números y devuelve la suma de los que son pares.

````python
def sum_even(numbers):
    '''Funcion que suma los numero pares de una lista'''
    result = 0
    for number in numbers:
        if number % 2 ==0:
            result = result + number
    return result
````

### Sumar hasta un máximo

Crea la función `sum_up_to(numbers, max_val)`  que recibe una lista de números y un valor máximo. Va sumando hasta que el valor acumulado es igual o superior a `max_val`, y devuelve ese número.

Recuerda que un `for` por defecto sigue iterando hasta el final de la secuencia. Si has logrado tu objetivo antes de terminar y deseas salir, tienes que usar la palabra clave `break`

````python
def sum_up_tp(numbers,max_val):
    result = 0
    for number in numbers:
        result = result + number
        if result >= max_val:
          break            
    return result


print(sum_up_tp([5,5,10,15,15,20],25))
````

### Sumar números y `None`

Recibes una lista que contiene números, ¡pero también puede contener le valor `None`!  Es decir, pueden ser listas como estas:

* `[2,3,None, 3, None, 4]`
* `[None, None, -1]`
* `[None, None, None]`
* `[8, -3]`
* `[]`

Se trata de una lista de números y de `None`.  Esta combinación (un tipo *normal* y `None`) es muy común y se llama una lista de *números opcionales*: a veces es un número, y a veces, opcionalmente, no hay número.

Si fuesen cadenas y `None`, entonces serían *cadenas opcionales*.

Esto se usa mucho para indicar que en un caso, nos falta el dato, y ese hueco lo rellenamos con un `None` para indicar precisamente eso. Es muy común en todos los lenguajes, no sólo en Python.

Veamos un ejemplo concreto. Supongamos que estás haciendo una app de navegación, como Waze o Google Maps. Recibes constantemente una secuencia de posiciones obtenidas con el GPS, pero *en algún momento eso puede fallar*. El usario entra en un túnel y deja de haber datos: en ese caso se envía un `None`. Tu aplicación no puede caerse por eso, tiene que ser capaz de *sobrevivir* a la llegada de un `None`.

Eso es precisamente lo que vas a hacer ahora.

Crea la función `flat_sum` que recibe una lista de *números opcionales* y devuelve el total.

Debes de recorrer todos los elementos, sumando los números e ignorando a los `None`s.

A estas versiones de funciones como `sum_all` que son capaces de gestionar un `None` ocasional, se les suele llamar funciones *flat*. Las encontrarás muy a menudo en Big Data (procesamiento de datos) y también en desarrollo Web y Mobile.

````python
def flat_sum(list_numbers):
    acum = 0
    for number in list_numbers:
        if number is not None:
            acum = acum + number
    return acum

def flat_sum(list_numbers):
    acum = 0
    for number in list_numbers:
        if type(number) == int:
            acum = acum + number
    return acum
````

### Compresores incompletos

Todas estas funciones tienen una cosa en común *reciben una lista de elementos* y devuelven un solo elemento. Son compresores.

La diferencia es que son incompletos: no necesitan todos los elementos y  

Para la operación de compresión, en este caso usamos solo algunos **seleccionados**. 

Para seleccionar, usamos un **predicado**.

Para ir construyendo el valor comprimido, usamos una variable donde vamos *acumulando* los resultados temporales. A esa variable le llamamos el **acumulador**. 

Cuando llegamos al final,  tenemos el *valor comprimido final* dentro del *acumulador*, y lo devolvemos.

El patrón general es:


```python

def compress(elements):

    # Acumulador con valor inicial
    accum = <initial>

    # iteramos
    for element in elements:
        # combrobamos si el elemento actual supera un test
        if <predicate>(element):
            # lo tenemos en cuenta
            accum = <combine>(element, accum)

    # Tenemos en accum el valor comprimido
    return accum
```


En el fondo, no es más que un caso especial de **compresor**.




## Resumen

Los *compresores* son un patrón computacional universal que aparece en multitud de aplciaciones. Se caracterizan por recibir una secuencia de elementos y devuelven 1, tras analizar de forma secuencial a los de antrada, combinándolos hasta crear el valor de salida.
