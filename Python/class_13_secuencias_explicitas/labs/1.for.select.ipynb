## (1)

Crea la función `select_even` que recibe una lista de números y devuelve una lista con *sólo los números pares*, en el mismo orden.

Es decir, si recibe la lista `[1,2,3,4, 5, 6]`, el valor correcto de retorno es `[2,4,6]`. No es correcto devolver `[4, 2, 6]`.

* ¿Qué debe de devolver si no hay ningún numero par?
* ¿Qué debe de devolver si recibe la lista vacía?



````python
def select_even(list_numbers):
    '''Lista de numero de pares'''
    list_even = []
    for number in list_numbers:
        if number % 2 ==0:
            list_even.append(number)
    return list_even
print(select_even([1,2,3,4,5,6,7,8,9]))
print(select_even([3,5,7]))
print(select_even([]))
````

Crea la función `select_long_words`. Recibe dos parámetros:

1. Un número
2. Una lista de cadenas

Devuelve la lista de las palabras cuya longitud sea igual o mayor que el número.

* ¿Qué devolverías si el número es negativo o cero?
* ¿Qué devolverías si recibes una lista vacía?




````python
def select_long_words(list_strings,number):
    '''Funcion que recibe una lista de palabras y devuelve la longitud de esas palabras 
    que corresponda al parametro number'''
    strings = []
    for string in list_strings:
        if len(string) > number:
            strings.append(string)
    return strings
````

````python
def select_contains(list_strings,sub_string):
    '''Funcion que contiene una lista de cadena donde cada elemente se compara si contiene
    una subcadena'''
    lists_str = []
    for string in list_strings:
        if len(string) >= len(sub_string) and sub_string in string:
            lists_str.append(string)
    return lists_str


````

````python
#Recibe una aguja y una lista de pajares
def select_contains(needle, list_haystacks):
    selected=[]
    for haystack in list_haystacks:
        if contains(needle,haystack):
            selected.append(haystack)
    return selected
````

````python
def select_stars(list_strings,string):
    list_strs=[]
    for elemnt in list_strings:
        if elemnt.startswith(string):
            list_strs.append(elemnt)
    return list_strs
````

````python
#Otra manera de hacer el ejercicio
#Realizo el predicado primero 
def stars_with(needle, haystack):
    '''Predicado que determina si haystack empieza por needle'''
    return haystack[0:len(needle)] == needle

#Luego realizo la busquedad dentro de la lista
def selec_sarts(needle, list_haystacks):
    selected=[]
    for haystack in list_haystacks:
        if stars_with(needle,haystack):
            selected.append(haystack)
    return selected

assert stars_with('hola','mundo') == False
assert stars_with('gato','gato negro') == True
````

## Selectores

Todas estas funciones reciben una lista y devuelven otra lista. La lista de salida es una *versión filtrada* de la de entrada.

Todas ellas *seleccionan* (mediante un predicado o comparación) aquellos elementos que pasan el test.

Todas ellas van *construyendo una nueva lista* con aquellos elementos que van pasando el test.

Para ello, se usa una variable que empieza con la lista vacía y  a ella se van añadiendo aquellos elementos que pasan el test.

Cuando llegas al final de la lista, todos los elementos seleccionados estarán en la variable.

El patrón general de un **selector** es:

```python
def select(elements):
    selected = []
    for element in elements:
        if <predicate>(element):
            selected.append(element)
    return selected
```

## Iterar por posiciones de una cadena y acceder a la última posición

Cuando trabajas con cadenas y necesitas iterar por sus posiciones o acceder a la última posición, tienes varias opciones en Python:

### Métodos principales:

1. **Índices negativos**: La forma más simple para acceder a la última posición
   - `cadena[-1]` → último carácter
   - `cadena[-2]` → penúltimo carácter
   - etc.

2. **Usar `len()`**: Para obtener el índice de la última posición
   - `len(cadena) - 1` → índice de la última posición

3. **Iterar con `range(len(cadena))`**: Para obtener todos los índices

4. **Slicing**: Para obtener subcadenas desde cualquier posición


````python
# Ejemplo: cadena de trabajo
cadena = "hola"

# ===== ACCEDER A LA ÚLTIMA POSICIÓN =====

# Método 1: Índice negativo (MÁS SIMPLE)
print("Último carácter (índice negativo):", cadena[-1])  # 'a'
print("Penúltimo carácter:", cadena[-2])  # 'l'

# Método 2: Usando len() para obtener el índice
ultima_posicion = len(cadena) - 1
print(f"Índice de última posición: {ultima_posicion}")
print("Último carácter (con len):", cadena[ultima_posicion])  # 'a'

# ===== ITERAR POR POSICIONES =====

# Método 1: Iterar con índices usando range()
print("\nIterando con índices:")
for i in range(len(cadena)):
    print(f"Posición {i}: '{cadena[i]}'")

# Método 2: Iterar con enumerate() (obtiene índice y valor)
print("\nIterando con enumerate():")
for i, caracter in enumerate(cadena):
    print(f"Posición {i}: '{caracter}'")

# Método 3: Iterar desde el final hacia el principio
print("\nIterando desde el final:")
for i in range(len(cadena) - 1, -1, -1):  # desde len-1 hasta 0, paso -1
    print(f"Posición {i}: '{cadena[i]}'")

# ===== OBTENER SUBCADENAS =====

print("\nSubcadenas:")
# Desde el inicio hasta una posición
print("Desde inicio hasta posición 2:", cadena[:3])  # 'hol'

# Desde una posición hasta el final
print("Desde posición 1 hasta el final:", cadena[1:])  # 'ola'

# Desde una posición hasta otra
print("Desde posición 1 hasta 3:", cadena[1:3])  # 'ol'

# Desde el final (usando índices negativos)
print("Últimos 2 caracteres:", cadena[-2:])  # 'la'
print("Todos menos el último:", cadena[:-1])  # 'hol'
````

````python
# Ejemplo práctico: Obtener todas las subcadenas desde el inicio
# Similar al ejercicio de la celda 32, pero corregido

def obtener_subcadenas_desde_inicio(cadena):
    """
    Devuelve todas las subcadenas desde el inicio:
    'hola' -> ['h', 'ho', 'hol', 'hola']
    """
    subcadenas = []
    for i in range(1, len(cadena) + 1):  # desde 1 hasta len(cadena)
        subcadenas.append(cadena[:i])  # desde inicio hasta posición i
    return subcadenas

print(obtener_subcadenas_desde_inicio('hola'))
# Resultado: ['h', 'ho', 'hol', 'hola']

# Ejemplo: Obtener todas las subcadenas desde el final
def obtener_subcadenas_desde_final(cadena):
    """
    Devuelve todas las subcadenas desde el final:
    'hola' -> ['a', 'la', 'ola', 'hola']
    """
    subcadenas = []
    for i in range(len(cadena), 0, -1):  # desde len hasta 1, paso -1
        subcadenas.append(cadena[-i:])  # desde posición -i hasta el final
    return subcadenas

print(obtener_subcadenas_desde_final('hola'))
# Resultado: ['hola', 'ola', 'la', 'a']
````
