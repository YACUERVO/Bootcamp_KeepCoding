# Listas en Python

--------------------

La lista, al ser un tipo súmamente común en Python, tiene una *representación literal*: los elementos separados por comas y entre corchetes.

* `[3,4,5,6]`: lista de enteros.
* `[True, False, False]`: lista de booleanos.
* `["Reservoir Dogs", "Pulp Fiction", 'Hateful Eight', "Django Unchained"]`: lista de cadenas.
  
Las listas del ejemplo anterior son *homogeneas*. Esto quiere decir que el tipo de todos sus elementos es el mismo.

En la mayoría de los lenguajes, las listas *siempre tienen que ser homogeneas*. En Python no es así, y puede haber listas *heterogeneas*, donde hay elementos de varios tipos distintos.

Ejemplo de una lista en Python con elementos de *diferentes tipos*. Pueden incluso ser otras listas.

`[3, 0.999, False, ["Kill Bill Vol. I", "Kill Bill Vol. 2"], "Pai Mei" ]`



## Tipo y Constructor

El tipo de la lista es `list` y el constructor se puede usar para converir otras secuencias en listas. En estos momentos no nos va a ser muy útil.

## Una lista muy especial: la lista vacía

La lista vacía `[]` es un caso especial y muy útil. Juega un papel similar al cero entre los números. La veremos muchas veces.

## Añadir elementos a una lista

Una de las operaciones más habituales es la añadir un elemento al final de una lista existente. Para eso usaremos el método `append`que acepta como parámetro el elemento que se quiere añadir.

* `[1,2,3].append(4)` => `[1,2,3,4]`
* `[].append("The path of the righteous man")` => `["The path of the righteous man"]`



## Comprobar la presencia con `in`

El operador `in` se usa para comprobar la presencia de un elemento dentro de la lista. Devuelve un booleano.

* `4 in [1,2,3]` => `False`
* `[5] in ["Python", [5], []]` => `True`

Es importante tener en cuenta `in` sólo dice si el elemento está o no dentro de la lista. No nos dice en qué posición lo encontró, ni tampoco cuantas veces aparece. Si necesitamos esa información, tendremos que hacer la búsqueda nosotros mismos.
