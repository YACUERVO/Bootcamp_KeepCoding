## Finalidad de las secuencias 🛎️

La **finalidad de toda estructura de datos es hacer que un algoritmo que antes era complejo, se convierta en algo trivial**. Reórdenando los datos de una forma diferente, logramaos hacer que una tarea se convierta en algo tan sencillo, que a veces puede parecer que no es importante.

**¿Qué tarea facilita la secuencia?**

La de recorrer todos los elementos sin olvidarnos de ninguno. Esto se logra, asignado a cada uno un índice consecutivo. Es decir, en una *secuencia todos los elementos tienen un índice o posición*.

Su busco un elemento, recorriendo toda la secuencia y el elemento está en la secuencia, sé que en algún momento va a aparecer. Si por el contrario, llego hasta el final y no lo he encontrado, sé a ciencia cierta que no está. Compara eso con buscar algo en la pila de libros de la primera imagen.

La *abstración* de la secuencia nos permite practicar una *amnesia selectiva*, olvidarnos de dónde está cada cosa, gracias a la seguridad absoluta de podee encontrarlo todo de forma trivial.



## Características de las secuencias 🛎️


* Todas las secuencias tienen una longitud.
* Se puede acceder directamente por posición a cualquier elemento.
* Todos los elementos tienen una posición, y el primero tiene la posición 0.
* Se puede *iterar* por todos los elementos, es decir, se puede recorrer la secuencia, de elemento en elemento.




## Secuencias en Python

1. Todas las secuencias tienen una longitud. 
   - **En python la longitud se obtiene con la función `len`**
2. Se puede acceder directamente por posición a cualquier elemento.
   - **En Python el elemento se obtiene por posición con el operador `[posicion]`**
3. Todos los elementos tienen una posición, y el primero tiene la posición 0. 
   - **El primer elemento de la secuencia `seq` es `seq[0]`**
4. Se puede *iterar* por todos los elementos, es decir, se puede recorrer la secuencia, de elemento en elemento.
   - **Veremos varios mecanismos, el más habitual siendo un bucle `for`**

Lo anterior se aplica a todos los lenguajes de programación y con una sintaxis muy similar. A veces la longitud se obtiene mediante un método en vez de una función, pero es todo muy parecido y estandar.

En Python sin embargo, hay algo particualr que no es tan común: el operador `in`.


#### Operador `in` en Python

El operador `in` permite saber si una secuencia contiene un elemento o no. Devuelve un `bool`, `True` si el elemento está en la secuencia y `False` si no lo está.

**Ejemplo:**

`3 in seq`: Devuelve `True` si la secuencia `seq` contiene el entero `3`, y `False` si no.

### Listas

Las listas son un tipo de Secuencia.

Las listas en Python se representan como una secuencia de datos separados por comas y entre corchetes:

* `[1,8,93]`
* `[True]`
* `[[5,6], ["hola", "mundo"]]`
* `[]`
* `[True, 45, 'hola']`

Las listas pueden ser de cualquier cosa, incluso otras listas o estar vacías.

Cada lista es un valor de Python: se evalúan a sí mismos.



#### El tipo lista

Las listas son un tipo de Python:

* Tienen infinitas instancias
* Tienen operaciones en común: 
  * tienen longitud (`len`)
  * se pueden recorrer
  * se pueden buscar
  * se puede seleccionar un elemento
  

Prueba a evaluar esto:

```python
type([3, "hol"])
```

#### Longitud

Se obtiene con la función `len` de Python

````python
l0 = len([[], [45, [56]]])
l1 = len("hola mundo")
````

#### Obtener un elemento

Se usa el operador `[]`, sabiendo que los **índices empiezan en 0** y llegan hasta su **longitud - 1**.


La cadena "hola" tiene longitud 4, su primer índice es el 0 y el último es el **3**.

|h|o|l|a|
|-|-|-|-|
|0|1|2|3|

Si intentas acceder más allá del último índices, te da un error.


````python
"hola"[10]

````

#### Buscar en una secuencia

Para buscar en una secuencia, se utiliza el operador `in` que nos devuelve un booleano.

```python
vowels = ["a", "e", "i", "o", "u"]
"a" in vowels
"w" in vowels
```
