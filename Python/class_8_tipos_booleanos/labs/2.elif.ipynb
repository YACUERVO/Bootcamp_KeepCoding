## Ejercicio 1

En Renfe te encargan una función para determinar el precio de los billetes con la siguiente promoción:

* El billete de ida y vuelta a cualquier ciudad de España cuesta €0.1 por Km recorrido en el primer tramo.
* Si la distancia es superior a 600 Km, el precio por Km baja a €0.08.
* Si la distancia es inferior a 100 km, el coste por km es de €0.17
* Si la vuelta es 7 días o más después de la ida, el precio total tiene un descuento del 12%.

Crea la función `ticket_cost` que recibe: distancia en km al destino final y número de dias entre la ida y la vuelta. Debe de devolver el coste total.

* ¿Es una función total o parcial?

```python
def ticket_cost(distancia, dias):
    '''Funcion que calcula el costo de un boleto de viaje de ida y vuelta'''
    if distancia < 100:
        precio_km = 0.17
    elif distancia <= 600:
        precio_km = 0.10
    else:
        precio_km = 0.08

    costo_total = 2 * distancia * precio_km

    if dias >= 7:
        costo_total *= 0.88  # aplica 12% de descuento

    return round(costo_total, 2)
```

```python
ticket_cost(150, 10)
```

## Ejercicio 6

Crea la función `is_leap` que recibe un año y determina si un año es bisiesto.

Un año es bisiesto si es divisible entre 4, excepto aquellos (que siendo divisibles por 4)  son divisibles entre 100 pero no entre 400.

![](bisiestos.png)

* Si no es divisible por 4, no es bisiesto.
* Si es divisible por 4, puede serlo (hay excepciones)
* ¿Es un predicado?



```python
def is_leap(year):
    '''Determina si un año es bisiesto'''
    year_bisiest = False    
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        year_bisiest = True
    else:
        year_bisiest = False
    return year_bisiest
is_leap(2002)
```

# Ejercicio 7

Función que recibe una edad y la clasifica de la siguiene manera:  
* "Infancia" (0-12 años), 
* "Adolescencia" (13-17 años), 
* "Juventud" (18-29 años), 
* "Adultez" (30-59 años) 
* "Vejez" (60+ años)

* ¿Es una función parcial o total?
* ¿Es un predicado?

```python
def age_classification(age):
    '''Clasifica la edad en diferentes categorias'''
    if 0 < age <= 12:
        category = 'Infancia'
    elif 13 <= age <= 17:
        category = 'Adolescencia'
    elif 18 <= age <= 29:
        category = 'Juventud'
    elif 30 <= age <= 59:
        category = 'Adultez'
    else:
        category = 'Vejez'
    return category
```

```python
age_classification(60)
```

## Ejercicio 8

¿Recuerdas la función `bmi` que recibia una altura y peso y te devolvía un índice?

Crea la función `classify_bmi`que recibe un bmi y devuelve una cadena indicando en qué rango está:

* "Delgado"  : bmi < 18.5 
* "Saludable": 18.5 < bmi < 24.9
* "Gordito"  : 25 a 29.9
* "Gordo"    : 30 a 34.9
* "Obeso"    : 35 a 39.9
* "Jabba the Hutt" : más de 40




```python
def bmi(peso, altura):
    '''Calcula el índice de masa corporal (BMI)'''
    return peso / (altura ** 2)


def classify_bmi(bmi_value):
    '''Clasifica el índice de masa corporal (BMI)'''
    classification = ''
    if bmi_value < 18.5:
        classification = 'Delgado'
    elif 18.5 <= bmi_value < 24.9:
        classification = 'Saludable'
    elif 25 <= bmi_value < 29.9:
        classification =  'Gordito'
    elif 30 <= bmi_value < 34.9:
        classification = 'Gordo'
    elif 35 <= bmi_value < 39.9:
        classification = 'Obeso'
    else:
        classification = 'Jabba the Hutt'
    return classification
```

```python
x = bmi(84,1.80)
classify_bmi(x)
```
