### Cortocircuito del `and`

En la siguiente expresión, tenemos a varias sebexpresiones booleanas combinadas mediante un `and`:

`(4.5 == 5) and ("hola" == "mundo") and (3 > 2)`

La primera (`4.5 == 5`) ya es `False`. Teniendo en cuenta que con el `and` *si uno sólo de sus operandos es `False` el
resultado será `False`*, ¿tiene sentido malgastar el tiempo evaluando el resto de la expresión?

La respuesta es **no**. Python inmediatamente pára cuando se encuentra con un `and` donde hay un `False` y no evalúa nada más. 
A esto se le llama el *cortocircuito del `and`*.

Esto tiene consecuencias interesante. Por ejemplo, esta expresión *no va a causar un error*, puesto que el `1 / 0` jamás
se va a evaluar.

```python
(4 != 4) and (1 / 0) != 4
```
