#### Mandamiento 0 de Dijkstra 🛎️

**UNA FUNCIÓN SÓLO HA DE TENER UN RETURN Y ÉSTE HA DE SER LA ÚLTIMA LINEA**

El tener varios puntos de salida (varios `returns`) hace que la lógica sea más compleja y sobre todo que tienda a complicarse muy rápidamente cuando aumentan los casos dentro del `if` (cuando aumenta la cantidad y complejidad de los `elifs`).

A partir de ahora, tus funciones sólo deben de tener *un `return`, tiene que estar en la última linea*. Además, no puede estar indentado. 

Por ejemplo, estas funciones están mal:

```python 
def is_even(num):
    if num % 2 == 0:
        return True
```

Aunque está en la última linea, está dentro del *ámbito* de un `if`. ¿Qué pasa sin `num % 2 != 0`? ¿Qué devuelve esa función?

```python 
def is_even(num):
    if num % 2 == 0:
        return True
    else:
        return False
```

Hay dos salidas de la función, cuando sólo debe de haber una.

Ésta **está bien**

```python 
def is_even(num):
    result = False
    if num % 2 == 0:
        result = True
    return result # Un solo punto de salida
```

De un vistazo, puedes saber donde y cómo sale la función.


#### Mandamiento 1 de Dijkstra 🛎️

**DECLARA TODAS TUS VARIABLES AL PRINCIPIO DE LA FUNCIÓN**

Con esto evitamos errores tontos de ámbitos y de un vistazo podemos localizar todas las variables que hayas creado para la función.



#### Resultado

Siguiendo los Mandamientos de la Programación Estructurada de Dijkstra, la función quedaría así:

```Python
def passes_test(grade):

    Recibe una nota numérica y devuelve `True` si es igual o superior a 7 y `False`en caso contrario.
    No despierta la furia de Dijkstra.
    """
```

En una empresa quieren subir el sueldo a todos los empleados cuyo sueldo sea inferior a un cierto nivel.

Crea la función `new_salary` que recibe el sueldo actual. Si dicho sueldo es inferior o igual a €1000, le aplica un incremento del 8%. Si no, se queda como está.

`````python
def new_salary(salary):
    '''Calcular nuevo salario con aumento del 15%'''
    if salary <= 1000:
        new_salary = salary * 1.08  
    
    return round(new_salary,2)

new_salary(1000)
`````

`````python
def new_total(precio):
    '''calcular el total del precio total'''

    if precio >= 1400:
        precio = precio - (precio * 0.15) 
        
    return precio
`````

## Discusión Ejercicios 5 y 6


Supongamos que has implementado `is_even` de la siguiente manera:

```python
def is_even(num):
    if num % 2 == 0:
        return "par"
    else:
        return "impar"
````

Habrias cometido dos errores imperdonables, ambos muy comunes en los novatos.

**Primer error: faltarle al respeto a Dijkstra**

Tu función tiene dos returns, lo cual va en contra del *Mandamiento 0 de Dijkstra*. Por suerte es fácil de arreglar.

**Segundo error: tipo equivocado**

Esto último es mucho más grave y extremadamente común entre novatos: estás **devolviendo una cadena cuando deberías de devolver un `bool`**. Esto hace que tu función, aunque correcta, sea inútil.

La función encargada es un test, algo que tiene que decir si un número es par, *si o no*.  Por lo tanto, su tipo de retorno debería de ser un `bool`, `True` o `False`.

Cuando se empieza, es común abusar de las cadenas a la hora de representar valores para los cuales hay un tipo más adecuado. Es algo tan común, que hasta hay un nombre para ello: *stringly-typed code* (es un juego de palabras que entenderás más adelante).

**¿Por qué está mal y por qué tantos novatos lo hacen?**

Abusas de las cadenas para representar valores que no son texto, *porque sobrestimas a la máquina*. Cuando tú ves una cadena `"par"` eres capaz de entender su significado, pero ni Python ni el ordenador pueden hacerlo. Ambos sólo ven que esa función devuelve un texto.

Si haces que devuelva un `bool`, **tú y la máquina entendeis**. El lenguaje de programación sólo ve los tipos, y si Python ve que devuelve `bool` sabrá que estás contestando a una pregunta de tipo si/no.

Por último, está muy mal porque cualquier función que llame a esa versión incorrecta de `is_even` será incapaz de descifrar qué demonios devuelve.

La versión correcta de `is_even`, `is_odd` y `g` sería la siguiente:

```python
def is_even(num):
    return num % 2 == 0 # devuelve un booleano

def is_odd(num):
    return not is_even(num) # explica por qué esto funciona

def g(num):
    result = 0 # un valor cualquier que sobre-escribiremos
    if is_even(num):
        result = num / 2
    else:
        result = num - 1
    return result
```

Debido a que las dos primeras funciones devuelven el tipo correcto, son muy fáciles de reutilizar desde `g`.

**¡DISEÑA TUS FUNCIONES PARA QUE SEAN COMO LEGOS: SENCILLOS Y FACILES DE COMBINAR ENTRE SI!**





* implementación de par e impar que devuelven cadenas y por qué está mal

**Pista**

Es un poco más complicado de lo que parece, así que miremos los *Pilares de la Ciberkinesis*: el primero, **Divide y Vencerás**, no parece que nos pueda ayudar mucho. Miremos el siguiente: **Transforma y Vencerás**: cuando no puedas resolver un problema, ¡transfórmalo en uno que sí sepas resolver!

Veamos cómo podemos aplicarlo en este caso:

* No sabes comprobar si un número es un entero o no. Acéptalo.
* Sin embargo, sabes *convertir un número cualquiera en un entero*.

Con el *constructor* de `int`, puedes transformar cualquier número en un entero: `int(3.5)` e `int(53)` devolverán `3` y `53`, respectivamente.

¿Cómo podemos usar esto para comprobar si un número es un entero?

> Un número `n` será un entero si `int(n) == n`

¡Hemos *transformado* el problema de *comprobar si un número es un entero*  en el problema trivial de *comprobar la igualdad entre dos números*!
