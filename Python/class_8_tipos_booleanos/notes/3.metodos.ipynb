## Métodos: funciones *unidas* a un tipo 🛎️

Casi todos los tipos tienen *métodos*. Se trata de funciones que están unidas a dicho tipo y que para acceder a ellas, tenemos que hacer algo parecido a como cuando escribíamos `math.` para acceder a los que había dentro de ese módulo.

Por ejemplo, para llamar al método `upper` de `str`, usaríamos la siguiente sintaxis:

````python
# upper cambia todas las letras de una cadena a mayúsculas
"hola".upper()
````

### Métodos de `str`

Todos los tipos tienen métodos, pero no siempre son útiles en el día a día. `str`, las cadenas, sin embargo sí que tiene métodos bastante útiles. Veamos algunos.

![](metodos.png)

Veamos algunos de los más útiles:


1. **`islower()`** - Verifica si todos los caracteres *con caja* (que tienen versión en mayúscula y minúscula) en la cadena están en minúsculas. Retorna `True` si la cadena está vacía.
   ```python
   s = "hola mundo".islower()  # True
   ```

2. **`index(sub, start, end)`** - Busca `sub` en la cadena desde el índice `start` hasta `end`. Retorna el índice de la primera aparición de `sub`. Si `sub` no se encuentra, se lanza una excepción `ValueError`.
   ```python
   s = "hola mundo".index("mundo")  # 5
   ```

3. **`capitalize()`** - Convierte la primera letra de la cadena en mayúscula y el resto en minúsculas. Devuelve una nueva cadena y no altera la original.
   ```python
   s = "hola mundo".capitalize()  # 'Hola mundo'
   ```

4. **`endswith(suffix, start, end)`** - Verifica si la cadena termina con el sufijo `suffix`, considerando solo la parte de la cadena que va desde el índice `start` hasta `end`.
   ```python
   s = "hola mundo".endswith("mundo")  # True
   ```

5. **`startswith(prefix, start, end)`** - Verifica si la cadena comienza con el prefijo `prefix`, considerando solo la parte de la cadena que va desde el índice `start` hasta `end`.
   ```python
   s = "hola mundo".startswith("hola")  # True
   ```

6. **`isnumeric()`** - Verifica si la cadena contiene solo caracteres numéricos (incluidos dígitos que no son del alfabeto latino).
   ```python
   s = "12345".isnumeric()  # True
   s = "五百".isnumeric()  # True (números en chino)
   ```

7. **`upper()`** - Convierte todos los caracteres *con caja* de la cadena a mayúsculas. Devuelve una nueva cadena y no altera la original.
   ```python
   s = "hola mundo".upper()  # 'HOLA MUNDO'
   ```

8. **`lower()`** - Convierte todos los caracteres *con caja* de la cadena a minúsculas. Devuelve una nueva cadena y no altera la original.
   ```python
   s = "Hola Mundo".lower()  # 'hola mundo'
   ```
