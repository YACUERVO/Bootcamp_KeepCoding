# Constructores


------------

Ya sabemos que los tipos tienen instancias. Por ejemplo, `4`, `-12`, `34517` y `0` son todos instancias del tipo `int`. Para representar instancias de un tipo, normalmente existe un tipo de sintaxis para representar directamente a esa instancia. A esa sintaxis especial, se le llama una *representación literal*

## Representación literal de una instancia

La representación literal de una instancia de un tipo, es una sintaxis especial, que permite que Python sepa instantaneamente Cuando ve un texto *con esa pinta* (esa sintaxis) en nuestro programa, que se trata de una instancia de un cierto tipo.

Por ejemplo, 
* siempre que se encuentra con un número entero (`3`, `-341`, etc), ya sabe que es un `int`.
* cuando ve un numero decimal (`3.14`, `1.0`, ó `-0.00312`), ya sabe que se trata de un `float`.
* cuando un texto está entre comillas (sencillas o dobles), tenemos a una cadena: `"Python"`, `'Lambda'`.
* si vemos las palabras clave `True` or `False`, como son reservadas, sabemos que sólo pueden ser instancias de `bool`.


Si te fijas, estamos buscando *triquiñuelas* para poder representar las instancias de una forma sencilla y rápida. Sólo hay un problema, hay muchos más tipos de los que hemos visto hasta ahora, e incluso el programador pueder *crear sus propios tipos*, lo cual es muy común y lo veremos más adelante. 

Llegará un momento en que nos quedemos sin trucos, y habrá que buscar **una forma universal de crear y representar instancias**, que funcione con todos, sin depender de *atajos*.



## Constructor  🛎️

**Un constructor es una función que crea instancias de un tipo**. Por convención, el *constructor* tiene el mismo nombre que el tipo:


| Tipo | Función Constructor |
|-----------|-----------|
| `str`   | `str(...)`   |
| `int`   | `int(...)`   |
| `float`   | `float(...)`   |
| `bool`   | `bool(...)`   |
| Continuará   | Continuará   |


Podemos usarlo para crear instancias de un tipo, a partir de otros valores. veamos algunos ejemplos:


* `str(4)` se evalúa a la cadena `"4"`
* `float(8)` se evalúa a `8.0`
* `int(3.94)` se evalúa a `3`
* `float("3.14159")` se evalúa al float `3.14159`
  


### Convertir valores de un tipo a otro 🛎️

Cómo posiblemente te hayas dado cuenta ya, los **constructores también sirven para convertir valores de un tipo a otro**.  

Usando el constructor del tipo de destino, puedo convertir (si es posible) un valor de otro tipo a ese nuevo tipo.

Por ejemplo, 

* `float("3.14159")` convierte una cadena al float `3.14159`
* `str(3.14159)` convierte un float a la cadena `"3.14159"`
  
No siempre esa conversión será posible y hay que saber las reglas que sigue Python para determinar qué es posible y qué no. En el futuro, Cuando ya sepas crear tus propios tipos, también podrás determinar esas reglas para tus tipos, creando constructores.

El caso más común es el de los `bool`.  Para convertir un valore de los tipos que conocemos, se usan estas reglas:


1. **Números**:
   - `0` se convierte en `False`.
   - Cualquier otro número (positivo o negativo, entero o flotante) se convierte en `True`.

2. **Cadenas de texto (strings)**:
   - Una cadena vacía (`""`) se convierte en `False`.
   - Cualquier otra cadena, incluso un espacio en blanco (`" "`) se convierte en `True`.

3. **`None`**:
   - El valor especial `None` se convierte siempre en `False`.



### ¿Qué pasa  con `None`?

`None` es la única instancia del tipo `NoneType`. No se pueden crear nuevas isntancias de ese tipo (no hay diferentes *nadas*) y por lo tanto **no tiene constructores**.
