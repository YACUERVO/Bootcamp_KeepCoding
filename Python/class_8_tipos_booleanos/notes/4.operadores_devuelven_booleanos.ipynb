# Operadores que devuelven Booleanos

-------------------

Se trata de operadores, que aunque no tratan sólo con booleanos, sí que devuelven un valor booleano. 

Veremos los siguientes:


* **Igualdad o Equivalencia**: `==` y `!=`
* **Comparación**: `<`, `>`, `<=`, `>=`

## Igualdad o Equivalencia: `==` y `!=`


Ambos operadores actúan sobre valores de **cualquier tipo**: números, cadenas, lo que sea. 

#### `==`

El primero, `==`, se usa para comprobar
que dos valores son *iguales o equivalentes entre sí*. Para que dos valores sean equivalentes, lo primero que tienen que darse, 
es que sean del mismo tipo.


* `3 == 4` devuelve `False`  Son del mismo tipo, pero no son el mismo valor.
* `"Hola" == 'Hola'` devuelve `True` Son del mismo tipo y son el mismo valor.
* `4 == 'fantástico'` devuelve `False`, porque son de tipos distintos.
* `"Python" == "python"` devuelve `False`  ¿porqué?
* `51 == "51"` devuelve `False` porque aunque parecen iguales, ni siquiera son del mismo tipo.
  
Veamos ahora el caso de dos valores, dos numeros que tienen en el fondo el mismo valor, pero tiene tipos distintos:

* `67 == 67.0` devuelve `True`, ¡a pesar de que `67` es un `int` y `67.0` es un `float`!

Aquí Python ha metido una excepción a la regla de que los *tipos tienen que ser el mismo* para que el funcionamiento sea
más parecido a lo que un humano se espera: 3 y 3.0 son lo mismo. 

No todos los lenguajes tienen ese detalle contigo y para muchos, 3 y 3.0 son cosas totalmente distintas.

En el futuro, tú también podrás manipular esas reglas de igualdad *con los tipos que tú crees* (se pueden crear tipos nuevos y lo 
veremos más adelante).



> No lo confundas con el operador de asignación (`=`) que se utiliza para crear variables y cambiar su valor.



#### `!=`


Es justo el opuesto del anterior.

* `54 != 45` Devuelve `True`
* `"Hola" != "Hola"` Devuelve `False`



Ambos se usan mucho en las *expresiones booleanas*.

## Comparación: `<`, `>`, `<=`, `>=`


### Comparación con números


Funcionan igual que en la aritmética, comparando dos números (tanto si son `int` como si son `float`).

* `34.0 < 34.0` Devuelve `False`
* `-1 > -100` Devuelve `True`
* `-1 >= -100` Devuelve `True`

### Comparación de cadenas

¿Tiene sentido aplicar esos operadores sobre cadenas? Lo cierto es que sí, aunque no se usa tanto como con los números.

En ese caso se hace una **comparación lexicográfica** entre las cadenas, es decir, se averigua quién viene antes en el *orden alfabético*.

Veamos algunos ejemplos:

1. `"agua" > "barco"` devuelve `False` puesto que agua viene *antes* que barco y se considera *menor*.
2. `"algo" > "alabarda"` devuelve `True`, puesto que la segunda *a* en *alabarda* viene antes que la *g* de *algo*. La g es *mayor* que la a.
3. `"agua" > "aguardiente"` devuelve `False` puesto que si empiezan igual, pero una es más pequeña, se considera *mayor* a la más larga.

Tampoco tienes que obsesionarte mucho con esto, sin embargo hay una sorpresa de la que tienes que estar al tanto: el efecto de las mayúsculas.

**Mayúsculas y minúsculas**

Se considera que las *mayúsculas son menores que las minúsculas*:

* `"Agua" < "agua"` devuelve `True` puesto que la A es menor que la a.


**Caracteres acentuados** 

Temía que lo preguntases... Pues sí, también influye, pero es más dificil de explicar. En realidad, cada letra o símbolo (`v`, `45`, `💩`)
está codificado en el ordenador mediante un número siguiendo un sistema llamado Unicode. Ese número es el que se usa realmente para comparar
cadenas. Por lo general, los caracteres acentuados son considerados *mayores* que su equivalente no acentuado.

No le des más importancia y simplemente recuerda que Python usa estos operadores para comparar cadenas en un orden lexicográfico que
más o menos tiene sentido.
