# Variable Hoisting

------------------

El *variable hoisting* es un concepto de nombre estrafalario y consecuencias perniciosas que con toda seguridad ya te has encontrado, sin darte cuenta, en los ejercicios. Traducido al Español, significa *"Elevación de variables"* aunque a menudo se le llama *Hoisting de variables* en *Spanglish*.

Antes de explicar lo que es y los problemas que te puede causar, vamos a repasar las normas de visibilidad en los ámbitos.

## ¿Cómo se buscan las definiciones?

Cuando Python se encuentra una referencia a alguna definición (variable o función, tanto da), hace lo siguiente:

1. La busca en el ámbito actual.
   1. Si la encuentra, devuelve el valor
   2. Si no la encuentra, sigue buscando en el ámbito superior
2. Si al llegar al ámbito global, no la encuentra, devuelve un error.

## ¿Un error?

Viendo esa definición y sabiendo que `if`, `elif` y `else` crean ámbitos, ¿este código debería de dar un error?

```python
def new_salary(old_salary):

  """
  recibe el actual salario y devuelve ´true´ si es igual o inferior a 1000 y se aplica su 
  aumento(8%) y ´false´ se mantiene con su salario,en cambio si es inferior a 1200 
  se le aplica otro tipo de aumento(9%)
  """
                                # ámbito de la función
  if old_salary <= 1000:
    new_salary=old_salary*1.08  # ámbito del if
  elif old_salary < 1200:         # ámbito del elif
    new_salary=old_salary*1.09
  return new_salary             # ámbito de la función (de nuevo)
```

Cuando se ejecuta la función, ocurre lo siguiente:

1. Al llamar a la función, se entra en el ámbito de la misma: no hay ninguna definición.
2. Si se entra en el `if`, entramos en un ámbito local y se crea una definición de la variable `new_salary` y se le asigna un valor.
   1. Al salir del `if` se destruyen todas las definiciones hechas dentro de él, y se destruye la variable `new_salary`.
3. Si por el contrario, entramos en el `elif`, entramos en un ámbito local y se crea una definición de una nueva variable `new_salary` y se le asigna otro valor.
   1. Al salir, se destruyen todas las definiciones hechas dentro de él, y se destruye la variable `new_salary`.
4. Cuando llegamos a la última linea, la del `return` ¿existe una definición para la variable `new_salary`?

Con lo que sabemos, **¿esta función causará un error or devolverá un valor correcto?**

Lamentablemente, no provocará un error, sino que retornará correctamente.

## ¿Qué está pasando?

Ocurre que las normas para buscar definiciones que hemos visto, tienen una triste excepción y letra pequeña que no habíamos mencionado.


**Cuando se define una variable dentro del ámbito de un `if` (o de un `elif` o `else`), Python la "eleva" al ámbito inmediatamente superior**.

Es decir, cuando se encuentra con código como el de la función anterior, Python lo convierte en algo equivalente a este:

```python
def new_salary(old_salary):

  """
  recibe el actual salario y devuelve ´true´ si es igual o inferior a 1000 y se aplica su 
  aumento(8%) y ´false´ se mantiene con su salario,en cambio si es inferior a 1200 
  se le aplica otro tipo de aumento(9%)
  """
  new_salary = None             # "eleva" a new_salary del ámbito interno al superior
  if old_salary <= 1000:
    new_salary=old_salary*1.08  # ámbito del if
  elif old_salary < 1200:       # ámbito del elif
    new_salary=old_salary*1.09
  return new_salary             # ámbito de la función (de nuevo)
```

A esto se le llama *variable hoisting* o *elevación de variables*, aunque un nombre más adecuado sería *truño*.💩

## ¿Por qué el *variable hoisting* es malo?

Por que rompe la regla general sin venir a cuento y peor aun, introduce una forma de *magia* que hace que las declaraciones se muevan de lugar sin que tú lo hayas indicado.

Es un *rareza* de Python que no se da en otros lenguajes (la excepción sería Javascript) y no debes de acostumbrarte a ello ni usarlo, puesto que **fomenta malos hábitos**.

## ¿Cómo deberíamos implementar esa función?

Sin *variable hoisting* y siguiendo las normas que ya hemos dado, que precisamente fomentan buenos hábitos y el código claro y correcto.

```python

def new_salary(old_salary):

  """
  recibe el actual salario y devuelve ´true´ si es igual o inferior a 1000 y se aplica su 
  aumento(8%) y ´false´ se mantiene con su salario,en cambio si es inferior a 1200 
  se le aplica otro tipo de aumento(9%)
  """
  new_salary = old_salary       # un buen valor por defecto
  if old_salary <= 1000:
    new_salary=old_salary * 1.08  
  elif old_salary < 1200:       
    new_salary=old_salary * 1.09
  return new_salary            
```
