# El Hechicero

--------------------



![](despertar.png)

Despiertas y no sabes muy bien dónde estás. El dolor de cabeza es tan fuerte que te cuesta abrir los ojos. Poco a poco consigues ver dónde te encuentras: es una habitación de hotel y no tienes idea de cómo has llegado aquí.

Haces un esfuerzo enorme y algunos recuerdos empiezan a formarse: has estado en una despedida de soltero y la cosa debió de desmadrarse pero bien.

¿Pero despedida de soltero de quién? ¡Ostras, el novio eres tú! Llegar tarde no es una opción, y es ¡dentro de media hora!

Tu vida empieza a pasar delante de tus ojos cuando de repente ves la solución: **¡eres aprendiz de magia!** Eso podría ser la solución porque sabes varios *hechizos*.

Miras en tu libreta de notas y tienes varios guardados. Los tienes organizados en dos grupos:

* *hechizos puros*  
* *hechizos impuros*  

---

## *Hechizos Puros*

A partir de una cosa, genera otra.

* *musca llamus:* mosca → llama  
* *llama expressus:* llama → cafetera  
* *deloreanus ipso:* cafetera → DeLorean  

---

## *Hechizos Impuros*

Modifican algo existente.

* *resacon expurgo:* elimina inmediatamente los efectos de la resaca.  
* *elegantia máxima:* te deja superpintón, listo para una boda.  

---

Los primeros transforman una cosa en algo completamente nuevo. Es decir, crean algo que antes no existía, sin alterar lo que había antes.

Los segundos no crean nada: lo que hacen es modificar algo que ya existía. Te han dicho que estos últimos son más peligrosos y hay que usarlos con cuidado, pero aún no entiendes muy bien por qué.

Eso sí, tienes claro que no crean nada nuevo, sino que cambian lo existente.

---



## ¡La solución!


![](musca.png)

Si usas tus *hechizos* en el orden adecuado, llegarás a tu boda super elegante, sintiéndote fresco y recuperado, y además a tiempo. ¡Quien tiene un DeLorean jamás llega tarde!

Lo primero es conseguir el coche, y por suerte hay una mosca pegada a la ventana. La miras fijamente y gritas:

> **¡Musca Llama!**

Se produce un fogonazo y aparece una llama en medio de la habitación.

¡Esto marcha! piensas, y sin pensar más, das otro grito:

> **¡Deloreanus Ipso!**

![](error.png)

La cagaste.

Se produce una explosión tremenda que te lanza contra la pared. La pobre llama ha explotado y tú y toda la habitación estáis cubiertos de sangre, tripas y lana.

Desde el más allá se oye una voz con marcado acento holandés diciendo:

> “TypeError: Se esperaba un objeto de tipo Cafetera, pero se recibió Llama.”

¡Mierda! Las prisas, siempre las prisas... Si los *hechizos* fuesen funciones de Python, las firmas serían las siguientes:

```python
def musca_llamus(m: Mosca) -> Llama
def llama_expressus(l: Llama) -> Cafetera
def deloreanus_ipso(c: Cafetera) -> Delorean
```

Antes de usar `deloreanus_ipso`, ¡hay que tener una cafetera! No se pueden lanzar en cualquier orden.

La llama está por todas las paredes... ¿qué haces?

Por suerte, **¡son *hechizos puros*!** No alteran nada, así que la mosca sigue intacta. ¡Solo tienes que encontrarla!

Mientras la localizas vas pensando que eso de los *hechizos puros* sí que son útiles; sin ellos estarías bien jodido.  

El asunto es llamarlos en el **orden adecuado**.

`musca llama -> llama expressus -> deloreanus ipso`

La encuentras y le lanzas:

> **¡Musca Llama!**

Una nueva llama aparece en tu habitación con cara de sorpresa. Ahora sí que no te equivocas:

> **¡Llama Expressus!**

Ya tienes una cafetera *fashion* en el suelo de la habitación.  
Solo queda un paso:

> **¡Deloreanus Ipso!**

Un DeLorean alicatado hasta el techo, con regulador de fluzo y *Mr. Fusion* te espera.

Solo falta arreglar tus pintas. Aquí hay que tener cuidado, porque son *hechizos impuros* y **van a alterar** a quien se los dirijas, y resulta que ese alguien eres tú. No te equivoques y no lances el temido *Calvorium Totalus*, que la cagas.

Te plantas frente al espejo y sueltas:

> **¡Resacon Expurgo!**

Instantáneamente las ojeras desaparecen y te sientes mucho mejor, pero sigues con una pinta horrible.

> **¡Elegantia Maxima!**

En un instante pareces un galán de cine y sales disparado en tu DeLorean a **88 millas por hora**.

**¡La magia te ha salvado!**
