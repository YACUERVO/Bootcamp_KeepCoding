## Hechizos y funciones puras e impuras

La ciberkinesis es una forma de magia y sus hechizos son las funciones. 

Al igual que en la magia, hay *funciones puras* que crean algo nuevo sin alterar sus parámetros y funciones impuras que *producen cambios en cosas existentes*. Las funciones impuras pueden serlo porque:

* alteran sus parámetros
* modifican el ámbito
* modifican el estado del ordenador.

Estas últimas son más peligrosas, tal como hemos visto y hay que usarlas sólo cuando resulta indispensable. Normalmente son necesarias cunado interactuamos con algo externo a nuestro software (el usuario, la pantalla, un servidor web,  una base de datos, el sistema de ficheros, etc).

Lo correcto es usar siempre que sea posible, las funciones *puras*.

Veamos unos ejemplos de funciones puras:


`````python
def total(numbers:list)-> int|float:
    """
    Recibe una lista de números y devuelve su total.
    """
    total = 0
    for number in numbers:
        total = total + number
    return total
`````

`total` es una función pura porque:

1. No altera sus parámetros.
2. Si le pasas un mismo parámetro, siempre devolverá el mismo resultado.
3. El ámbito donde ha sido llamada no se ve alterado, por muchas veces que la llames.

Es parecida a `musca llama` y a `deloreanus ipso`: si le pasas el parámetro que se espera, hace su trabajo sin alterar nada a su alrededor. Es fiable y determinista.

Veamos algunos ejemplos de *funciones impuras*:

`````python
def quitar_cadenas(datos):
      """
      Recorre una lista y elimina todas las cadenas de ella
      """
      for i in range(len(datos) - 1, -1, -1):
          if type(datos[i]) is str:
              del datos[i]
      return datos
`````

`````python
import os

def ls(ruta:str)->list:
      """
      Devuelve una lista de nombres en el directorio `ruta`.
      """
      return os.listdir(ruta)
`````

* `quitar_cadenas` es una calamidad: destruye el parámetro que recibe, como si fuese suyo. 
* `ls` también es peligrosa, pero no está mal hecha. Realmente tiene que ser así. Devuelve una lista de cadenas con las carpetas y archivos que hay en una carpeta. Eso puede cambiar y no es culpa de nadie, así que si la llamas hoy y la llamas mañana, los resultados podrían ser diferentes. Es un caso de una función que interactúa con el *mundo exterior*. 

Por último, otro ejemplo de función impura:

`````python
def saludo(nombre:str)->None:
    print(f"Hola {nombre}!")
`````

* `saludo` es innecesariamente impura. Cambia el estado de la pantalla. A veces es indispensable, cuando hay que comunicarse con el mundo exterior y el usuario. Sin embargo si su propósito es *crear una cadena de saludo*, entonces es una calamidad, puesto que no crea nada y altera el estado de la pantalla.
* Suponiendo que el propósito es *crear un saludo*, ¿cómo la modificarías para que fuese pura?

# El orden importa

Hemos visto que en los hechizos el orden es vital. ¡No puedes llamar `musca llama` si no tienes un mosca!

`musca llama -> llama expresus -> deloreanus ipso`

es el orden adecuado, puesto que cada hechizo se espera un *parámetro* de un cierto tipo, sino terminas cubierto de sangre y tripas de llama.

Si fuesen funciones de Python, tendríamos que llamarlas de la siguiente manera:

```python
llama = musca_llama(mosca)
cafetera = llama_expressus(llama)
dmc = deloreanus_ipso(cafetera)
```

Aunque también se podría hacer así:

```python
dmc = deloreanus_ipso(llama_expressus(musca_llama(mosca)))
````

Asegúrate de entender que ambas hacen lo mismo, de una forma un poco distinta.



# Si el orden importa, una función depende de la otra

Para que esto funcione, `llama_expresus` tiene que confiar en `musca_llama`. Tiene que saber que hará su trabajo, y pase lo que pase, devolverá siempre una llama. Si dependiendo de la fase de la luna, va y devuelve un rinoceronte o un orinal, vamos dados.

La situación de `deloreanus_ipso` no es distinta. Tiene que confiar en los que vienen antes que ella. Tiene que saber que las funciones que han pasado antes que ella, han hecho bien su trabajo y no han dejado "cáscaras de plátano" en las que pueda resbalar. 


# Cuidar el porvenir

Para que toda función pueda hacer su trabajo, es necesario que las demás hagan bien su trabajo, sean fiables, y cuiden en el entorno de cara *al porvenir*.

Cuando vas a un aseo público, te aseguras de dejarlo limpio por los que vengan depsués. Es de civismo elemental.

Lo mismo ocurre en la Ciberkinesis: toda función, después de *"hacer lo suyo"*, tiene que asegurarse de dejar todo como estaba, y que si promete devolver un cierto tipo, que eso siempre ocurra. Siempre.
